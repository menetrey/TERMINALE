## I) Rappel sur les probabilités
### 1) Définition
<div class="alert alert-blok alert-danger">
    <b>Espace de calcul de probabilités discrètes</b><br><br>
<b>Expérience aléatoire </b>: Protocole précis qui vérifie des conditions qui sont
non prévisible, d’issues possibles connues et renouvelable dans des conditions
identiques.  <br>
    
À partir d’une expérience aléatoire, on définit l’espace probabilisé par :<br>
    
 - L’<b>univers $\Omega$ </b>: Ensemble (discret et fini) des issues possibles  $\Omega = \{e_1, e_2,\ldots , e_n\}$<br>
 - Les <b>parties de $\Omega$</b>, noté $P(\Omega)$.  <br>
 Les éléments de $P(\Omega)$ sont appelé événements. Les événements élémentaires sont les singletons de $P(\Omega)$
 - L’événement <b>contraire de $A$</b>, noté $\overline{A}$ est le complémentaire de A dans $\Omega$.<br><br>
    
Deux événements A et B sont <b>incompatibles</b> si $A \cap B = \emptyset $.
L’événement certain est $\Omega$ et l’événement impossible est l’ensemble vide, $\emptyset$.<br><br>
    
<b>Une loi de probabilité $p$ </b>: fonction définie de $P(\Omega)$ dans $[0 ; 1]$ qui vérifie les conditions suivantes :
- $p(\Omega)$ = 1
- additivité : si A et B sont incompatibles alors $p(A \cup B) = p(A) + p(B)$
    </div>  
    
**Exemples**:  
1) On lance un dé cubique : $\Omega = \{1, 2, 3, 4, 5, 6\}$  
- Soit les événements  A: « obtenir un nombre impair » et B :« obtenir un 6 ».  
$A = \{1, 3, 5\}$ et $\overline{A} = \{2, 4, 6\}$.
$B = \{6\}$ est un événement élémentaire.  

A et B sont incompatibles, car $A \cap B = \emptyset$, et donc $p(A \cup B) = p(A) + p(B$).  

2) On tire successivement, sans remise, deux boules d’une urne qui en contient 6 :   
- $\Omega = \{(b_i , b_j), i, j \in [[ 1, 6 ]] $ et $i \neq j \}$. Il y a $6 \times 5 = 30$ issues   

## 2) Loi de probabilité    
 <div class="alert alert-block alert-danger">
    <b>Théorème 1 :</b>  <br>
Soit $p$ une loi de probabilité, on a alors :  
    
- $$ p(e_1) + p(e_2) + \ldots + p(e_n) = \sum_{i=1}^{n} p(e_i) = 1$$ <br>  
-  $p(\emptyset) = 0$  <br><br>
    
Pour tous événements A et B, on a les relations :  <br>
- $p(\overline{A}) = 1 − P(A)$  <br><br>
- $p(A \cup B) = p(A) + p(B) − p(A \cap B)$<br>
 </div>  
 
**Exemples :**  
1) On lance un dé truqué. Après un relevé statistique, on a pu déterminer que les probabilités d’apparition de chaque face sont telles que :  

$p(1) = p(2) = p(3) = p(4) = p(5)$ et $p(6) = 3p(1)$  
Déterminer la loi de probabilité p.     
  
  
  
2) On donne : $p(A) = 0,3$ , $p(A \cup B) = 0,7$ et $p(A \cap B) = 0,2$.   Calculer $p(B)$  


### 3) Loi équiprobable
<div class="alert alert-block alert-danger">
<b>Théorème 2 : </b>    <br>
Une loi de probabilité est équiprobable si chaque événement élémentaire a la même probabilité d’apparition.<br>  

Si $\Omega$ est composé de n éléments alors pour tout $i \in [[1, n]]$, $p(e_i) = \dfrac{1}{n}$ <br> 
et pour tout événement A on a :   
   $$ p(A) = \dfrac{\mbox{nombre de cas favorables}}{\mbox{nombre de cas possibles}}$$.  <br>
</div>   

### 4) Variable aléatoire
<div class="alert alert-block alert-danger">
    <b>Définition 2 :</b><br>
 Une variable aléatoire $X$ est une fonction de $\Omega$ dans $\mathbb{R}$, qui à chaque issue $e_i$ associe un réel $x_i$.<br>
Une loi de probabilité $p$ de $X$ est une fonction de $\mathbb{R}$ dans $[0; 1]$ qui à $x_i$ associe la probabilité $p(X=x_i) = p_i$.<br>
    <b>L’espérance</b> mathématique de $X$ est le réel noté $E(X)$
définie par :<br>
$$E(X) = \sum_{i=1}^{n} p_ix_i = p_1x_1 + p_2x_2 + \ldots + p_nx_n$$  <br>
 La <b>variance</b> et <b>l’écart-type</b> de $X$ sont les réels notés $V(X)$ et $\sigma(X)$ définies par : <br>
   $$ V(X) = \sum_{i=1}^{n} p_ix_i^2 - E^2(X) \quad \mbox{ et } \quad \sigma(X)=\sqrt{V(X)}$$

 
</div>    

**Remarque :**  
L’espérance mathématique correspond à une moyenne des valeurs $x_i$ , pondérées par les probabilités de la loi $p$ définie sur $X$. Si $X$ représente le gain pour un jeu, $E(X)$ représente le gain moyen que peut espérer le joueur.  
Pour le joueur :  
$E(X) \gt 0$ : jeu favorable,  
$E(X) \lt 0 $: jeu défavorable,  
$E(X) = 0$ : jeu équitable.  
  
### 5) Probabilité conditionnelle
<div class="alert alert-block alert-danger">  
    <b>Définition 3 :</b>  <br>
La probabilité de $B$ sachant que $A$ est réalisé, notée $p_A(B)$ vaut :  
    
$$p(A) \neq 0 \quad \mbox{ et } \quad p_A(B) =\dfrac{p(A \cap B)}{p(A)}$$
</div>


### 6) Probabilité totale
<div class="alert alert-block alert-danger" >
    <b> Théorème 3 : Probabilités totales</b><br> 
    
Soit $A_1, A_2, \ldots, A_n$  une partition de l’univers $\Omega$ alors, pour tout événement B :<br>
$$P(B) = P(A_1 \cap B) + P(A_2 \cap B) + \ldots + P(A_n \cap B)$$
 </div>     
 
 ### 7) Événements indépendants
<b>Définition 5 :</b><br>  
Deux événements A et B sont indépendants si et seulement si :<br>
$$p(A \cap B) = p(A) \times p(B) \quad \mbox{ou} \quad p_A(B) = p(B)$$
</div>

**Exemple :**  
Un appareil ménager peut présenter après sa fabrication deux types
de défauts : un défaut d’apparence (A) et un défaut de fonctionnement (F).  
La probabilité que l’appareil présente un défaut d’apparence vaut 0, 02 et la probabilité que l’appareil présente au moins l’un des deux défauts vaut 0, 069.  

On suppose que les deux types de défauts sont indépendants.
On choisit un des appareils.  
Quelle est la probabilité que l’appareil présente le défaut F?

Soit une succession de $n$ épreuves indépendantes dont les univers associés sont respectivement $\Omega_1, \Omega_2, \ldots , \Omega_n$.<br>

### 2) Probabilité d’une succession d’épreuves indépendantes
<div class="alert alert-block alert-danger">

    Soit $x = (x_1, x_2, \ldots , x_n)$  une issue possible alors :  
    $$p(x) = p_1(x_1) \times p_2(x_2) \times \ldots \times p_n(x_n)$$<br>
    
 </div>   


### 3) Épreuve, loi et schéma de Bernoulli
<div class="alert alert-block alert-danger">  
    <b>Définition 7 : Épreuve de Bernoulli</b><br>
Une épreuve de Bernoulli est une expérience aléatoire qui admet exactement
deux issues : succès ou échec. 
</div>  

**Exemple :**  
On lance un dé à 6 faces et l’on appelle succès « obtenir un six » et échec le cas contraire.  

<div class="alert alert-block alert-danger"> 
    <b>Théorème 5 : Loi de Bernoulli</b><br>
Soit $X$ la variable aléatoire associée à une épreuve de Bernoulli qui prend la valeur 1 pour un succès et 0 pour un échec. <br>  
 Soit p la probabilité de succès.<br>  
    <br>
La loi de probabilité de $X$ est appelée loi de Bernoulli :<br>
    
Elle est notée $\mathcal{B}(p)$.  <br>

 <br><br>   
On a alors : $E(X) = p$ , $V(X) = p(1 − p)$ et $\sigma(X) = \sqrt{p(1 − p)}$  
 </div>    
 
**Démonstration**:  
<br><br><br><br>

<div class="alert alert-block alert-danger"> 
<b>Définition 8 : Schéma de Bernoulli</b><br>
La répétition de $n$ épreuves de Bernoulli identiques et indépendantes est appelée schéma de Bernoulli d’ordre $n$ et de paramètre $p$.
</div>    

```python
from random import randint

def droite():
    D = 0
    for i in range(3):
        if randint(0,1)== 1:
            D = D+1
    return D        
            

def galton(n):
    L =[]
    for i in range(4):
        L.append(0)
    for k in range(n):
        D = droite()
        L[D] = L[D] + 1
    for D in range(4):
        L[D]=L[D]/n
    return L    
```

```python
galton(100000)
```

#### b. Propriétés des coefficients binomiaux
<div class="alert alert-block alert-danger">
    
<b>Théorème 5 :</b><br>  
Pour tous entiers naturels n et p, tels que $0 \lt p \lt n$, on a :  <br>
    $$\binom{n}{0} = \binom{n}{n}= 1$$<br>
    $$\binom{n}{1}=\binom{n}{n-1}= n$$<br>  
    $$\binom{n}{n-p}=\binom{n}{p}$$<br><br>

<b>Relation de Pascal :</b><br>
    $$\binom{n}{p}+ \binom{n}{p+1} =\binom{n+1}{p+1}$$
</div>   

**Exemple :**  
$\binom{7}{0}=\binom{7}{7}=1$

$\binom{6}{4}=\binom{6}{2}=15$  

  
**Remarque :**

# IV) Loi binomiale
## 1) Loi binomiale de paramètre n et p 
<div class="alert alert-block alert-danger">
    <b>Théorème 6 : Loi binomiale</b><br>
Dans un schéma de Bernoulli d’ordre n et de paramètre p, la loi de probabilité de la variable X qui à chaque issue associe le nombre k de succès est définie par :  <br>
    
$$p(X = k) = \binom{n}{k}p^k(1 − p)^{n−k}$$ <br>
    
On dit alors que la variable aléatoire $X$ suit une loi binomiale $\mathcal{B}(n, p)$
</div>  

**Exemple :**  
On lance 10 fois de suite un dé cubique.  
Quelle est la probabilité d’obtenir :
1) exactement 3 fois un six?  
2) au moins 2 fois un six?  
3) entre 2 et 5 fois un six?
