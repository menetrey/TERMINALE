En mathématiques, il est nécessaire d'utiliser un langage logique et de s'appuyer sur des raisonnements pour réussir une démonstration claire et concise.    
Une rédaction ne s’improvise pas, il faut s’y être préparé, car elle mêle
des automatismes qui ne s’acquièrent que par la pratique et des définitions et
théorèmes qu’il faut savoir citer au bon moment et précisément.
  
Voici quelques indications pour améliorer votre rédaction et apprendre quelques
automatismes qu’il est bon de connaître.
D’une façon générale, la rédaction d’une question doit comporter trois parties :  
• L’introduction  
• Le raisonnement  
• La conclusion  
  
  
# I) Introduire ce dont on parle
Introduire toutes les variables utilisées, même si elles sont définies dans l’énoncé.  

Par exemple pour introduire un entier naturel non nul quelconque, on peut écrire :  

• Soit $n \in \mathbb{N^∗}$  
• Pour tout $n \in \mathbb{N^∗}$

Cette étape nécessaire nous permet de contextualiser.   
Par exemple, si vous écrivez hors de tout contexte :  
"Ils sont colinéaires" , la première question que l'on se pose est : Qui ça "les vecteurs", quels vecteurs ?  


# II) Mettre en évidence les articulations logiques
Quelques petits mots bien utile dans la rédaction :

«   

• or, on sait que, de plus, en outre, ensuite, enfin,     
• mais, cependant, toutefois, puisque, comme, car,    
• donc, alors, il vient, d’où, par conséquent, ainsi,   
• . . .     
»

Ces petits mots vous permettent de mettre du liant dans votre raisonnement et
rend la lecture plus claire. Attention toutefois à la signification logique de ces petits mots, ils ont en effet une implication dans votre raisonnement.  

**Exemple :**  

Montrer que : Pour tout $ x \in [0 ; 1], \sqrt{1 − x^2} \in [0 ; 1]$.  

  
Soit $x \in [0 ; 1]$  

• Par croissance de la fonction carrée sur $\mathbb{R^+}$, on a : $0 \lt x^2 \lt 1$  en conséquence $0 \lt 1 − x^2 \lt 1$  


• Par croissance de la fonction racine carrée sur $\mathbb{R^+}$ : $0 \lt \sqrt{1 − x^2} \lt 1$  

En conséquence, pour tout  $ x \in [0 ; 1], \sqrt{1 − x^2} \in [0 ; 1]$.

# III) Annoncer ce que l’on fait
Rédiger correctement une question en mathématiques, c’est aussi expliquer ce que l’on fait. Annoncer la méthode de résolution au début de votre réponse :   

"Montrons que. . .",  
"Démontrons par récurrence . . .",  
"Montrons par l’absurde que . . .",  
"Il ne reste plus qu’à montrer que. . .", etc.   

Votre travail n’en sera que plus compréhensible.

# IV) Citer une définition ou un théorème
Citer une définition ou un théorème doit se faire avec précision . Il faut donner clairement et sans faute les hypothèses, les notations et la conclusion. Un théorème mal rédigé, imprécis, une hypothèse omise tout cela donne une impression de manque de rigueur et peut mener à une conclusion erronée!    


**Exemple :**  
Soit la fonction $f$ définie sur $\mathbb{R}$ par : $f (x) = x^3 + x −1$.  
Montrer que l’équation $f (x) = 0$ admet une unique solution sur $\mathbb{R}$.  

**Réponse incorrecte :**  
$f (0) = −1$ et $f (1) = 1$ donc la fonction f change de signe, d’après le théorème des valeurs intermédiaires, l’équation $f (x) = 0$ admet une
unique solution $\alpha$ sur $\mathbb{R}$.  

Quelles sont les hypothèses du théorème des valeurs intermédiaires ? Pourquoi
cette solution est-elle unique ?  

**Réponse correcte :**   
La fonction cube et la fonction affine $x \to x − 1$ sont deux
fonctions définies et croissantes sur $\mathbb{R}$.  

La fonction $f$ est continue sur $\mathbb{R}$ car $f$ est un polynôme.  
$f (0) = −1$ et $f (1) = 1$ donc la fonction $f$ change de signe sur $\mathbb{R}$.  
La fonction $f$ est **continue, monotone, et change de signe** sur $\mathbb{R}$ donc d’après le théorème des valeurs intermédiaires, l’équation $f (x) = 0$ admet une unique solution $\alpha \in [0 ; 1]$ sur $\mathbb{R}$.  


# V) Faire la différence entre f et f(x)
**Rédaction incorrecte :**  
"La fonction $\dfrac{x}{ x^2 + 1}$ est dérivable sur $\mathbb{R}$"


**Rédaction correcte :**   
"La fonction $x \to \dfrac{x}{ x^2 + 1}$ est dérivable sur $\mathbb{R}$"

En effet $\dfrac{x}{ x^2 + 1}$ n’est pas une fonction mais une **expression algébrique**.

Une fonction est une relation qui à une quantité $x$ appelée variable associe la quantité $f(x)$. On l’a note alors : $x \to f (x)$.  


**Remarque :**
Parfois la fonction a un nom comme $f$ , $g$, fonctions carrée, cube et
racine carrée, exp, ln, cos, sin. On pourra alors écrire "La fonction exp est croissante sur $\mathbb{R}"$.

# VI) Implication ou une équivalence
## 1) Montrer une implication
Quand on veut montrer que $p \implies q$, on procède par l’un des deux procédés
suivant :  
<div class="alert alert-block alert-danger">
 On suppose que p est vraie et l’on montre <b>qu’alors</b> q est vraie.  
</div>  

**Exemple :**  

Si $A$ et $B$ sont deux événements indépendants alors il en est de même
pour $\overline{A}$ et $B$.  
Soient $A$ et $B$ deux événements indépendants alors : 
$$P(A ∩ B) = P(A)\times P(B)$$
  
D’après les probabilités totales :
$$\begin{array}\\
P(B) &=& P(A ∩ B) + P(\overline{A} ∩ B) \\
&=& P(A)P(B) + P(\overline{A} ∩ B) \quad (1)
\end{array}$$  

D’après la probabilité de l’événement contraire : $P(A) = 1 − P(\overline{A}) \quad (2)$  

En remplaçant (2) dans (1) :  
$$\begin{array}\\
P(B) &=& [1 − P(\overline{A})]P(B) + P(\overline{A} ∩ B) \\
&=& P(B) − P(\overline{A})P(B) + P(\overline{A} ∩ B)
\end{array}$$  

On a alors : $P(\overline{A} ∩ B) = P(\overline{A})P(B)$.  
Les événements $\overline{A}$ et $B$ sont indépendants.  

    
**Par la contraposée.**
<div class="alert alert-block alert-danger">
On suppose que <b>non q</b> est vraie et l’on montre qu’alors <b>non p</b>
est vraie 
</div>


**Exemple :**

Montrer que 109 est un nombre premier.    

Test de primalité :   
si un entier $n$, avec $n \ge 2$, n’est pas premier alors il admet un diviseur premier $p$ tel que $2 \le p \le \sqrt{n}$.    

On vérifie que 109 n’est pas divisible par 2, 3, 5, 7. De plus $10 \le \sqrt{109} \le 11$.   

109 n’admet pas de diviseur premier $p$ tel que $2 \le p \le \sqrt{109}$, d’après la contraposée du test de primalité, 109 est premier.    



**Remarque :**  
La contraposée du test de primalité peut s’énoncer ainsi :  
« si un entier $n$, avec $n \ge 2$ n’admet pas de diviseur premier $p$ tel que $2 \le p \le \sqrt{n}$ alors $n$ est premier »

## 2)  Montrer qu’une implication est fausse
<div class="alert alert-block alert-danger">
Pour montrer que l’implication $p \implies q$ est fausse, il suffit de trouver un contre-exemple où la proposition $p$ est vraie et la proposition $q$ fausse.
</div>


**Exemple :**  
Soit la proposition :  
« La suite $(U_n)$ est croissante donc la suite $(U_n)$ est divergente vers $+\infty$»  

La proposition est bien sûr fausse.  
Il faut donc trouver un contre-exemple d’une suite qui est croissante et non divergente vers $+\infty$ i.e. convergente.  

Soit la suite $(U_n)$ définie sur $\mathbb{N^∗}$ par :  
$$U_n = 1 − \dfrac{1}{n}$$  

  
• La fonction inverse étant décroissante sur $]0 ;+\infty[$, la fonction $x \to −\dfrac{1}{x}$ est croissante sur $]0 ;+\infty[$. Il en résulte que la suite $(U_n)$ est croissante.    

• $\lim \limits_{n\to -\infty} \dfrac{1}{n}= 0$, par somme $\lim \limits_{n \to +\infty} U_n = 1$.  
La suite $(U_n)$ est donc convergente vers 1.  

                      
                      
Conclusion :  
On a trouvé une suite croissante convergente, donc la proposition
est fausse.  


**Remarque :**  
La locution latine « i.e. » signifie « id. est » soit « c’est à dire ». Cette
locution est fréquemment utilisée dans la rédaction mathématique.  


## 3)  Montrer une équivalence
<div class="alert alert-block alert-danger">
Pour montrer que : $p \iff q$, on peut procéder de deux façons :<br>
    
• Soit on raisonne par équivalence, comme c’est le cas dans la résolution d’équations.<br>
• Soit on raisonne par double implication lorsque le raisonnement par équivalence s’avère périlleux :  <br>
$\quad$ On suppose que p est vraie et l’on montre alors que q est vraie.<br>
$\quad$ Réciproquement, on suppose que q est vraie et l’onmontre alors que p est vraie.
</div>
                      
**Exemple :**  
Montrer le théorème de Bézout à l’aide de l’identité de Bézout :  
$$pgcd(a, b) = 1 \iff \exists(u, v) \in \mathbb{Z^2}, au + bv = 1$$

Raisonner par équivalence, s’avère ici impossible car l’identité de Bézout est une implication. On raisonne alors par double implication.    

Rappelons l’identité de Bézout :  
$$pgcd(a, b) = d \implies \exists(u, v) \in \mathbb{Z^2}, au + bv = d$$
• Supposons que $pgcd(a, b) = 1$, d’après l’identité de Bézout, il existe $u, v \in \mathbb{Z}$, tels que $au + bv = 1$.  

• Réciproquement, $a$ et $b$ étant deux entiers, supposons qu’il existe $u, v \in \mathbb{Z}$, tels que $au + bv = 1$.  
Soit $d$ le $pgcd(a, b)$, $d$ divise $a$ et $b$ donc divise toute combinaison linéaire de $a$ et de $b$ soit $au + bv$.  
En conséquence $d$ divise $1$, donc $d = 1$.  

                      
# V)  Le raisonnement par l’absurde
<div class="alert alert-block alert-danger">
Quand on veut montrer qu’une propriété $p$ est vraie, on peut raisonner par l’absurde, c’est-à-dire supposer $p$ fausse et arriver à une contradiction. 
</div>


**Exemple :** Montrer l’irrationalité de $\sqrt{2}$  

Supposons que $\sqrt{2}$ est un nombre rationnel. Il existe donc deux entiers $p$ et $q$ premiers entre eux tels que : $\sqrt{2} =\dfrac{p}{q}$  

On élève au carré, on obtient alors :   
$$\dfrac{p^2}{q^2} = 2 \iff p^2 = 2q^2$$  

On en déduit que $p^2$ est pair. Comme un nombre et son carré ont même parité, $p$ est pair. On peut donc écrire : $p = 2p′$  
On a alors $p^2 = 2q^2 \iff 4p′^2 = 2q^2 \iff q^2 = 2p′^2$.   
On en déduit que $q^2$ est pair et par suite $q$ est pair.  

$p$ et $q$ sont pairs. Il ne sont donc pas premiers entre eux. Contradiction  

Conclusion : $\sqrt{2}$ n’est pas un nombre rationnel.

La rédaction par récurrence doit être précise faute de quoi sa validité peut être contestée.  
Le raisonnement par récurrence obéit au principe suivant :  
 
<div class="alert alert-block alert-danger">
Si $P_0$ est vraie et si :  pour tout $n \in \mathbb{N}$, $P_n \implies P_{n+1}$, alors : pour tout $n \in \mathbb{N}$, $P_n$ est vraie.  
</div>  


# VII) Analyse-synthèse

Parfois pour déterminer les éléments d’un ensemble qui vérifie une propriété
ou pour trouver une accroche afin de commencer une démonstration, on peut
raisonner par analyse-synthèse.  

Ce raisonnement se fait en deux parties :
<div class="alert alert-block alert-danger">
• L’analyse : <br>
On suppose qu’il existe un élément $x$ d’un ensemble $E$ qui vérifie la
propriété $P$. Par implication, on détermine un critère que doit vérifier $x$. On réduit « le champs des possibles » pour $x$. Il peut même arriver qu’il n’existe qu’une solution possible.<br>
• La synthèse :<br>
On vérifie alors les valeurs de $x$, parmi celles déterminées pendant
l’analyse, qui vérifient la propriété $P$.  
</div>


**Exemple :**  
Résoudre dans $\mathbb{N}$, l’équation : $x(x + 1)(2x + 1) = 84$.  

• Analyse :   
Soit $x$ une solution de l’équation. D’après la factorisation, $x$ et $x + 1$
sont des diviseurs de 84. Les solutions de l’équation sont donc à chercher parmi les entiers consécutifs diviseurs de 84.  

$D_{84} = {1 , 2 , 3 , 4 , 6 , 7 , 12 , 14 , 21 , 28 , 42 , 84}$  

Les valeurs de $x$ possibles sont donc : 1, 2, 3 et 6.  

• Synthèse :  
On vérifie parmi ces valeurs celles qui sont solutions de l’équation :

|x| 1| 2| 3| 6|
|:---:|---|---|---|---|
|x(x + 1)(2x + 1) | 6 | 30 | 84 | 546 |


L’équation admet une unique solution : $x = 3$
