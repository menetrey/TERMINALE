# I)  Langages des ensembles
## 1) Ensemble

<div class="alert alert-block alert-danger">
<b>Définir un ensemble</b><br>
Un ensemble est une collection d’éléments que l’on peut énumérer ou définir par une propriété.   <br>
    
Certains ensembles ont des notations particulières (ex. $\mathbb{N}, \mathbb{Z}, \mathbb{Q}, \mathbb{R}, \mathbb{C}$). <br> 
    
Un ensemble est défini par extension si l’on énumère tous ses éléments :
$$E = \{a , b , c , \ldots \}$$<br>
Un ensemble est défini par compréhension si ses éléments sont définis par une propriété.<br>
$$E = {x \in A, P(x)}$$
qui se lit : « E est l’ensemble des éléments de A vérifiant la propriété P ».<br><br>
L’ensemble ne contenant aucun élément s’appelle : l’ensemble vide noté «$\emptyset$».<br>
 </div>  
 
**Exemples :**  
1) Ensembles définis par extension :  
• $\{1, 3, 5, 7, 9\}$ ensemble des chiffres impairs.  
• $\{b_1, b_2, \ldots , b_n\} = \{b_i\}_{1\le i \le n}$ ensemble de $n$ boules.    

2) Ensembles définis par compréhension :  
• ${n \in \mathbb{N}, 1 \le n \le 49}$ ensemble des nombres d’une grille de loto  
• $2\mathbb{N} =\big \{ x \in \mathbb{N}, \dfrac{x}{2} \in \mathbb{N} \big \}$ ensemble des entiers naturels pairs.  

**Remarque :**   
Dans un ensemble les éléments sont non ordonnés et distincts.
Un ensemble contenant qu’un élément s’appelle un singleton , deux éléments
une paire, trois éléments un triplet.  

## 2) Appartenance, inclusion, parties d’un ensemble
<div class="alert alert-block alert-danger">
<b>Définition  :</b><br>  
Soient E et F deux ensembles avec E non vide .  <br>
• Si un élément $a$ appartient à $E$, on écrit : $a \in E$<br>
• Si F est inclus dans E, alors F est un sous-ensemble ou une partie de E, soit :<br>
$$F \subset E \iff \forall a \in F, a \in E \mbox{ ou } F = \emptyset$$  
    
• L’ensemble des parties de $E$ est l’ensemble, noté, $\mathcal{P}(E)$, constitué de tous les sous-ensembles de E.  
 </div>
 
**Remarque :**  
Une partie est un autre nom pour sous-ensemble.
- Ne pas confondre les symboles $\in$ et $\subset$ qui sont très proche.  

  
**Exemples :**  
• On a la suite des inclusions suivantes : $\mathbb{N} \subset \mathbb{Z} \subset \mathbb{Q } \subset \mathbb{R} \subset \mathbb{C}$.  
• Soit l’ensemble $E = \{a, b, c\}$. L’ensemble des parties de $E$ est :
$\mathcal{P}(E) = \{\emptyset , \{a\} , \{b\} , \{c\} , \{a, b\} , \{a, c\} , \{b, c\} , E\}$
    
 
 
  
## 3) Opérations sur les ensembles
<div class="alert alert-block alert-danger">  
<b>Définition  :</b><br>  
On définit trois opérations élémentaires dans $\mathcal{P}(E)$ :<br>
• Complémentaire de $A$ dans $E$ : $\overline{A}$<br>
$x \in \overline{A} \iff x \in E \mbox{ et } x \notin A$<br>
    
• Intersection de $A$ et $B$ : $A \cap B$<br>
$x \in A \cap B \iff x \in A \mbox{ et } x \in B$<br>
    
• Union de $A$ et $B$ : $A \cup B$<br>
$x \in A \cup B \iff x \in A \mbox{ ou } x \in B$<br>
    
</div>


**Remarque :**  
Quelques cas particuliers.  

• On dit que $A$ et $B$ sont disjoints si et seulement si $A \cap B = \emptyset$.  
• Si $A \cap B = B$ alors, $B \subset A$.  
• Si $A \cup B = B$ alors, $A \subset B$.  

## 4) Produit cartésien, p-uplet et p-liste
<div class="alert alert-block alert-danger"> 
<b>Définition  :</b><br>  
Le produit cartésien de deux ensembles $E$ et $F$, noté $E \times F$, est
l’ensemble des couples $(x, y)$ tels que $x \in E$ et $y \in F$.<br>
On généralise le produit cartésien à $p$ ensembles : $E_1 \times E_2 \times \ldots \times E_p$ où les éléments sont des p-uplets $(x_1, x_2, \ldots , x_p)$  tel que : $ \forall i \mbox{, avec } 1 \le i \le p, x_i \in E_i$. <br> 
    
Lorsque qu’il s’agit du même ensemble, on note alors :<br>
    $E^p = \underset{\mbox{p ensembles}}{\underbrace{E \times E \times \ldots \times E}}$  
    
et les p-uplets sont appelés des p-listes. 
</div>

**Remarque :**  
$E \times F$ se lit "E croix F".   
Les éléments d’un couple, d’un p-uplet ou d’une p-liste sont appelés termes.  

Une p-liste peut aussi se définir comme une suite de p éléments ordonnés, distincts ou non, d’un ensemble E.

- Dans un couple, un p-uplet ou une p-liste, les éléments sont ordonnés et
notés entre parenthèses contrairement à un ensemble où les éléments ne sont pas
ordonnés et notés entre accolades.  

**Exemples :**  
Soit $A = \{1, 2, 3\}$ et $B = \{r, b\}$ alors : $A \times B = \{(1, r), (1, b), (2, r), (2, b), (3, r), (3, b)\}$  

Un couple d’entiers naturels est un élément de $\mathbb{N^2}$.  
Le couple de coordonnées d’un point du plan $(x, y)$ est un élément de $\mathbb{R}^2$.  
Le triplet de coordonnées d’un point de l’espace $(x, y, z)$ est un élément de $\mathbb{R}^3$.  
  
  
Un mot de 5 lettres, compréhensible ou non, est une 5-liste de l’ensemble $E^5$ où $E$ est l’ensemble des 26 lettres de l’alphabet.
 
 
## 5) Principes additif et multiplicatif
<div class="alert alert-block alert-danger">
<b>Théorème  :</b><br>  
On appelle cardinal d’un ensemble fini $E$, noté $|E|$ ou Card(E), le
nombre d’éléments de l’ensemble $E$.<br>
Soit $E$ et $F$ deux ensembles tels que $|E| = n$ et $|F| = p$.  <br>
    
• Principe additif : Si $E \cap F = \emptyset$ alors, $|E \cup F| = n + p$.<br>
• Principe multiplicatif : $|E \times F| = n \times p$.<br>
    
</div>    
    
**Remarque :**  
Par convention $|\emptyset| = 0$.  

**Exemples :**  
Si un plat est composé d’une viande parmi 4 possibles ou d’un
poisson parmi 3 possibles alors, il y a 4 + 3 = 7 plats différents possibles.  

En effet si $V=\{v_1, v_2, v_3, v_4\}$ et $P=\{p_1, p_2, p_3\}$ alors, $V \cap P=\emptyset$ donc $|V \cup P|=7$.   

Si à une boule on associe le numéro 1, 2 ou 3 et la couleur rouge ou bleue alors, il y a $3 \times 2$ boules différentes possibles.  

En effet si $N = \{1, 2, 3\}$ et $C = \{r, b\}$ alors $|N \times C| = 6$

Avec cette propriété, on peut proposer le programme récursif en Python pour calculer n !



```python
def fact(n) :
    if n==0:
        return 1
    else:
        return n*fact(n-1)
```

## 4) Nombre de p-listes d’éléments distincts - Arrangements
<div class="alert alert-block alert-danger">

</div>


**Remarque :**  
Une p-listes d’élément distincts peut être associée à p tirages successifs
**sans remise** dans une urne qui contient n boules.  

Une p-listes d’éléments distincts est parfois appelée arrangement de p éléments
parmi n. Le nombre d’arrangements est alors noté : $A^p_n$  

**Exemples :**

départ est de :  
$$\dfrac{18!}{(18 − 3) !}=\dfrac{18!}{15!}= 4 896$$

2) Le nombre de tirages successifs, sans remise, de 4 boules dans une urne comportant 9 boules numérotées de 1 à 9 est de : 
$$\dfrac{9!}{(9 − 4) !}=\dfrac{9!}{5!}= 3 024$$

# III) Combinaisons
## 1) Nombre de combinaisons

<div class="alert alert-block alert-danger">
<b>Définition :</b><br>  
Soit $E$ un ensemble de $n$ éléments et $p \le n$. 
Une combinaison de $p$ éléments de $E$ est une partie de $E$ à $p$ éléments.<br>  
    
Le nombre de combinaisons de $p$ éléments de $E$, noté:$\begin{pmatrix} n \\ p \end{pmatrix}$ , est égal à :<br>
$$\begin{pmatrix} n \\ p \end{pmatrix}=\dfrac{n!}{p!(n − p)!}$$    
</div>

    
**Remarque:**       
Le nombre de combinaison est parfois noté : $C^p_n=\begin{pmatrix} n \\ p \end{pmatrix}$.

Une combinaison de $p$ éléments parmi $n$ peut-être associée à un tirage simultané de $p$ boules dans une urne qui en comporte $n$.  
    

**Exemples :**  
1) Soit un ensemble $E = {a, b, c, d}$. Les combinaisons de 2 éléments de E sont : ${a, b} , {a, c} , {a, d} , {b, c} , {b, d} , {c, d}$  

Le nombre de combinaisons de 2 parmi 4 vaut :
$$\begin{pmatrix} 4 \\ 2 \end{pmatrix}=\dfrac{4!}{2!(4 − 2)}=\dfrac{4 \times 3}{2}= 6$$  

2) Le nombre de délégations de 4 personnes que l’on peut désigner dans un
groupe de 15 personnes est de :
$$\begin{pmatrix} 15 \\ 4 \end{pmatrix}=\dfrac{15!}{4!(15 − 4)!}=\overset{4 facteurs}{\overbrace{\dfrac{15 \times 14 \times 13 \times 12}{4 \times 3 \times 2}}}= 15 \times 7 \times 13 = 1 365$$  

## 2) Propriétés des coefficients binomiaux
<div class="alert alert-block alert-danger">
<b>Théorème:</b><br> 
Pour tous entiers naturels $n$ et $p$, tels que $0 \le p \le n$, on a :<br>
    
$$\begin{pmatrix} n \\ 0 \end{pmatrix}=\begin{pmatrix} n \\ n \end{pmatrix}=1 \quad\quad
\begin{pmatrix} n \\ 1 \end{pmatrix}=\begin{pmatrix} n \\ n-1 \end{pmatrix}=n
$$  
    <br>
    
$$\begin{pmatrix} n \\ n-p \end{pmatrix}=\begin{pmatrix} n \\ p \end{pmatrix} $$
<br><br>

$\mbox{• Relation de Pascal :}\quad \begin{pmatrix} n \\ p \end{pmatrix}+\begin{pmatrix} n \\ p+1 \end{pmatrix}=\begin{pmatrix} n+1 \\ p+1 \end{pmatrix} $

</div>


**Exemple :**  
$\begin{pmatrix} 7 \\ 0 \end{pmatrix} = \begin{pmatrix} 7 \\ 7 \end{pmatrix}=1$
  
  
$\begin{pmatrix} 8 \\ 1 \end{pmatrix} = \begin{pmatrix} 8 \\ 7 \end{pmatrix}=8$
  
$\begin{pmatrix} 6 \\ 4 \end{pmatrix} = \begin{pmatrix} 6 \\ 2 \end{pmatrix}=15$  

**Remarque :**  
Construire une partie de $p$ éléments de $E$ revient à construire une
partie à $(n − p)$ éléments de $E$ (ceux qui restent) donc :  

$$\begin{pmatrix} n \\ n-p \end{pmatrix}=\begin{pmatrix} n \\ p \end{pmatrix} $$  

<div class="alert alert-block alert-info">
<b>Démonstration : Relation de Pascal :</b><br>
    
• Par une méthode combinatoire :<br>
Dans un ensemble à $(n + 1)$ éléments, on choisit un élément $a$.<br>
Dans les parties à $(p+1)$ éléments on distingue les parties contenant $a$ ou non.  
    
1) Pour les parties contenant $a$, il reste à choisir $p$ éléments parmi les $n$ restants.  <br>
    
Il y a alors  $\begin{pmatrix} n \\ p \end{pmatrix}$ parties de $(p + 1)$ éléments qui contiennent $a$. <br> 
    
2) Pour les parties ne contenant pas $a$, on choisit $(p+1)$ éléments parmi les $n$ restants.  <br>
    
Il y a alors $\begin{pmatrix} n \\ p +1 \end{pmatrix}$ parties de $(p + 1)$ éléments qui ne contiennent pas $a$.  <br>
    
3) Conclusion : on obtient l’égalité<br>
$\begin{pmatrix} n +1 \\ p +1 \end{pmatrix} = \begin{pmatrix} n  \\ p  \end{pmatrix} + \begin{pmatrix} n  \\ p +1 \end{pmatrix}$<br>

• Par le calcul :<br><br>
On rappelle que :   $(p+1) ! = (p+1)p !\quad$ et $\quad (n−p) ! = (n−p)(n−p−1) !$.  
    <br>
$\begin{array}\\
\begin{pmatrix} n \\ p  \end{pmatrix}+\begin{pmatrix} n \\ p +1 \end{pmatrix}&=&\dfrac{ n! }{p!(n-p)!}+\dfrac{n !}{(p+1) !(n − p-1) !}\\ \\
&=&\dfrac{n!(p+1)}{(p+1)!(n-p)!} + \dfrac{n!(n-p)}{(p+1)!(n-p)!} \\ \\
&=&\dfrac{n !(p + 1) + n !(n − p)}{(p + 1) !(n − p) !} \\ \\
&=&\dfrac{n !(n + 1)}{(p + 1) !(n − p) !} \\ \\
&=&\dfrac{(n + 1) !}{(p + 1) !(n − p) !}\\ \\
&=&\dfrac{(n + 1) !}{(p + 1) ! [(n+1)−(p+1)] !}\\ \\
&=&\begin{pmatrix} n + 1 \\ p + 1 \end{pmatrix}
\end{array}$
</div>

## 4) Nombre de parties d’un ensemble
<div class="alert alert-block alert-danger">
<b>Théorème :</b><br> 
Le nombre de parties d’un ensemble à $n$ éléments est : $2^n$.<br>
On a alors la relation :$$\sum_{p=0}^n\begin{pmatrix} n \\ p \end{pmatrix} =2^n$$
</div>

<div class="alert alert-block alert-info">
<b>Démonstration :</b><br> 
Soit un ensemble $E = \{e_1, e_2, \ldots , e_n\}$.
Pour construire un sous-ensemble de $E$, on procède de la façon suivante :<br>
• À chaque élément de $E$ on associe le nombre 1 si l’on sélectionne l’élément et le nombre 0 dans le cas contraire.<br>
• À chaque sous-ensemble de $E$ correspond alors une unique n-liste de l’ensemble ${0, 1}$. <br>
Par exemple :$(0, 0, \ldots , 0)$ correspond à $\emptyset$,     $(1, 0, \ldots , 0)$ correspond à ${e_1}$,<br>
    $(1, 1, 0,\ldots , 0)$ correspond à ${e_1, e_2}$ et     $(1, 1, \ldots , 1)$ correspond à $E$.  
    
Le nombre de parties de $E$ correspond au nombre de n-listes de la paire $\{0, 1\}$ :  $2^n$
    <br><br>
$\begin{pmatrix} n  \\ p \end{pmatrix}$ correspond au nombre de parties de $E$ à $p$ éléments, <br>
    donc la somme $\sum_{p=0}^n\begin{pmatrix} n \\ p \end{pmatrix}$ correspond au nombre de toutes les parties de $E$, d’où :  
    
$$\sum_{p=0}^n\begin{pmatrix} n \\ p \end{pmatrix} =2^n$$

</div>

# IV) En résumé


Pour dénombrer toutes les éventualités d’un problème, les deux critères à prendre en compte sont : 
- les éléments peuvent-ils être répétés ? 
- L’ordre des éléments est-il à prendre en compte ? 

On résume alors les différentes réponses par le tableau :


|Critères | Répétition | Non répétition |
| :--- | :---: | :---: |
| Ordre  |p-listes | p-listes d’éléments distincts |
|Pas d’ordre | Combinaisons avec répétition (hors programme) | Combinaisons |

**Exemples :**  

1) **Le loto :** on tire, au hasard, 6 boules parmi 49. Combien de tirages possibles ?  

• Peut-on obtenir plusieurs fois le même numéro lors d’un tirage ? Non donc pas de répétition.  
• L’ordre d’apparition des différents numéros a-t-il de l’importance ? Non on considère les six numéros globalement ! Donc pas d’ordre.  

On utilise les combinaisons : $\begin{pmatrix} 49  \\ 6 \end{pmatrix}= \dfrac{49!}{6!(49 − 6)!}= 13 983 816$.  

2) **Podium :** Huit coureurs participent à la finale du 100 m. On décerne trois médaillés (or, argent et bronze). Combien y a-t-il de podiums possibles ? 

• Peut-on obtenir plusieurs fois le même coureur sur un podium? Non un même coureur ne peut pas être à la fois médaillé d’or et d’argent ! Donc pas de répétition.  
• L’ordre d’apparition des coureurs sur le podium a-t-il de l’importance ?
Oui car les médailles sont différentes. L’ordre est ici déterminant.  

On utilise des p-listes d’éléments distincts :
$\dfrac{8!}{(8 − 3)!}= 8 \times 7 \times 6 = 336$  

3) **QCM:** un QCM de 15 questions comporte 4 choix (a, b, c, d) dont un seul exact par question. Combien y a-t-il de façons de répondre à ce QCM?  

• Peut-on obtenir plusieurs fois le même choix? Oui car on peut avoir deux fois le choix a choisi. Donc répétition.  

• L’ordre dans dans les choix a-t-il de l’importance ? Oui car un choix donné est associé à une question.  

On utilise des p-listes : $4^{15} = 1 073 741 824$  

4) **le jeu de cartes** 
On tire cinq cartes d’un jeu de 32 (du 7 à l’as). Nombre de mains contenant :
1) Le valet de trèfle ?  
2) Exactement deux coeurs ?  
3) Exactement un roi, une dame et deux valet ?  
4) Ni le roi de trèfle, ni un pique?  
5) Au moins un roi ?  
6) L’as de pique et au moins deux trèfles ?  
7) Exactement un roi et deux carreaux?
