<div class="alert alert-block alert-danger">

**Remarque :**  
Construire une partie de $p$ lments de $E$ revient  construire une
partie  $(n  p)$ lments de $E$ (ceux qui restent) donc :  

$$\begin{pmatrix} n \\ n-p \end{pmatrix}=\begin{pmatrix} n \\ p \end{pmatrix} $$  

<div class="alert alert-block alert-info">
<b>Dmonstration : Relation de Pascal :</b><br>
    
 Par une mthode combinatoire :<br>
Dans un ensemble  $(n + 1)$ lments, on choisit un lment $a$.<br>
Dans les parties  $(p+1)$ lments on distingue les parties contenant $a$ ou non.  
    
1) Pour les parties contenant $a$, il reste  choisir $p$ lments parmi les $n$ restants.  <br>
    
Il y a alors  $\begin{pmatrix} n \\ p \end{pmatrix}$ parties de $(p + 1)$ lments qui contiennent $a$. <br> 
    
2) Pour les parties ne contenant pas $a$, on choisit $(p+1)$ lments parmi les $n$ restants.  <br>
    
Il y a alors $\begin{pmatrix} n \\ p +1 \end{pmatrix}$ parties de $(p + 1)$ lments qui ne contiennent pas $a$.  <br>
    
3) Conclusion : on obtient lgalit<br>
$\begin{pmatrix} n +1 \\ p +1 \end{pmatrix} = \begin{pmatrix} n  \\ p  \end{pmatrix} + \begin{pmatrix} n  \\ p +1 \end{pmatrix}$<br>

 Par le calcul :<br><br>
On rappelle que :   $(p+1) ! = (p+1)p !\quad$ et $\quad (np) ! = (np)(np1) !$.  
    <br>
$\begin{array}\\
\begin{pmatrix} n \\ p  \end{pmatrix}+\begin{pmatrix} n \\ p +1 \end{pmatrix}&=&\dfrac{ n! }{p!(n-p)!}+\dfrac{n !}{(p+1) !(n  p-1) !}\\ \\
&=&\dfrac{n!(p+1)}{(p+1)!(n-p)!} + \dfrac{n!(n-p)}{(p+1)!(n-p)!} \\ \\
&=&\dfrac{n !(p + 1) + n !(n  p)}{(p + 1) !(n  p) !} \\ \\
&=&\dfrac{n !(n + 1)}{(p + 1) !(n  p) !} \\ \\
&=&\dfrac{(n + 1) !}{(p + 1) !(n  p) !}\\ \\
&=&\dfrac{(n + 1) !}{(p + 1) ! [(n+1)(p+1)] !}\\ \\
&=&\begin{pmatrix} n + 1 \\ p + 1 \end{pmatrix}
\end{array}$
