# LIMITES

## I)  Limite à l’infini 
### 1)  Limite finie à l’infini 
<div class="alert alert-block alert-danger">
    Une fonction $f$ a pour limite $\mathcal{l}$ en $+\infty$, <b>si</b><br>
tout intervalle ouvert incluant $\mathcal{l}$, contient toutes les valeurs
de f(x) pour <b>x  grand </b>($x \in ]A;+\infty[$). <br><br>
    
On note alors :<br>
$$\lim\limits_{x \to +\infty} f(x)= \mathcal{l}$$
<br>    
On définit de façon analogue 
 $$\lim\limits_{x \to -\infty} f(x)= \mathcal{l}$$ avec $x \in ] −\infty ; B[$
   
 </div>   




**notion d'asymptote:**   
La droite $D$ d’équation $y = \mathcal{l}$ est dite asymptote horizontale à $\mathcal{C_f}$ .   

<div class="alert alert-block alert-success">
    <b>Exemple :</b> <br> 
$x \to \dfrac{1}{x}$ , $\quad   x \to \dfrac{1}{x^n}, n \in \mathbb{N}^*$ et $\quad  x \to \dfrac{1}{\sqrt{x}}$
ont des limites nulles en $+\infty$. <br> 
    
$x \to e^x$ a pour limite 0 en $-\infty$.<br><br>
Leurs courbes admettent alors la droite d’équation $y = 0$ (l’axe des abscisses) comme asymptote horizontale.
</div>    

### 2) Limite infinie à l’infini 
<div class="alert alert-block alert-danger">
    <b>Définition 2 :</b><br>
    Une fonction f a pour limite $+\infty$ en $+\infty$, <b>si</b><br>
tout intervalle $]M;+\infty[$ contient toutes les valeurs de $f(x)$ pour <b>x grand </b>($ x \in ]A;+\infty[$).<br>
    <br>
On note alors :<br>
    $$\lim \limits_{x \to +\infty}f(x)=+\infty$$
<br>
On définit de façon analogue :<br>
$\lim \limits_{x \to -\infty}f(x)=+\infty$ de $]M ; +\infty[$ avec x dans $] − \infty ; B[$<br>
$\lim \limits_{x \to +\infty}f(x)=-\infty$ de $]-\infty ; m[$ avec x dans $] A ; +\infty[$<br>
$\lim \limits_{x \to -\infty}f(x)=-\infty$ de $]-\infty ; m[$ avec x dans $] -\infty ; B[$<br>    
 
    
</div> 

**exemple:**  
Une fonction peut tendre vers $+\infty$ en $+\infty$ de plusieurs façons.  
C’est le cas par exemple des fonctions $x \to x^2$, $\quad x \to x$ et $\quad x \to \sqrt{x}$.  

• $x \to x^2$ tend « rapidement » vers l’infini.
La concavité est tournée vers le haut.    

• $x \to x$ tend «moyennement » vers l’infini.
Pas de concavité.  
• $x \to \sqrt{x}$ tend « lentement » vers l’infini.
La concavité est tournée vers le bas  



## II)  Limite en un point 
### 1) Limite infinie en un point 
<div class="alert alert-block alert-danger">
<b> Définition :</b>  <br>
Une fonction $f$ a pour limite $+\infty$ en $a$, <b>si</b><br>
tout intervalle $]M;+\infty[$ contient toutes les valeurs de $f(x)$ pour $x$  proche de $a$ ( $x$ dans un intervalle ouvert contenant $a$).<br> 
On note alors :<br>
$$\lim\limits_{x \to a} f(x)=+\infty$$

On définit de façon analogue:<br>
$$\lim\limits_{x \to a} f(x)=-\infty$$
    
</div>  

**notion d'asymptote:**  

La droite $D$ d’équation $x = a$ est dite asymptote verticale à $\mathcal{C_f}$  


**remarque:**  
On définit la limite à gauche ou à droite de $x = a$ lorsque la limite en $ x = a$ n’existe pas :  

limite à gauche :  
$\lim\limits_{\substack{x \to a \\ x<a}}f(x) $ ou $\lim \limits_{x \to a^-}f(x)$
  
limite à droite :   
$\lim\limits_{\substack{x \to a \\ x>a}}f(x) $ ou $\lim \limits_{x \to a^+}f(x)$  

**exemple:**  

La fonction $x\to \dfrac{1}{x}$ n’admet pas de limite en 0,
mais admet une limite à gauche et à droite de 0.

### 2)  Limite finie en un point 
<div class="alert alert-block alert-danger">
    <b>Définition</b><br>
 Une fonction $f$ a pour limite $ℓ$ en $a$, <b>si</b><br>
 tout intervalle ouvert contenant $ℓ$ contient toutes les valeurs de $f(x)$ pour $x$ assez proche de $a$.<br>
On note alors :
$$\lim \limits_{x \to a}f (x) = ℓ$$
</div>

**Exemple :**  
$\lim \limits_{x \to 2} x^2-1=3$.  

Une fonction peut ne pas être définie et admettre une limite en $a$.  

Par exemple la fonction $x \to \dfrac{e^x − 1}{x}$ n’est pas définie en $0$ mais :
$$\lim \limits_{x \to 0} \dfrac{e^x-1}{x}=1$$

Taux d’accroissement de exp en 0.

## III) Opérations sur les limites
### 1) Somme de fonctions
La limite d'une somme de fonctions est la somme de leur limite.   

Il existe une forme indéterminée pour $f+g$ avec...  


**Exemples :**  
1) Limite en $+\infty$ de la fonction $f$ définie sur $\mathbb{R^∗}$ par : $f (x) = x + 3 + \dfrac{1}{x}$  
2) Limite en $+\infty$ et $−\infty$ de la fonction $f$ définie sur $\mathbb{R}$ par : $f (x) = x^2 + x$  



### 2) Produit de fonctions   
La limite d'un produit de fonctions est le produit de leur limite. 

**Exemples :**  
1) Limite en $−\infty$ de la fonction précédente : $f (x) = x^2 + x$  
Pour lever la forme indéterminée, on factorise $f(x)$ par le terme prédominant :  
$f (x) = x^2 + x = x^2\Big(1+\dfrac{1}{x}\Big)$  

2) Limite en $+\infty$ de la fonction définie sur $\mathbb{R^+}$ par : $f(x) = x − \sqrt{x}$  

3) Limite à droite de $0$ de la fonction définie sur $\mathbb{R^∗}$ par : $f(x) =\dfrac{1}{x}sin(x)$  



### 3) Quotient de fonctions
La limite d'un quotient de fonctions est le quotient de leur limite.   

**Exemples :**  
1) Limite en $−2$ de la fonction définie sur $\mathbb{R }− \{−2\}$ par : $f (x) =\dfrac{
2x − 1}{x + 2}$.  
  
2) Limite en $+\infty$ de la fonction f définie par : $f (x) =\dfrac{2x + 1}{3x + 2 }$  


## IV)  Limite d’une fonction composée 
<div class="alert alert-block alert-danger">
    <b>Théorème 1 :</b><br>  
Soit deux fonctions $f , g$ et $a, b, c$ des réels ou $±\infty$.  <br>
    $$\lim \limits_{x \to a} f(x)=b$$ et comme $$\lim \limits_{x \to b}g(x)=c$$
    alors $$\lim \limits_{x \to a} g[f(x)]=c$$
</div>  


**Exemples:**  
1) $\lim \limits_{x \to +\infty}cos\Big( \dfrac{1}{x^2+1}\Big)$  
  
<div class="alert alert-block alert-danger">
<b>Théorème 2 :</b><br>
Limites de fonctions et de suites .<br>  
Soit une suite $(U_n)$ définie par : $u_n = f (n)$, de façon explicite. $f$ est alors la fonction réelle associée à
la suite $(u_n)$. <br>
    Soit $a$ un réel ou $±\infty$.<br>  
$$ \lim \limits_{x \to +\infty} f(x) = a \quad \Rightarrow \lim \limits_{n \to +\infty} u_n = a$$ 
</div>  

**Exemple :**  
Soit la suite $(u_n)$ définie pour tout $n \in \mathbb{N^∗}$ par : $u_n = \sqrt{2 + \dfrac{1}{n^2}}$ .  

Soit $f$ la fonction définie sur $]0;+\infty[$ par : $f(x) = \sqrt{2 + \dfrac{1}{x^2}}$ .  

$ \begin{aligned}
\lim \limits_{x \to +\infty} 2+\dfrac{1}{x^2} = & 2 & \\
& \lim \limits_{x \to 2} \sqrt{x}& = \sqrt{2}\\
\end{aligned} $

Par composition
$\lim \limits_{x \to +\infty} f(x)=\sqrt{2} \Rightarrow \lim \limits_{n \to +\infty} u_n=\sqrt{2}$

## V)  Théorèmes des gendarmes et de comparaison

<div class="alert alert-block alert-danger">
<b>Théorème </b><br>
$f$ , $g$, $h$ trois fonctions définies sur $I =]b ;+\infty[$ et $ℓ$ un réel.<br>
Si pour tout $x \in I$ :<br>
    
<b>Théorème des «Gendarmes »</b> (pour montrer une limite finie)<br><br>
$ \begin{array} \\
g(x) \le f(x) \le h(x) &\\
\lim \limits_{x \to +\infty} g(x)  =  \lim \limits_{x \to +\infty} h(x) =  l & \\
\end{array}
\Bigg \rbrace \lim \limits_{x \to +\infty} f(x) =  l $ 
<br>      
<b>Théorèmes de comparaison</b> (pour montrer une limite infinie) <br>
 <br>
$ \begin{array} \\  
f(x) \ge g(x) \\
\lim \limits_{x \to +\infty} g(x) = +\infty \\
\end{array} \Bigg \rbrace \lim \limits_{x \to +\infty} f(x) = +\infty$ et inversement <br><br> 
    
$ \begin{array} \\  
f(x) \le g(x) \\
\lim \limits_{x \to +\infty} g(x) = -\infty \\
\end{array} \Bigg \rbrace \lim \limits_{x \to +\infty} f(x) = -\infty$
    
</div> 

**Exemples :**   

1) Montrer que $\lim \limits_{x \to +\infty} \dfrac{sin(x)}{x}=0$ 
  
2) Déterminer $\lim \limits_{x \to +\infty} x + cos (x)$  
    
  
### théorème des croissances comparées
<div class="alert alert-block alert-danger">
Pour tout entier naturel n, on a :
$$\lim \limits_{x \to +\infty} \dfrac{e^x}{x^n}=+\infty$$ 
<br>
$$\lim \limits_{x \to -\infty} x^ne^x=0$$  
</div>

Ce théorème illustre le fait que la fonction exponentielle croît en $+\infty$ bien plus vite que toute fonction puissance.   

    
    
## VI)  Continuité 
### 1) Continuité en un point  
<div class="alert alert-block alert-danger">
Soit une fonction $f$ définie sur un intervalle ouvert $I$ contenant
le réel $a$. On dit que $f$ est continue en $a$ si et seulement si :<br>
    $$\lim \limits_{x \to a}f (x) = f (a)$$<br>  
    
La fonction $f$ est continue sur un intervalle $I$ si, et seulement si, $f$ est continue en tout point de $I$.  
</div>    
  
Graphiquement, la continuité d’une fonction $f$ sur un intervalle I se
traduit par une courbe en « un seul morceau ».  



### 2) Continuité des fonctions usuelles  
  
On a le tableau de continuité des fonctions de référence suivant :


|Fonctions usuelles | Intervalle de continuité |
|---|---|
|$f (x) = x^n, n \in \mathbb{N}$ | $\mathbb{R}$|
|$f (x) = \dfrac{1}{x^n}$ | $ ]−\infty ; 0[$  et $]0 ;+\infty[$|
|$f (x) = \sqrt{x}$| $[0 ;+\infty[$|
|$f (x) = |x|$|$\mathbb{R}$|
|$f (x) = e^x$| $\mathbb{R}$|
|$f (x) = sin(x)$ et $f (x) = cos(x)$|  $\mathbb{R}$|

<div class="alert alert-block alert-danger">
    <b>Théorème:</b>  <br>
Toutes fonctions construites par somme, produit, quotient ou composition à partir des fonctions de référence sont continues sur leur ensemble de définition.
</div>




### 3) Continuité et suite 

<div class="alert alert-block alert-danger">
 <b> Théorème du point fixe</b><br>
Soit une suite $(U_n)$ définie par la relation $U_{n+1} = f (U_n)$ convergente vers $ℓ$.<br>
Si la fonction associée $f$ est continue en $ℓ$, alors la limite de la suite $ℓ$ est solution de l’équation $f (x) = x$.
</div>   
 
**Exemple :**   Soit la suite $(U_n)$ définie par $\begin{cases} u_0= 0 \\ u_{n+1}=\sqrt{3u_n+4}\end{cases}$

  
On montre par récurrence que $(u_n)$ est positive, croissante etmajorée par $4$, d’après le théorème des suites monotone, $(u_n)$ est convergente vers une limite $ℓ$.  

La fonction $x \to \sqrt{3x + 4}$ est continue sur $[ \dfrac{-3}{4};+\infty[$ donc sur $[0; 4]$, d’après le théorème du point fixe, sa limite $ℓ$ est solution de l’équation $f (x) = x$.  


 
### 4) Continuité et dérivabilité 

<div class="alert alert-block alert-danger">
<b>Théorème</b><br>
• Si $f$ est dérivable en $a$ alors la fonction $f$ est continue en $a$. <br> 
    
• Si $f$ est dérivable sur un intervalle $I$ alors la fonction $f$ est continue sur $I$.<br>
 </div>   

### 5) Continuité et équation   

<div class="alert alert-block alert-danger">
 <b>Théorème des valeurs intermédiaires </b>
Soit une fonction <b>continue</b> sur un intervalle $I = [a, b]$.<br>
Pour tout réel $k$ compris entre $f(a)$ et $f(b)$, l’équation $f(x) = k$ admet au moins une solution $c$ dans $I$.<br>
</div>    



    
<div class="alert alert-block alert-danger">
<b>Théorème des valeurs intermédiaires bis - TVI</b><br>
Soit une fonction $f$ <b>continue et strictement monotone</b> sur $I = [a, b]$.<br>
Pour tout réel $k$ compris entre $f(a)$ et $f(b)$, l’équation $f(x) = k$ admet une unique solution dans $I$.<br>
</div>





**Exemple :**  
Soit la fonction $f$ définie sur $\mathbb{R}$ par : $f (x) = x^3 + x − 1$.
Montrer que l’équation $f(x) = 0$ n’admet qu’une solution sur $\mathbb{R}$.   
On donnera un encadrement à l’unité de cette solution.  
Trouver ensuite, à l’aide d’un algorithme un encadrement à $10^{−6}$ de cette solution.
