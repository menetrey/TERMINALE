### 1)  Limite finie  linfini 
<div class="alert alert-block alert-danger">

On note alors :<br>
$$\lim\limits_{x \to +\infty} f(x)= \mathcal{l}$$
<br>    
On dfinit de faon analogue 
 $$\lim\limits_{x \to -\infty} f(x)= \mathcal{l}$$ avec $x \in ] \infty ; B[$
   
 </div>   




**notion d'asymptote:**   
La droite $D$ dquation $y = \mathcal{l}$ est dite asymptote horizontale  $\mathcal{C_f}$ .   

<div class="alert alert-block alert-success">
    <b>Exemple :</b> <br>

$x \to e^x$ a pour limite 0 en $-\infty$.<br><br>

On note alors :<br>
$$\lim\limits_{x \to a} f(x)=+\infty$$

On dfinit de faon analogue:<br>
$$\lim\limits_{x \to a} f(x)=-\infty$$
    
</div>  

**notion d'asymptote:**  

La droite $D$ dquation $x = a$ est dite asymptote verticale  $\mathcal{C_f}$  


**remarque:**  
On dfinit la limite  gauche ou  droite de $x = a$ lorsque la limite en $ x = a$ nexiste pas :  

limite  gauche :  
$\lim\limits_{\substack{x \to a \\ x<a}}f(x) $ ou $\lim \limits_{x \to a^-}f(x)$
  
limite  droite :   
$\lim\limits_{\substack{x \to a \\ x>a}}f(x) $ ou $\lim \limits_{x \to a^+}f(x)$  

**exemple:**  

La fonction $x\to \dfrac{1}{x}$ nadmet pas de limite en 0,
mais admet une limite  gauche et  droite de 0.

## VI)  Continuit 
### 1) Continuit en un point  
<div class="alert alert-block alert-danger">
Soit une fonction $f$ dfinie sur un intervalle ouvert $I$ contenant
le rel $a$. On dit que $f$ est continue en $a$ si et seulement si :<br>
    $$\lim \limits_{x \to a}f (x) = f (a)$$<br>  
    
La fonction $f$ est continue sur un intervalle $I$ si, et seulement si, $f$ est continue en tout point de $I$.  
</div>    
  
Graphiquement, la continuit dune fonction $f$ sur un intervalle I se
traduit par une courbe en  un seul morceau .
