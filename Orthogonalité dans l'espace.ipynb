# Produit Scalaire Orthogonalité


## I)  Relations entre droites et plans
### 1) Relations entre deux droites
<div class="alert alert-block alert-danger">
Deux droites, dans l’espace, peuvent être :<br>
• sécantes, si ces deux droites se coupent en un
point, par exemple (AF) et (BE) ;<br>
• parallèles, si ces deux droites sont coplanaires et
n’ont aucun point commun ou si ces deux droites
sont confondues, par exemple (AB) et (HG) ;<br>
• non coplanaires par exemple (AB) et (DG).
       
</div> 

 

### 2) Relations entre une droite et un plan
<div class="alert alert-block alert-danger">
Une droite et un plan peuvent être :<br>
• parallèles : si la droite et le plan n’ont aucun point
commun ou si la droite est contenue dans le plan,
par exemple (EF) et (P) ;<br>
• sécantes : si la droite et le plan ont un seul point commun,
par exemple (HI) et (P).<br>
</div>    
    
    
    

### 3) Relation entre deux plans
<div class="alert alert-block alert-danger">

Deux plans peuvent être :<br>
• parallèles : si les deux plans n’ont aucun point
commun ou si les deux plans sont confondus, par
exemple (P1) et (P2).<br>
• sécants : si les deux plans ont une droite en commun,
par exemple (P1) et (P3) d’intersection (BC).<br>
</div>


### 4) Parallélisme
<div class="alert alert-block alert-danger">

<b>Théorème: </b> <br>
Parallélisme de droites et de plans<br>
• Une droite est parallèle à un plan si et seulement si elle est parallèle à une droite de ce plan.<br>
• Deux plans sont parallèles si et seulement si deux droites sécantes de l’un sont parallèles à deux droites sécantes de l’autre.<br>
</div>





## II)  Produit scalaire
### 1)  Définition
<div class="alert alert-block alert-danger">

<b>Définition:</b><br>
Le produit scalaire dans le plan se généralise à l’espace.<br>
Le produit scalaire de deux vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ est le nombre réel, noté $\overrightarrow{u} ·\overrightarrow{v}$, tel que :<br>
• Par le cosinus : $\overrightarrow{u} ·\overrightarrow{v} = ||\overrightarrow{u}|| × ||\overrightarrow{v}|| \times cos(\overrightarrow{u},\overrightarrow{v})$. <br>
• Par le projeté : $\overrightarrow{u} ·\overrightarrow{v} = \overrightarrow{AB} · \overrightarrow{AC} = ±AB \times AH$ <br>
avec $\overrightarrow{u} = \overrightarrow{AB}$ et $\overrightarrow{v} = \overrightarrow{AC}$ et H le projeté orthogonal de C sur la droite $(AB)$.<br>
• Par la norme : $\overrightarrow{u} ·\overrightarrow{v} = \dfrac{1}{2}\Big ( ||\overrightarrow{u} + \overrightarrow{v}||^2 − ||\overrightarrow{u}||^2 − ||\overrightarrow{v}||^2 \Big )$ . <br>

• Par les coordonnées : $\overrightarrow{u} ·\overrightarrow{v}= \begin{pmatrix}x \\ y  \\ z \end{pmatrix} . \begin{pmatrix} x′ \\ y′ \\ z′ \end{pmatrix} = xx′ + yy′ + zz′$  
    
avec $\overrightarrow{u}(x ; y ; z)$ et $\overrightarrow{v}(x′ ; y′ ; z′)$ . 
    
</div>


**Exemple :**    
Soit les vecteurs $\overrightarrow{u}(2 ; \sqrt{3} ; 1)$ et $\overrightarrow{u}(3 ; \sqrt{3} ; 2)$.  
Calculer $\overrightarrow{u} ·\overrightarrow{v} $, puis déterminer une mesure de l’angle $(\overrightarrow{u},\overrightarrow{v})$ au degré près.  




**Exemple :**   
ABCDEFGH est un cube d’arête a. O est le centre de la face EFGH.  
Calculer $\overrightarrow{AE} · \overrightarrow{AO}$ en fonction de $a$.  






### 2) Propriétés et orthogonalité de deux vecteurs
<div class="alert alert-block alert-info">
Le produit scalaire est une forme :<br>
• Symétrique :  $\overrightarrow{u} ·\overrightarrow{v}  =  \overrightarrow{v} ·\overrightarrow{u}$<br> 
• Bilinéaire :  $\overrightarrow{u} ·(\overrightarrow{v} + \overrightarrow{w}) = \overrightarrow{u} ·\overrightarrow{v} + \overrightarrow{u} ·\overrightarrow{w} \quad $     et $(a\overrightarrow{u}) · (b\overrightarrow{v}) = ab \times (\overrightarrow{u} ·\overrightarrow{v})$. <br>
</div>    
    
**Remarque :** La bilinéarité du produit scalaire est une sorte de « distributivité ».  

Symétrie et bilinéarité permettent d’écrire les « identités remarquables » suivantes :  

$(\overrightarrow{u} ± \overrightarrow{v})^2 = \overrightarrow{u}^2 + \overrightarrow{v}^2 ± 2 \overrightarrow{u} ·\overrightarrow{v} $ et   
$(\overrightarrow{u} − \overrightarrow{v})(\overrightarrow{u} + \overrightarrow{v}) = \overrightarrow{u}^2 − \overrightarrow{v}^2$  

Que l’on peut transposer avec les normes :  

$||\overrightarrow{u} ± \overrightarrow{v}||^2 = ||\overrightarrow{u}||^2 + ||\overrightarrow{v}||^2 ± 2 \overrightarrow{u} ·\overrightarrow{v} \quad$ et $(\overrightarrow{u} − \overrightarrow{v})(\overrightarrow{u} + \overrightarrow{v}) = ||\overrightarrow{u}||^2 − ||\overrightarrow{v}||^2$  

<div class="alert alert-block alert-info">
    
<b>Colinéarité et orthogonalité de deux vecteurs</b><br>
    
• Si $\overrightarrow{u}$ et $\overrightarrow{v}$ sont colinéaires et de même sens : $\overrightarrow{u} ·\overrightarrow{v} = ||\overrightarrow{u}|| \times ||\overrightarrow{v}||$<br>
• Si $\overrightarrow{u}$ et $\overrightarrow{v}$ sont colinéaires et de sens contraires :  $\overrightarrow{u} ·\overrightarrow{v} = - ||\overrightarrow{u}|| \times ||\overrightarrow{v}||$<br>
• $\overrightarrow{u}$ et $\overrightarrow{v}$ sont orthogonaux si, et seulement si : $\overrightarrow{u} ·\overrightarrow{v} = 0$ <br>
 </div>
 
 
**Remarque :**   
Le vecteur nul est orthogonal à tout vecteur.  


**Exemple :**   
Soit les points $A(6 ; 8 ; 2)$, $B(4 ; 9 ; 1)$ et $C(5 ; 7 ; 3)$.
Montrer que le triangle $ABC$ est rectangle en $A$.  
  
## II)  Orthogonalité dans l’espace
### 1)  Droites orthogonales
<div class="alert alert-block alert-danger">
    
Deux droites $d_1$ et $d_2$ de vecteurs directeurs $\overrightarrow{u_1}$ et $\overrightarrow{u_2}$ sont :<br>
• orthogonales si, et seulement si : $\overrightarrow{u_1} · \overrightarrow{u_2} = 0$.<br>
• perpendiculaires si et seulement si $d_1$ et $d_2$ sont orthogonales et sécantes.
</div>


**Remarque :**  

On écrit indistinctement $d_1 \perp d_2$ dans le deux cas.  

Dans l’espace, on distingue droites « orthogonales » et droites «perpendiculaires ».  

Dans le cube :   
• les droites $d_1$ et $d_2$ sont orthogonales mais pas perpendiculaires.  
• les droites $D$ et $d_2$ sont perpendiculaires donc orthogonales. 

  


**Exemple :**  
Soit les points $A(2 ;−5 ; 1)$ et $B(0 ; 2 ; 6)$.
Démontrer que la droite d de vecteur directeur $\overrightarrow{u}(−4 ; 1 ;−3) $ est orthogonale à la droite $(AB)$.  

### 2) Droite et plan orthogonaux
<div class="alert alert-block alert-danger">
            
<b>Définition:</b><br>
Un plan $(P)$ de vecteurs directeurs $(\overrightarrow{u_1}, \overrightarrow{u_2})$ est orthogonal à une droite $d$ de vecteur directeur $\overrightarrow{v}$ si, et seulement si, $\overrightarrow{u_1} ·\overrightarrow{v} = 0$ et $\overrightarrow{u_2} ·\overrightarrow{v}= 0$
</div>




**Exemple :**  
Soit les points $A(2 ; 0 ; 2)$, $B(4 ; 0 ; 0)$, $C(1 ;−2 ; 1)$, $D(−1 ; 1 ; 0)$ et $E(1 ;−1 ; 2)$.  

Le plan $(ABC)$ et la droite $(DE)$ sont-ils orthogonaux?  



### 3) Plans orthogonaux
<div class="alert alert-block alert-danger">
    
<b>Définition:</b><br>
Un plan $(P)$ est orthogonal à un plan $(Q)$ si, et seulement si, il
existe une droite d du plan $(Q)$ orthogonale au plan $(P)$.  
</div>


## III) Équation cartésienne d’un plan
### 1)  Vecteur normal

<div class="alert alert-block alert-danger">
<b>Définition:</b><br>
    
Un vecteur $\overrightarrow{n}$ est normal à un plan $(P)$ si $\overrightarrow{n}$ est orthogonal à un couple de vecteurs directeur $(\overrightarrow{u},\overrightarrow{v})$ de $(P)$.  
</div>

**Remarque :**  
$(\overrightarrow{u},\overrightarrow{v},\overrightarrow{n})$ forme alors une base de l’espace.  

<div class="alert alert-block alert-danger">
    
<b>Théorème :</b><br>
Deux plans de vecteurs normaux respectifs $\overrightarrow{n_1}$ et $\overrightarrow{n_2}$ sont orthogonaux si et seulement si : $\overrightarrow{n_1} · \overrightarrow{n_2} = 0$
</div>

**Remarque :**

<b>Théorème :</b><br>
Le plan $(P)$ passant par le point $A$ et de vecteur normal $\overrightarrow{n}$ est l’ensemble des points $M$ tels que : $\overrightarrow{AM}·\overrightarrow{n} = 0$
 </div>
 
 
**Démonstration :**  
Si $(\overrightarrow{u},\overrightarrow{v})$ est un couple de vecteurs directeur de $(P)$ alors pour tout point M, il existe $a, b \in \mathbb{R}$ tels que : $\overrightarrow{AM}= a\overrightarrow{u} + b\overrightarrow{v}$.  


On a alors : $\overrightarrow{AM}·\overrightarrow{n} = (a\overrightarrow{u} + b\overrightarrow{v}) \overrightarrow{n} = a\overrightarrow{u} ·\overrightarrow{n} +  b\overrightarrow{v} . \overrightarrow{n} = 0$ 


### 2) Équation d’un plan
<div class="alert alert-block alert-danger">
    
<b>Théorème:</b><br>
    
Une équation cartésienne d’un plan est de la forme :<br>
$ax + by + cz + d = 0$ avec $a, b, c$ non tous nuls.  <br>
    
Le vecteur $\overrightarrow{n}(a ; b ; c)$ est alors un vecteur normal au plan.
    
</div>


**Démonstration : Par une double implication.**  

Soit le plan $(P)$ passant par $A$ et de vecteur normal $\overrightarrow{n}(a ; b ; c)$.  

Un point $M(x ; y ; z)  \in (P)$ vérifie alors :  

$$\overrightarrow{AM}·\overrightarrow{n} = 0 \iff a(x − x_A) + b(y − y_A) + c(z − z_A) = 0 \iff  ax + by + cz − (ax_A + by_A + cz_A) = 0$$  

On pose $d = −(ax_A + by_A + cz_A)$, on a alors $ax + by + cz + d = 0$.  


**Exemple :**
Déterminer une équation cartésienne du plan (P) passant par le point 
$A(3 ; 5 ; 2)$ et de vecteur normal $\overrightarrow{n} (2 ;−3 ;−1)$.

### 3) Distance d’un point à un plan
<div class="alert alert-block alert-danger">
<b>Projeté orthogonal.</b><br>
Le projeté orthogonal d’un point A sur  un plan (P) est le point
d’intersection H du plan (P) et de la perpendiculaire à ce plan passant par le point A.
</div>

<div class="alert alert-block alert-danger">
    
<b>Théorème: Distance d’un point à un plan.</b><br>
On appelle distance d’un point M au plan (P), la longueur MH où H est le projeté orthogonal de M sur le plan (P). Cette distance est la plus courte distance entre le point M et un point du plan (P).
 </div>
 
**Démonstration :**  
Soit $H$ le projeté orthogonal du point $M$ sur le plan $(P)$ et $A$ un point de $(P)$ **distinct** de $H$.  
La droite $(MH)$ est orthogonale au plan $(P)$ donc elle est orthogonale à toutes droites du plan $(P)$ et donc à la droite $(AH)$.  
Le triangle $AMH$ est rectangle en $H$, d’après le théorème
de Pythagore : $$AM^2 = AH^2 +MH^2$$  

Comme $AH \neq 0$ alors $AM > MH$.   
La distance $MH$ est la plus courte distance de $M$ à un point du plan $(P)$.  



**Exemple :**   
Calculer les coordonnées du projeté orthogonalHdu point $A(7 ; 0 ; 4)$ sur le plan $(P)$ d’équation : $2x − y + 3z + 1 = 0$.  
En déduire la distance du point $A$ au plan $(P)$.
