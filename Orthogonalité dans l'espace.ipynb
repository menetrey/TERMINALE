## I)  Relations entre droites et plans
### 1) Relations entre deux droites
<div class="alert alert-block alert-danger">
Deux droites, dans lespace, peuvent tre :<br>
 scantes, si ces deux droites se coupent en un
point, par exemple (AF) et (BE) ;<br>
 parallles, si ces deux droites sont coplanaires et
nont aucun point commun ou si ces deux droites
sont confondues, par exemple (AB) et (HG) ;<br>
 non coplanaires par exemple (AB) et (DG).
       
</div> 

 

### 2) Relations entre une droite et un plan
<div class="alert alert-block alert-danger">
Une droite et un plan peuvent tre :<br>
 parallles : si la droite et le plan nont aucun point
commun ou si la droite est contenue dans le plan,
par exemple (EF) et (P) ;<br>
 scantes : si la droite et le plan ont un seul point commun,
par exemple (HI) et (P).<br>
</div>    
    
    
    

### 3) Relation entre deux plans
<div class="alert alert-block alert-danger">

Deux plans peuvent tre :<br>
 parallles : si les deux plans nont aucun point
commun ou si les deux plans sont confondus, par
exemple (P1) et (P2).<br>
 scants : si les deux plans ont une droite en commun,
par exemple (P1) et (P3) dintersection (BC).<br>
</div>


### 4) Paralllisme
<div class="alert alert-block alert-danger">

<b>Thorme: </b> <br>
Paralllisme de droites et de plans<br>
 Une droite est parallle  un plan si et seulement si elle est parallle  une droite de ce plan.<br>
 Deux plans sont parallles si et seulement si deux droites scantes de lun sont parallles  deux droites scantes de lautre.<br>
</div>





## II)  Produit scalaire
### 1)  Dfinition
<div class="alert alert-block alert-danger">

<b>Dfinition:</b><br>
Le produit scalaire dans le plan se gnralise  lespace.<br>
Le produit scalaire de deux vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ est le nombre rel, not $\overrightarrow{u} \overrightarrow{v}$, tel que :<br>
 Par le cosinus : $\overrightarrow{u} \overrightarrow{v} = ||\overrightarrow{u}||  ||\overrightarrow{v}|| \times cos(\overrightarrow{u},\overrightarrow{v})$. <br>
 Par le projet : $\overrightarrow{u} \overrightarrow{v} = \overrightarrow{AB}  \overrightarrow{AC} = AB \times AH$ <br>
avec $\overrightarrow{u} = \overrightarrow{AB}$ et $\overrightarrow{v} = \overrightarrow{AC}$ et H le projet orthogonal de C sur la droite $(AB)$.<br>
 Par la norme : $\overrightarrow{u} \overrightarrow{v} = \dfrac{1}{2}\Big ( ||\overrightarrow{u} + \overrightarrow{v}||^2  ||\overrightarrow{u}||^2  ||\overrightarrow{v}||^2 \Big )$ . <br>

 Par les coordonnes : $\overrightarrow{u} \overrightarrow{v}= \begin{pmatrix}x \\ y  \\ z \end{pmatrix} . \begin{pmatrix} x \\ y \\ z \end{pmatrix} = xx + yy + zz$  
    
avec $\overrightarrow{u}(x ; y ; z)$ et $\overrightarrow{v}(x ; y ; z)$ . 
    
</div>

### 2) Proprits et orthogonalit de deux vecteurs
<div class="alert alert-block alert-info">
Le produit scalaire est une forme :<br>
 Symtrique :  $\overrightarrow{u} \overrightarrow{v}  =  \overrightarrow{v} \overrightarrow{u}$<br> 
 Bilinaire :  $\overrightarrow{u} (\overrightarrow{v} + \overrightarrow{w}) = \overrightarrow{u} \overrightarrow{v} + \overrightarrow{u} \overrightarrow{w} \quad $     et $(a\overrightarrow{u})  (b\overrightarrow{v}) = ab \times (\overrightarrow{u} \overrightarrow{v})$. <br>
</div>    
    
**Remarque :** La bilinarit du produit scalaire est une sorte de  distributivit .  

Symtrie et bilinarit permettent dcrire les  identits remarquables  suivantes :  

$(\overrightarrow{u}  \overrightarrow{v})^2 = \overrightarrow{u}^2 + \overrightarrow{v}^2  2 \overrightarrow{u} \overrightarrow{v} $ et   
$(\overrightarrow{u}  \overrightarrow{v})(\overrightarrow{u} + \overrightarrow{v}) = \overrightarrow{u}^2  \overrightarrow{v}^2$  

Que lon peut transposer avec les normes :  

$||\overrightarrow{u}  \overrightarrow{v}||^2 = ||\overrightarrow{u}||^2 + ||\overrightarrow{v}||^2  2 \overrightarrow{u} \overrightarrow{v} \quad$ et $(\overrightarrow{u}  \overrightarrow{v})(\overrightarrow{u} + \overrightarrow{v}) = ||\overrightarrow{u}||^2  ||\overrightarrow{v}||^2$  

<div class="alert alert-block alert-info">
    
<b>Colinarit et orthogonalit de deux vecteurs</b><br>
    
 Si $\overrightarrow{u}$ et $\overrightarrow{v}$ sont colinaires et de mme sens : $\overrightarrow{u} \overrightarrow{v} = ||\overrightarrow{u}|| \times ||\overrightarrow{v}||$<br>
 Si $\overrightarrow{u}$ et $\overrightarrow{v}$ sont colinaires et de sens contraires :  $\overrightarrow{u} \overrightarrow{v} = - ||\overrightarrow{u}|| \times ||\overrightarrow{v}||$<br>
 $\overrightarrow{u}$ et $\overrightarrow{v}$ sont orthogonaux si, et seulement si : $\overrightarrow{u} \overrightarrow{v} = 0$ <br>
 </div>
 
 
**Remarque :**   
Le vecteur nul est orthogonal  tout vecteur.  


**Exemple :**   
Soit les points $A(6 ; 8 ; 2)$, $B(4 ; 9 ; 1)$ et $C(5 ; 7 ; 3)$.
Montrer que le triangle $ABC$ est rectangle en $A$.  
  
## II)  Orthogonalit dans lespace
### 1)  Droites orthogonales
<div class="alert alert-block alert-danger">
    
Deux droites $d_1$ et $d_2$ de vecteurs directeurs $\overrightarrow{u_1}$ et $\overrightarrow{u_2}$ sont :<br>
 orthogonales si, et seulement si : $\overrightarrow{u_1}  \overrightarrow{u_2} = 0$.<br>
 perpendiculaires si et seulement si $d_1$ et $d_2$ sont orthogonales et scantes.
</div>


**Remarque :**  

On crit indistinctement $d_1 \perp d_2$ dans le deux cas.  

Dans lespace, on distingue droites  orthogonales  et droites perpendiculaires .  

Dans le cube :   
 les droites $d_1$ et $d_2$ sont orthogonales mais pas perpendiculaires.  
 les droites $D$ et $d_2$ sont perpendiculaires donc orthogonales. 

  


**Exemple :**  
Soit les points $A(2 ;5 ; 1)$ et $B(0 ; 2 ; 6)$.
Dmontrer que la droite d de vecteur directeur $\overrightarrow{u}(4 ; 1 ;3) $ est orthogonale  la droite $(AB)$.  

### 2) Droite et plan orthogonaux
<div class="alert alert-block alert-danger">
            
<b>Dfinition:</b><br>
Un plan $(P)$ de vecteurs directeurs $(\overrightarrow{u_1}, \overrightarrow{u_2})$ est orthogonal  une droite $d$ de vecteur directeur $\overrightarrow{v}$ si, et seulement si, $\overrightarrow{u_1} \overrightarrow{v} = 0$ et $\overrightarrow{u_2} \overrightarrow{v}= 0$
</div>




**Exemple :**  
Soit les points $A(2 ; 0 ; 2)$, $B(4 ; 0 ; 0)$, $C(1 ;2 ; 1)$, $D(1 ; 1 ; 0)$ et $E(1 ;1 ; 2)$.  

Le plan $(ABC)$ et la droite $(DE)$ sont-ils orthogonaux?  



### 3) Plans orthogonaux
<div class="alert alert-block alert-danger">
    
<b>Dfinition:</b><br>
Un plan $(P)$ est orthogonal  un plan $(Q)$ si, et seulement si, il
existe une droite d du plan $(Q)$ orthogonale au plan $(P)$.  
</div>
