## 4) Oprations sur les limites

F.I. $$<br>

## 6)  Convergence dune suite monotone
### 1) Suites majores, minores et bornes
<div class="alert alert-block alert-danger">
    <b>Dfinition 4 :</b><br>  
Soit une suite $(u_n)$  <br>
-  $(u_n)$ est majore si, et seulement si, il existe $M \in \mathbb{R}$ tel que : $\forall n \in \mathbb{N}, u_n \le M$.<br>
- $(u_n)$ est minore si, et seulement si, il existe $m \in \mathbb{R}$ tel que : $ \forall n \in \mathbb{N}, u_n \ge m$<br>
- $(u_n)$ est borne si, et seulement si, il existe $m, M \in \mathbb{R}$ tel que : $ \forall n \in \mathbb{N}, M \ge u_n \ge m$<br>

</div>

**Remarque :**  
Le majorant ou le minorant nest pas unique. En effet si 2 majore la
suite $(u_n)$ alors tout rel suprieur  2 majore aussi $(u_n)$.  


**Exemple :**  
Montrer que la suite $(u_n)$ dfinie sur $\mathbb{N^}$ par :  
$u_n=\dfrac{1}{n+1}+\dfrac{1}{n+2}+\ldots+\dfrac{1}{2n}$ est borne par lintervalle $[\dfrac{1}{2};1]$
