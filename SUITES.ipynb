### 2) Principe de récurrence (axiome)

$P(n)$ désigne une propriété mathématique sur un entier naturel $n$.
Pour démontrer que cette propriété est vraie quelque soit le rang $n$ (on dit: pour tout $n$ entier naturel), il faut procéder en 2 étapes.

<div class="alert alert-block alert-warning">
 Une des deux étapes est de prouver que si $P$ est vraie pour un rang donné $n$, alors elle est vraie pour le rang suivant $n+1$, il s'agit de <b>l'hérédité </b> de cette propriété.<br>
 L'autre étape est triviale! Car pour que la propriété soit vraie pour tout $n$, faut-il qu'il y ait un premier rang pour lequel cette propriété soit vraie. Il s'agit de <b>l'initialisation.</b><br>
  </div>  
  
<div class="alert alert-block alert-info">
<b>Démontrons que $1+2+... + n = \frac{n(n+1)}{2}$ .</b><br>


Pour $ n = 1$, on a $\frac{1(1+1)}{2}=1.$ Donc $P(1)$ est vraie.<br><br>


On suppose que $P(n)$ est vraie pour un rang $n$ donné.<br>
    On cherche à prouver que :
$ P(n)$ est vraie $\Longrightarrow P(n+1)$ est vraie<br>

C'est-à-dire que $ 1+2+...+n = \frac{n(n+1)}{2} \Longrightarrow 1+2+...+ (n+1) = \frac{(n+1)(n+2)}{2}$<br><br><br>

Remplaçons les n premiers termes de notre somme par $P(n)$: 
$$ 1+2+...+n+(n+1) = \frac{n(n+1)}{2}+(n+1)$$<br><br>
$$ 1+2+...+n+(n+1) = \frac{n(n+1)}{2} + \frac{2(n+1)}{2} $$<br><br>
$$ 1+2+...+n+(n+1) = \frac{n(n+1)+2(n+1)}{2}$$<br><br>
$$ 1+2+...+n+(n+1) = \frac{(n+1)(n+2)}{2}$$<br>
On en déduit que: $P(n)$ est vraie $\Longrightarrow$ $P(n+1)$ est vraie pour un rang donné $n$.<br><br>


<b>Comme $P(n)$ est initialisée et héréditaire, elle est vraie.</b> <br>

## 5) Situations amenant à une conclusion erronée
- Situation 1 :  Hérédité seulement vérifiée  

Soit la propriété suivante : $\forall n \in \mathbb{N}, 3 divise 2^n$  

Hérédité : on suppose que 3 divise $2^n$, montrons que 3 divise $2^{n+1}$.
Si 3 divise $2^n$, alors il existe un entier naturel $k$ tel que :   
$2^n = 3k$  

En multipliant par 2 : $2^{n+1} = 2 \times 3k = 3(2k)$. 3 divise donc $2^{n+1}$.    
  
$\quad$


- Situation 2 : Initialisation vérifiée jusqu’à un certain rang.  

Soit la propriété suivante : $\forall n \in \mathbb{N}\mbox{,} \quad n^2 − n + 41$ est un nombre premier.  

L’initialisation est vérifiée car pour $n = 0$ on obtient 41 qui un nombre premier.  

Mais l’hérédité n’est pas assurée bien que la propriété soit vraie jusqu’à
n = 40. On peut le vérifier avec une table de nombres premiers et la
liste des premiers termes de la suite $(U_n)$ définie par $U_n = n^2 − n + 41$.  

Pour $n = 41$, on a : $41^2 − 41 + 41 = 41^2$ qui n’est pas un nombre premier. La propriété est donc fausse

# III) Limite d’une suite
## 1) Limite finie 
<div class="alert alert-block alert-danger">

<b>Définition :</b><br>
On dit que la suite $(U_n)$ a pour limite $ℓ$ si, et seulement si,
tout intervalle ouvert contenant $ℓ$ contient tous les termes de la suite à partir d’un certain rang.<br>
On note alors : $$ \lim\limits_{n \to +\infty} U_n=ℓ$$ et l’on dit que la suite $(U_n)$ converge vers $ℓ$.  
</div>    

**Suites de références**  
Les suites définies pour tout entier naturel $n \neq 0$ par :  
$\Big(\frac{1}{\sqrt{n}}\Big)$, $\Big(\frac{1}{n}\Big)$, $\Big(\frac{1}{n^2}\Big)$, $\ldots$ , $\Big(\frac{1}{n^k}\Big)$ avec $k \in \mathbb{N}^*$, ont pour limite 0.


**Algorithme  :**   
Déterminer à partir de quel entier $n$, le terme $U_n$ est dans un
intervalle centré en $ℓ$ et de rayon $10^{−p}$.  

$$ (U_n): \quad \left \{ \begin{array} \\
 U_0 &=& 0,1 \\
U_{n+1} &=& 2U_n(1 − U_n) \\
\end{array}
\right.$$  
On admet que cette suite est croissante et converge vers 0,5.   
À partir de quel entier n le terme $(U_n)$ est-il dans l’intervalle ouvert centré en 0,5 et de rayon $10^{−p}$? 

et si vous êtes fort en pyhton, tentez de me tracer la représentation graphique de cette suite!! ;)

```python
from math import *
def lim(p):
    u = 0.1
    n = 0
    while abs(u- 0.5) >= 10**( -p):
        u = 2*u*(1-u)
        n = n +1
    return n
```

```python
lim(10)
```

## 2) Limite infinie
<div class="alert alert-block alert-danger">
<b>Définition 3 : </b><br>  
On dit que la suite $(U_n)$ a pour limite $+\infty$ (resp. $-\infty$) si, et seulement si, tout intervalle $]A ;+\infty[$ (resp. $] − \infty; B[$) contient tous les termes de la suite à partir d’un certain rang.<br>
On note alors : $$\lim \limits_{ n \to +\infty }U_n = + \infty \quad
    \mbox{(resp. }\quad \lim \limits_{n\to +\infty} U_n = − \infty)$$
On dit que la suite diverge vers $+\infty$ (resp. $−\infty$)
 </div>
 
**Remarque :**  Pour une limite $+\infty$, cette définition traduit l’idée que les termes de la suite arrivent à dépasser $A$, aussi grand soit-il.  

Une suite peut n’avoir aucune limite : $U_n = (−2)^n$. La suite diverge sans limite.   

**Suites de référence :**  Les suites définies pour tout entier naturel $n$ par :  
$\Big(\sqrt{n} \Big), \Big(n \Big), \Big(n^2\Big), \ldots , \Big(n^k \Big) \quad \mbox{avec} \quad k \in \mathbb{N^∗}$, ont pour limite $+\infty$.  
  
**Algorithme :**  Déterminer à partir de quel entier $n$, $U_n$ est supérieur à $10^p$.  
$$ (U_n): \quad \left \{ \begin{array} \\
 U_0 &=& -2 \\
U_{n+1} &=& \dfrac{4}{3}U_n +1 \\
\end{array}
\right.$$ 

On admet que $(U_n)$ est croissante et diverge vers $+\infty$. À partir de quel entier $n$, le terme $U_n$ est-il supérieur à $10^p$ ?  

La fonction en Python $lim(p)$ suivante permet de trouver $n$ avec **une boucle conditionnelle**.    


```python
def lim(p):
    u = -2
    n = 0
    while u <= 10**p:
        u = 4/3*u + 1
        n+=1
    return n
```

```python
lim(6)
```

## 3) Limites par comparaison et par encadrement
<div class="alert alert-block alert-danger">  
 <b>Théorème 2 :</b><br>
Soit trois suites $(U_n), (V_n), (W_n)$. Si à partir de $k \in \mathbb{N}$, on a :<br>
    
 <b>Théorème d’encadrement ou « des gendarmes »</b><br>  
$$\begin{array} \\ 
 v_n \leq  u_n   \leq w_n \\
\lim \limits_{n \to +\infty}v_n  = \lim \limits_{n \to +\infty}w_n = l \\
\end{array} \Bigg \}
 \Rightarrow \lim \limits_{n \to +\infty} U_n = l.$$<br>
<b> Théorèmes de comparaison   </b><br>
$$\begin{array} \\ 
 u_n \leq  v_n  \\
\lim \limits_{n \to +\infty}u_n  = + \infty \\
\end{array} \Bigg \}
 \Rightarrow \lim \limits_{n \to +\infty} v_n = +\infty.$$<br> 
 $$\begin{array} \\ 
 v_n \geq  w_n  \\
\lim \limits_{n \to +\infty}v_n  = - \infty \\
\end{array} \Bigg \}
 \Rightarrow \lim \limits_{n \to +\infty} w_n = -\infty.$$<br> 
</div>  

**Remarque :**  
Le théorème des gendarmes s’utilise pour montrer une limite finie
et les théorèmes de comparaison une limite infinie.  

  
**Exemples :**
- Démontrer que la suite $(u_n)$ définie par : $u_n =\dfrac{sin( n)}{n + 1} $ est convergente.   
  
$$\forall n \in \mathbb{N}, \dfrac{-1}{n + 1} \leq \dfrac{sin( n)}{n + 1} \leq \dfrac{1}{n + 1} $$  

or $\lim \limits_{n \to +\infty} \dfrac{-1}{n+1} = \lim \limits_{n \to +\infty} \dfrac{1}{n+1} = 0$  

Donc, d’après le théorème des gendarmes on a :   
$$\lim \limits_{n \to + \infty} u_n = 0$$  

• Montrer que la suite $(v_n)$ définie par : $v_n = n + sin (n)$ diverge vers $+\infty$.  
$$\forall \in \mathbb{N}, n + sin(n) \gt n-1$$

or $\lim \limits_{n \to +\infty} n-1 = +\infty$

Donc, d’après les théorèmes de comparaison, on a :   
$$\lim \limits_{n \to +\infty} v_n = +\infty$$  

## 4) Opérations sur les limites
Il est assez intuitif de penser que la limite de a somme, du produit ou du quotient est la somme, le produit ou le quotient des limites. Seuls 4 cas représentent des formes indéterminées et donc ne peuvent se résoudre par opération. Il faudra alors soit essayer de changer la forme de $u_n$, soit
utiliser les théorèmes de comparaison ou des gendarmes, soit le théorème sur les suites monotones (voir plus loin) pour pouvoir conclure.  
### a. forme indéterminée d'une somme

$$\begin{array} \\ 
 \lim \limits_{n \to +\infty} u_n = +\infty\\
 \lim \limits_{n \to +\infty} v_n = -\infty\\
\end{array} \Bigg \}
F.I. $$<br>  

**Exemples:**  
$U_n = 3n +1 +\dfrac{2}{n}$ avec $n \in \mathbb{N^*}$  
$V_n = \bigg( \dfrac{1}{3} \bigg)^n + 5 - \dfrac{1}{n}$ avec $n \in \mathbb{N^*}$   
$W_n = n^2 - n +2$ avec $n \in \mathbb{N}$


### b. forme indéterminée d'un produit

$$\begin{array} \\ 
 \lim \limits_{n \to +\infty} u_n = \infty\\
 \lim \limits_{n \to +\infty} v_n = 0\\
\end{array} \Bigg \}
F.I. $$<br>  

**Exemples:**  
$U_n = (2-n)\times 3^n$  
$V_n = \dfrac{1}{n+1} \times (n^2 +3)$ 

### c. forme indéterminée d'un quotient
$$\begin{array} \\ 
 \lim \limits_{n \to +\infty} u_n = 0\\
 \lim \limits_{n \to +\infty} v_n = 0\\
\end{array} \Bigg \}
F.I. $$<br>  

$$\begin{array} \\ 
 \lim \limits_{n \to +\infty} u_n = \infty\\
 \lim \limits_{n \to +\infty} v_n = \infty \\
\end{array} \Bigg \}
F.I. $$<br>

 
**Exemples:**   
$U_n = \dfrac{5}{2n^2+1}$    
  
$V_n = \dfrac{1-n}{0,5^n}$   
  
$W_n = \dfrac{n^2+3}{n+1}$


## 5) Limite d’une suite géométrique
<div class="alert alert-block alert-danger">
<b>Théorème 3 : </b>  
 Soit q un réel. On a les limites suivantes : <br> 
- Si $q \gt 1$ alors $\lim \limits_{n \to +\infty} q^n = +\infty$<br>
- Si $q = 1$ alors $\lim \limits_{ n \to +\infty} q^n = 1$<br>
- Si $- 1 \lt q \lt 1$  alors $\lim \limits_{n \to +\infty} q^n = 0$<br>
- Si $q \lt −1$ alors $\lim \limits_{n \to +\infty} q^n $ n'existe pas.<br>
</div>  

<div class="alert alert-block alert-info">  
<b>Démonstration :</b><br>  <br>
<b>Pour $q \gt 1$</b><br>
D’après l’inégalité de Bernoulli, on a, pour $a \gt 0$:<br>
$$\forall n \in \mathbb{N}, (1 + a)^n \ge 1 + na$$<br>
$q \gt 1$ donc il existe $a \gt 0$ tel que $q = 1 + a$.<br> 
 L’inégalité devient : $q^n \ge 1 + na$. Comme $a \gt 0$ on a :<br>  
    $$\lim \limits_{n \to =\infty} 1+na = +\infty$$

D’après le théorèmes de comparaison on a :<br>
    $$\lim \limits_{n \to +\infty} q^n = +\infty$$
</div>  

**Remarque :**  
Pour $−1 \lt q \lt 1$ et $q \neq 0$, on peut poser $Q = \dfrac{1}{|q|}$, donc $Q \gt 1$.  
On revient alors à la première limite et l’on conclut avec le quotient sur les limites.  


**Exemple:**  

Soit une suite $(u_n)$ définie par :  
$$ \left \{ \begin{array} \\
 u_0 &=& 2 \\
u_{n+1} &=& 2u_n + 5 \\
\end{array}
\right.$$ 

On pose la suite $(v_n)$ telle que $v_n = u_n + 5$
1) Montrer que la suite $(v_n)$ est géométrique  
2) Exprimer $v_n$ puis $u_n$ en fonction de n  
3) En déduire la limite de $(u_n)$


## 6)  Convergence d’une suite monotone
### 1) Suites majorées, minorées et bornées
<div class="alert alert-block alert-danger">
    <b>Définition 4 :</b><br>  
Soit une suite $(u_n)$  <br>
-  $(u_n)$ est majorée si, et seulement si, il existe $M \in \mathbb{R}$ tel que : $\forall n \in \mathbb{N}, u_n \le M$.<br>
- $(u_n)$ est minorée si, et seulement si, il existe $m \in \mathbb{R}$ tel que : $ \forall n \in \mathbb{N}, u_n \ge m$<br>
- $(u_n)$ est bornée si, et seulement si, il existe $m, M \in \mathbb{R}$ tel que : $ \forall n \in \mathbb{N}, M \ge u_n \ge m$<br>

</div>

**Remarque :**  
Le majorant ou le minorant n’est pas unique. En effet si 2 majore la
suite $(u_n)$ alors tout réel supérieur à 2 majore aussi $(u_n)$.  


**Exemple :**  
Montrer que la suite $(u_n)$ définie sur $\mathbb{N^∗}$ par :  
$u_n=\dfrac{1}{n+1}+\dfrac{1}{n+2}+\ldots+\dfrac{1}{2n}$ est bornée par l’intervalle $[\dfrac{1}{2};1]$

### 2) Théorèmes de divergence et de convergence
<div class="alert alert-block alert-danger"> 
 <b>Théorème 4 :Divergence</b>  <br>  
 Soit $(u_n)$ une suite.<br>
- Si $(u_n)$ est croissante et non majorée alors $(u_n)$ diverge vers $+\infty$.<br>
- Si $(u_n)$ est décroissante et non minorée alors $(u_n)$ diverge vers $-\infty$.<br>

 </div>     
 
**Exemple :**  
Soit la suite $(u_n)$ définie par : $u_0 = 1$ et $u_{n+1} = u_n + 2n + 3$.  
$ \forall n \in \mathbb{N}, u_{n+1} − u_n = 2n + 3 \gt 0$.  
La suite $(u_n)$ est croissante.
Montrons par récurrence que $\forall n \in \mathbb{N}, u_n > n^2$.  


<div class="alert alert-block alert-danger">  
    <b>Théorème 5 : Convergence</b><br>
- Si une suite $(u_n)$ est croissante et majorée alors la suite $(u_n)$ converge.<br>
- Si une suite $(u_n)$ est décroissante et minorée alors la suite $(u_n)$ converge.
 </div>
 
**Exemple :**  
Soit la suite $(u_n)$ définie par :  

$u_0 = 1$
$u_{n+1} = \sqrt{2 + u_n}$  

On a montré par récurrence précédemment que $\forall n \in \mathbb{N}, 0 < u_n < u_{n+1} < 2$.  

La suite $(u_n)$ est croissante et majorée par 2,   
donc d’après le théorème des suites
monotones,   la suite $(u_n)$ est convergente vers $0 \lt ℓ \lt 2$. 


**Algorithme :**  
On admet que la suite un converge vers 2, déterminer à l’aide d’un  
algorithme, l’entier $n$ à partir duquel $u_n > 1, 999$.  


```python
from math import ∗

def lim(p) :
    u = 1
    n = 0
    while abs(u−2)>=10∗∗(−p):
        u = sqrt(u+2)
        n+=1
    return n
```
