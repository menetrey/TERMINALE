## 1) Fonctions sinus et cosinus
### a) Définition
<div class="alert alert-block alert-danger">
 À tout réel $x$, on associe un point unique M du cercle unité ou cercle trigonométrique de centre O, dont les coordonnées sont :<br>
    $$M(cosx; sinx)$$ 
</div>

<div class="alert alert-block alert-danger">
On appelle fonctions sinus et cosinus les fonctions respectives : $$x \to sin x \mbox{  et  } x \to cos x$$  
 </div>  
 
**Remarque:**   
$$-1 \leq cos(x) \leq 1$$  
$$ -1 \leq sin(x) \leq 1$$  


### b) Propriétés

<div class="alert alert-block alert-danger">
 <b> Théorème sur la parité:</b><br>
  D’après les formules de trigonométrie,<br>
- La fonction sinus est impaire : $\forall x \in \mathbb{R} \quad sin(−x) =−sin(x) $<br>  
- La fonction cosinus est paire :$ \forall x \in \mathbb{R} \quad cos(−x) =cos(x)$<br>
</div>

<div class="alert alert-block alert-danger">
<b> Théorème sur la périodicité</b><br>
D’après la définition des lignes trigonométriques dans le cercle,les fonctions sinus et cosinus sont $2\pi$ périodiques :    $T=2 \pi $ <br>
    
 $$\forall x \in \mathbb{R}, \quad sin(x+2\pi) =sin(x) \quad \mbox{     et    } \quad  cos(x+2 \pi) =cos(x)   $$  
</div>    
 
 
<div class="alert alert-block alert-danger">
<b>de cosinus à sinus</b><br>
 D’après les formules de trigonométrie, on a :<br><br>
    $$sin\Big(\dfrac{\pi}{2}−x\Big)=cos(x) \quad \mbox{et} \quad  cos\Big(\dfrac{\pi}{2}−x\Big)=sin(x)$$   
</div>

**Exemple:**  
Résoudre dans l’intervalle $]−\pi \ ;\pi]$, l’équation suivante : $$sin\Big(x+\dfrac{\pi}{4}\Big)=cos(x)$$


## 2) Etude des fonctions sinus et cosinus
### a) Dérivées
<div class="alert alert-block alert-danger">
Les fonctions sinus et cosinus sont dérivables sur $\mathbb{R}$:<br>
    $$sin′(x)=cos(x) \quad \mbox{et} \quad   cos′(x)=−sin(x)$$  
</div> 
  
  
**Exemple:**  
Déterminer la dérivée de la fonction suivante :  
$$f(x) =cos(2x)+cos^2(x)$$
  
  
### b) Applications aux calculs de limites
<div class="alert alert-block alert-danger">
D’après les fonctions dérivées des fonctions sinus et cosinus, on a :<br>
    $$\lim \limits_{x\to 0} \dfrac{sinx}{x}=1 \quad \mbox{et} \quad \lim \limits_{x \to 0}\dfrac{cos(x)-1}{x}=0$$  
</div>

## II) Application aux ondes progressives

### a) ondes sonores
Un son pur est une onde sinusoïdale caractérisée par :  
 - Sa fréquence F (en Hertz, nombre de pulsations par seconde) qui détermine lahauteur du son.   
- Son amplitude (pression acoustique) P (en Pascal).  

La fréquence $F$ est relié à la période $T$ de la sinusoïde par la relation : $F=\dfrac{1}{T}$. La fonction $f$ associée est donc de la forme :  $$f(t) =P sin(2\pi Ft)$$

La note de référence (donnée par un diapason) sur laquelle s’accordent les instruments de l’orchestre est le $la_3$ qui vibre à 440 Hz. Pour une amplitude de 1 Pa,cette note peut être associé à la fonction $f$ définie par : $$f(t) =sin(880  \pi t)$$.

### b) Harmoniques
Une bonne technique pour analyser les ondes a été conçu en 1807 par le physicien français Jean-Baptiste Fourier. Il a établi que toute onde rencontrée dans la nature, qu’elle soit une impulsion ondulatoire ou une onde périodique entretenue peut être considérée comme résultant de la superposition d’ondes sinusoïdales. Cela peut se réaliser, dans le cas du son, par un analyseur de spectre et, dans le cas de la lumière, par un prisme.   
Selon Fourier, toute fonction périodique de fréquence F peut être considérée comme une somme de termes sinusoïdaux avec des amplitudes et des phases appropriées.   
Le premier d’entre eux a la même fréquence ($F_1=F$), c’est le **fondamental** ou le premier harmonique. Le terme suivant, de fréquence $F_2=2F$ est appelé deuxième harmonique puis vient le troisième terme de fréquence $F_3=3F$, appelé troisième harmonique et ainsi de suite. Notons que, pendant le temps $(\dfrac{1}{F_1})$ que met le fondamental pour décrire un cycle complet, le deuxième harmonique a décrit deux cycles et le $n^e$ harmonique $n$ cycles.

**Exemple:**   
Le signal en "dents de scie", une des formes d’ondes fréquemment utilisées pour la synthèse sonore, a pour expression :
  
$$f_n(t)=\dfrac{2}{n}\sum_{k=1}^n \dfrac{sin(2 \pi kFt +(k-1)\pi)}{k}$$  avec n qui tend vers l'infini.
