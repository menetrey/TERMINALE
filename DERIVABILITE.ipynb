# I)  Rappels sur la dérivabilité
## 1)  Définition
<div class="alert alert-block alert-danger">
Soit une fonction $f$ définie sur un intervalle ouvert $I$ contenant $a$.<br>
La fonction $f$ est dérivable en $a$ si et seulement si le taux d’accroissement de $f$ en $a$ admet une limite finie $ℓ$, c’est à dire :<br>
    $$\lim_{h \to 0} \dfrac{f (a + h) − f (a)}{h} = ℓ$$
    ou encore <br>
    $$\lim_{x \to a}  \dfrac{f (x) − f (a)}{x − a} = ℓ$$
    


Dans ce cas, on appelle $ℓ$ le nombre dérivé de $f$ en $a$ et on le note $f ′(a)$.  

 </div>
Lorsque la fonction $f$ est dérivable sur un intervalle $I$, on note $f ′$, la fonction dérivée qui à tout $x$ de $I$ associe son nombre dérivée $f ′(x)$.  

**Remarque :**  
• Si la fonction $f$ est dérivable en $a$ alors la fonction $f$ est continue en $a$  

• Les physiciens expriment volontiers une variation à l’aide des symboles $\Delta$ et $d$. Ils notent $\Delta x = x−a$ et $\Delta y = f (x)− f (a)$ ; ou $dx$ et $dy$ pour une variation très petite. On obtient alors la notation différentielle de la dérivée :   

$$f ′ =\dfrac{dy}{dx}$$ et $$f ′(a) =\dfrac{dy}{dx}(a)$$  

**Exemple :**  
Soit la fonction $f$ définie par :  
$$
\begin{cases} 
f (x) &=& x^2 − 2x − 2 &, \mbox{si }x \le 1 \\
f (x) &=& \dfrac{x − 4}{x}&, \mbox{si }x \gt 1 \\
\end{cases}$$


• Continuité de f en 1 : $$\lim_{x \to1−} f (x) = \lim_{x \to 1+} f (x) = f (1) = −3$$  


• Dérivabilité en 1 :  
on étudie les limites à gauche et à droite du taux de variation en 1 :
$ x \le 1 $

$$\begin{array}\\
\dfrac{f (1 + h) − f (1)}{h}&=& \dfrac{(1 + h)^2 − 2(1 + h) − 2 − (−3)}{h}=\dfrac{h^2}{h}= h \quad \lim_{h \to 0} h = 0\\  
 \dfrac{f (1 + h) − f (1)}{h}&=&\dfrac{\dfrac{1 + h − 4}{1 + h}+ 3}{h}=\dfrac{4 h}{h(1 + h)}=\dfrac{4}{1 + h} \quad  \lim_{h \to 0} \dfrac{4}{1+h}=4\\
\end{array}
$$

Donc les limites ne sont pas égales donc la fonction f n’est pas dérivable en 1.  

Graphiquement la fonction $f$ est en un seul morceau (continue en 1) et possède un point anguleux en 1 (non dérivable en 1).

### a) Interprétation graphique

<div class="alert alert-block alert-danger">

Lorsque $f$ est dérivable en $a$, la courbe représentative $\mathcal{C_f}$ de la fonction $f$ admet au point $A(a, f (a))$ une tangente de coefficient directeur $f ′(a)$ dont l’équation est :  
    $$(T) : y = f ′(a)(x − a) + f (a)$$
</div>

    
    
    
    
    
    
**Remarque :** Retenir que $f ′(a)$ représente le coefficient directeur de la tangente.   

### c) Interprétation numérique
<div class="alert alert-block alert-danger">
<b>Théorème :</b><br>
 Si une fonction $f$ est dérivable en $a$, une approximation affine,
lorsque $a + h$ est voisin de $a$ est : $f (a + h) ≈ f (a) + h f ′(a)$.<br>
 </div>
 
 
**Exemple :**  
Déterminer une approximation affine de $\sqrt{4,03}$.  
Soit $f (x) = \sqrt{x}$, donc $f ′(x) =\dfrac{1}{2\sqrt{x}}$  avec $a = 4$ et $h = 0,03$.
$\sqrt{4,03} \approx f (4) + 0,03 f ′(4) \approx 2 + 0,03 \times \dfrac{1}{4} \approx 2, 007 5$  

La calculatrice donne $\sqrt{4,03} \approx 2, 007\ 486$, la précision est donc de $10^{−4}$.


## 3) Interprétation cinématique
Si $x(t)$ est la loi horaire d’un mouvement sur $(Ox)$, alors $x′(t)$ représente la vitesse instantanée à l’instant $t$. De même, si $v(t)$ est la vitesse instantanée à l’instant $t$, alors $v′(t)$ représente l’accélération a à l’instant $t$.  

En notations différentielles :   
$v =\dfrac{dx}{dt}$ et $a =\dfrac{dv}{dt}=\dfrac{d^2x}{dt^2}$ .  

 
# 3)  Signe de la dérivée, sens de variation
<div class="alert alert-block alert-danger">
<b>Théorème  :</b><br> Soit une fonction $f$ dérivable sur un intervalle $I$.<br>
• Si $f ′ = 0$, alors $f$ est constante.   <br>
• Si $f ′ \gt 0$ (sauf en quelques points isolés de $I$ où $f ′$ s’annule), alors $f$ est strictement croissante sur $I$. <br>  
• Si $f ′ \lt 0$ (sauf en quelques points isolés de $I$ où $f ′$ s’annule), alors $f$ est strictement décroissante sur $I$. <br> 
</div>

**Remarque :**   
L’étude des variations d’une fonction dérivable consiste à étudier le
signe de la dérivée.  

**Exemple :** Soit la fonction $f$ définie sur $\mathbb{R}$ par :  
$$f (x) = x^3 − 6x^2 + 1$$.

• $f$ est dérivable sur $\mathbb{R}$ et $f ′(x) = 3x^2 − 12x = 3x(x − 4)$.  
• $f ′(x) = 0 \iff x_1 = 0 \mbox{ ou } x_2 = 4$   
• $f ′(x)$ est positive à l’extérieur des racines et négative à l’intérieur.   
  
  Proposer le tableau de signe de $f'$ et de variation de $f$.
  
  
## 4) Dérivée et extremum
<div class="alert alert-block alert-danger">

<b>Théorème  :</b><br>  
Soit une fonction $f$ dérivable sur $I$ ouvert contenant $a$. <br> 
    
• Si $f$ admet un extremum en $a$ alors $f ′(a) = 0$.<br>
• Si $f ′(a) = 0$ et si $f ′$ change de signe en $a$ alors $f$ admet un extremum en $a$.<br>
</div>

**Remarque :**  
Les extremum d’une fonction sont à chercher parmi les racines de $f ′$,
mais $f ′(a) = 0$, n’implique pas nécessairement un extremum en $a$.  

Si $f (x) = x^3$ alors $f ′(x) = 3x^2$ donc $f ′(0) = 0$ et 0 n’est pas extremum.  


**Conséquence:**  
Les problèmes d’optimisation consistent à déterminer une fonction
dérivable et à déterminer les extremum locaux.   

# II) Dérivée d’une fonction composée
## 1) Composition de deux fonctions

<div class="alert alert-block alert-danger">
Soient $u$ et $v$ deux fonctions définies respectivement sur $I$ et $J$ tels
que $u(I) \subset J$, c’est à dire que pour tout $x \in I, u(x) \in J$.<br>
On appelle composée de $u$ par $v$, la fonction notée $v \circ u$, définie sur I par :<br>
    $$v ◦ u(x) = v [u(x)]$$  
    
</div>


**Exemple :**     
$u(x) = 3x−4$ et $v(x)=\sqrt{x}$ alors $v \circ u(x) = v(3x−4) = \sqrt{3x − 4}$  

$v$ est définie sur $J=R^{+}$, la condition sur $I$ est $3x − 4\gt0$ soit 
$I =]\dfrac{4}{3};+\infty [$  

On peut faire le schéma suivant :   
$x  \rightarrow 3x − 4 \rightarrow \sqrt{3x − 4}$  

    
**Attention:**  
La composition de deux fonctions n’est pas commutative en effet :  
$u(x) = 3x−4$ et $v(x)=\sqrt{x}$ alors $u \circ v(x) = u\big(\sqrt{x}\big) = 3\sqrt{x} − 4$  


## 2) Monotonie d’une fonction composée

<div class="alert alert-block alert-danger">
Soit les fonctions $u$ et $v$ définies respectivement sur $I$ et $u(I)$.<br>
• Si $u$ et $v$ ont même variation alors $v \circ u$ est croissante sur $I$.<br>
• Si $u$ et $v$ ont des variations opposés alors $v \circ u$ est décroissante sur $I$.
    </div>
    
<div class="alert alert-block alert-info">
<b>Démonstration : </b><br>
Dans le cas où u croissante sur $I$ et $v$ décroissante sur $u(I)$.<br>
$u$ est croissante :<br>  pour tout $x_1, x_2 \in I, x_1 \lt x_2 \implies u(x_1) \lt u(x_2)$<br><br>
$v$ est décroissante :<br>  pour tout $u(x_1), u(x_2) \in u(I), u(x_1) \lt u(x_2) \implies v[u(x_1)] \gt v[u(x_2)]$<br><br>
On a donc :  <br>Pour tout $x_1, x_2 \in I, x_1 \lt x_2 \implies v[u(x_1)] \gt v[u(x_2)]$<br>
La fonction $v \circ u$ est décroissante sur $I$.<br>
 </div>
 
**Exemple :**  
Soit $f$ définie sur $] − \infty; 1]$ par : $f (x) = \sqrt{1 − x}$.   
La fonction $f$ se décompose en $v \circ u$, avec : $u(x) = 1 − x$ et $v(x) = \sqrt{x}$  

La fonction $u$ est affine décroissante sur $]−\infty ; 1]$ et $v$ est croissante sur $[0 ;+\infty[$  

d’après la monotonie des fonctions composées, f est décroissante sur $] − \infty ; 1]$  
Il n’est donc pas nécessaire pour ce cas de déterminer le signe de la dérivée
pour connaître les variations de la fonction f .   


## 3) Dérivée de la composée
<div class="alert alert-block alert-danger">
Soit $u$ et $v$ deux fonctions dérivables respectivement que $I$ et $J$
tels que $u(I) ⊂ J$, alors la fonction $v \circ u$ est dérivable sur $I$ et : $$(v \circ u)′ = u′ \times v′ \circ u$$ 
</div>

    
**Exemple :**  
Soit la fonction $f$ définie sur $\mathbb{R}$ par : $f (x) = cos(x^2 − 1)$.   
• On décompose la fonction $f = v \circ u$ avec $u(x) = x^2 − 1$ et $v(x) = cos x$  
• $u$ et $v$ sont dérivable sur $\mathbb{R}$ donc $f$ est dérivable sur $\mathbb{R}$.  

• $f ′(x) = u′(x) \times v′ \circ u(x) = 2x[−sin(x^2 − 1)] = −2x sin(x^2 − 1)$  
<div class="alert alert-block alert-info">  
<b>Démonstration :</b><br>
L’idée de la démonstration est d’étudier la limite du taux de variation de $f = v \circ u$ au voisinage d’un point $a \in I$ avec pour tout $x \neq a, u(x) \neq u(a)$.  <br>
$$\dfrac{f (x) − f (a)}{x − a}=\dfrac{v \circ u(x) − v \circ u(a)}{x − a } = \dfrac{v[u(x)] − v[u(a)]}{ u(x) − u(a) }\times \dfrac{u(x) − u(a)}{ x - a }$$  


d’où $(v \circ u)′(a) = v′[u(a)] \times u′(a)$  
</div>

# IV)  Convexité
## 1) Définitions
<div class="alert alert-block alert-danger">  
Soit une fonction $f$ de courbe représentative $C_f$ .<br>
• On appelle sécante de $C_f$ la droite passant par deux points de $C_f$ .<br>
• On dit que $f$ est convexe sur $I$ si $C_f$ est en dessous de ses sécantes sur $I$.<br>
• On dit que $f$ est concave sur $I$ si $C_f$ est au dessus de ses sécantes sur $I$.
</div>





**Exemples :**  
La fonction inverse $x \to \dfrac{1}{x}$ est concave sur $\mathbb{R}^{∗−}$ et convexe sur $\mathbb{R}^{∗+}$.  
Les fonctions carrée et exponentielle $x \to x^2$ et $x \to e^x$ sont convexes sur $\mathbb{R}$. 
  
La fonction racine $x \to \sqrt{x}$ est concave sur $\mathbb{R}^+$.  




## 2) Dérivée seconde
<div class="alert alert-block alert-danger">
Soit une fonction $f$ deux fois dérivable sur $I$.<br>
On appelle dérivée seconde de $f$ , notée $f′′$, la fonction dérivée de $f ′$ :<br>
    $$f ′′ = ( f ′)′.$$
</div>


**Remarque :**  
Le signe de la dérivée seconde donne les variations de la fonction $f ′$.  

• Si $f′′ \gt 0$ l’accroissement de $f$ augmente avec $x$ : cas de la fonction $x \to e^x$.  

• Si $f′′ \lt  0$ l’accroissement de $f$ diminue avec $x$ : cas de la fonction $x \to \sqrt{x}$  


En cinématique, la dérivée seconde de la loi horaire donne l’accélération.  

**Exemple :**  
Soit la fonction $f$ définie que $\mathbb{R}$ par : $f (x) = 2x^3 + 5x^2 − 3x − 1$.  

$f$ est deux fois dérivable sur $\mathbb{R}$ : $f ′(x) = 6x^2 + 10x − 3$ et $f ′′(x) = 12x + 10$  

<div class="alert alert-block alert-danger">
Soit $f$ une fonction deux fois dérivable sur $I$.<br>
• $f$ est convexe sur $I$ si, et seulement si, $f ′′ \gt 0$.<br>
• $f$ est concave sur $I$ si, et seulement si, $f ′′ \lt 0$.
</div>

  

**Exemple :**  
Reprenons : $f (x) = 2x^3 + 5x^2 − 3x − 1$, on a : $f′′(x) = 12x + 10$.  

• $f′′(x) = 0 \iff 12x + 10 = 0 \iff x = −\dfrac{5}{6}$  

• Si $x > −\dfrac{5}{6} \iff f′′(x) \gt 0$, la fonction $f$ est convexe.  
  
• Si $x < −\dfrac{5}{6} \iff f′′(x) \lt 0$, la fonction $f$ est concave.

<div class="alert alert-block alert-danger">
<b>Théorème : Tangentes et dérivée seconde</b><br>
Soit une fonction $f$ deux fois dérivables sur $I$ de courbe représentative $C_f$ .<br>
• Si $f ′′ \gt 0$, la courbe $C_f$ est au dessus de ses tangentes. <br>
• Si $f ′′ \lt 0$, la courbe $C_f$ est en dessous de ses tangentes.<br>
</div>

<div class="alert alert-block alert-info">
<b>Démonstration :</b><br> 
Dans le cas où $f ′′ \gt 0$.<br>
Soit $a \in I$, montrons que $C_f$ est au dessus de la tangente $T_a$ en $a$.<br>
$$T_a : y = f ′(a)(x − a) + f (a)$$,
    posons $\varphi(x) = f (x) − [ f ′(a)(x − a) + f (a)]$. <br> 
    
$\varphi$ est dérivable sur $I$ par opérations sur des fonctions dérivables.
$$\varphi(x) = f (x) − f ′(a)x + a f ′(a) − f (a) \implies \varphi′(x) = f ′(x) − f ′(a)$$ 
    
$f′′ \gt 0$ donc la fonction $f ′$ est croissante sur $I$ donc :<br>
    
• Si $x \gt a \implies f ′(x) \gt f ′(a) \implies \varphi′(x) \gt 0$.<br>
    
• Si $x \lt a \implies f ′(x) \lt f ′(a) \implies \varphi′(x) \lt 0$.<br>
    
• $\varphi(a) = f (a) − f (a) = 0$  
$$\forall x \in I, \varphi(x) \gt 0$$. 
    
La courbe $C_f$ est au-dessus de $T_a$.
</div>    


<div class="alert alert-block alert-danger">
<b> Point d’inflexion</b><br>  
    
Soit $f$ une fonction deux fois dérivable sur $I$ de courbe représentative $C_f$ .<br>
Soit A un point de $C_f$ de tangente $T_a$.<br>
• On dit que $C_f$ admet un point d’inflexion en $A$ si la tangente $T_a$ traverse $C_f$ .<br>
• Si $f′′(a) = 0$ en changeant de signe alors $C_f$ admet un point d’inflexion en $a$.
</div>


**Remarque :**  
Le fait que la tangente $T_a$ traverse la courbe $C_f$ cela signifie que $T_a$
change de position par rapport à $C_f$ (au dessus puis en dessous ou inversement).  

Si $f ′′$ change de signe en $a$ alors $f ′$ ne change pas de signe en $a$.
