# IV)  Convexit
## 1) Dfinitions
<div class="alert alert-block alert-danger">  
Soit une fonction $f$ de courbe reprsentative $C_f$ .<br>
 On appelle scante de $C_f$ la droite passant par deux points de $C_f$ .<br>
 On dit que $f$ est convexe sur $I$ si $C_f$ est en dessous de ses scantes sur $I$.<br>
 On dit que $f$ est concave sur $I$ si $C_f$ est au dessus de ses scantes sur $I$.
</div>





**Exemples :**  
La fonction inverse $x \to \dfrac{1}{x}$ est concave sur $\mathbb{R}^{}$ et convexe sur $\mathbb{R}^{+}$.  
Les fonctions carre et exponentielle $x \to x^2$ et $x \to e^x$ sont convexes sur $\mathbb{R}$. 
  
La fonction racine $x \to \sqrt{x}$ est concave sur $\mathbb{R}^+$.
