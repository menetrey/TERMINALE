### a) Somme de deux vecteurs  

La somme de deux vecteurs est dfinie par la relation de Chasles :  
$\overrightarrow{AC} = \overrightarrow{AB} + \overrightarrow{BC}$  

Ingalit triangulaire : $|| \overrightarrow{u} +\overrightarrow{v}|| \le ||\overrightarrow{u}||  + ||\overrightarrow{v}||$  


Construction de la somme de deux vecteurs de mme origine. On trace un paralllogramme, afin de reporter le deuxime vecteur permettant dappliquer la relation de Chasles.  



<div class="alert alert-block alert-danger">
<b>Proprit  :</b> Somme de deux vecteurs :<br>
 Commutativit : $\quad \quad \overrightarrow{u} + \overrightarrow{v} = \overrightarrow{v} + \overrightarrow{u}$<br>
 Associativit : $\quad \quad  (\overrightarrow{u} +\overrightarrow{v}) + \overrightarrow{w} = \overrightarrow{u} + (\overrightarrow{v} + \overrightarrow{w}) = \overrightarrow{u} + \overrightarrow{v} + \overrightarrow{w}$<br>

 Oppos $\quad - \overrightarrow{u}$ : $\quad \overrightarrow{AB} = \overrightarrow{BA}$  
 </div>   
 
### b) Multiplication dun vecteur par un scalaire
Lorsquon multiplie un vecteur par un rel $k$, appel scalaire, le vecteur ainsi form $k\overrightarrow{u}$ est tel que :      
 Sa longueur est multipli par $|k|$  
 Si $k \gt 0$ son sens est inchang  
 Si $k \lt 0$ son sens est invers.  
 Si $k = 0$ on a : $0\overrightarrow{u} = 0$      

<div class="alert alert-block alert-danger">
<b>Proprit :</b> Bilinarit.<br>
La multiplication par un scalaire est distributive par rapport  laddition de deux vecteurs ou la somme de deux rels.  
    $$ k(\overrightarrow{u} + \overrightarrow{v}) = k \overrightarrow{u} + k \overrightarrow{v}$$  <br>
    $$ (k + k) \overrightarrow{u} = k \overrightarrow{u} + k\overrightarrow{u}$$  
</div>    
    
## 3) Colinarit de deux vecteurs
<div class="alert alert-block alert-danger">
<b>Dfinition  :</b><br>
On dit que deux vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ sont colinaires, si et seulement si, il existe un rel $k$ tel que : <br>
    $\overrightarrow{v} = k \overrightarrow{u}$
</div>


**Remarque :**  
Le vecteur nul $\overrightarrow{0}$ est colinaire  tout vecteur car : $\overrightarrow{0} = 0 \overrightarrow{u}$  

<div class="alert alert-block alert-danger">
<b>Proprit :</b>  
La colinarit permet de montrer le paralllisme et lalignement.<br>
$\overrightarrow{AB} \quad \mbox{et} \quad \overrightarrow{CD} \quad \mbox{ colinaires} \quad \iff (AB) \ // \ (CD)$<br>
$\overrightarrow{AB} \quad \mbox{et} \quad \overrightarrow{AC} \quad \mbox{colinaires} \quad  \iff A, B, C \quad \mbox{aligns}$    
</div>    
    
**Exemple :**    

Soit ABC un triangle, E, I et F tels que :  
$\overrightarrow{AE} = \dfrac{1}{3} \overrightarrow{BC},\quad  \overrightarrow{CI} =\dfrac{2}{3}\overrightarrow{CB} \quad \mbox{et} \quad \overrightarrow{AF}=\dfrac{1}{3}\overrightarrow{AC}$  . 

Dmontrer que I, E et F sont aligns.



Exprimons $\overrightarrow{EI}$ et $\overrightarrow{EF}$ en fonction de $\overrightarrow{AB}$ .  

 d'aprs l'nonc $\overrightarrow{CI} =\dfrac{2}{3} \overrightarrow{CB}$ donc $\overrightarrow{BI} = \dfrac{1}{3} \overrightarrow{BC}$  . 

On en dduit que $\overrightarrow{AE} = \overrightarrow{BI}$ donc que AEIB est un paralllogramme. On a alors : $\overrightarrow{EI} = \overrightarrow{AB}$  
  
 De plus :  $ \overrightarrow{EF} = \overrightarrow{EA} + \overrightarrow{AF} =
\dfrac{1}{3}\overrightarrow{CB} + \dfrac{1}{3} \overrightarrow{AC} = \dfrac{1}{3} ( \overrightarrow{AC} + \overrightarrow{CB}) = \dfrac{1}{3} \overrightarrow{AB}$  

On en dduit alors : $\overrightarrow{EF} = \dfrac{1}{3} \overrightarrow{EI}$ .  
Les vecteurs $\overrightarrow{EF}$ et $\overrightarrow{EI}$ sont colinaires et donc les points E, F et I sont aligns.

## 5) Ensemble de points
<div class="alert alert-lbock alert-danger">
Soient deux points A et B et leur milieu I, pour tout point M on
a la relation :<br>
    $$
\overrightarrow{MA} \bullet \overrightarrow{MB} = MI^2  \dfrac{1}{4}AB^2$$  
</div>


**Dmonstration :**  
On introduit le point I dans le produit scalaire :  
$$\begin{array} \\
\overrightarrow{MA} \bullet \overrightarrow{MB}  
&= &   (\overrightarrow{MI} + \overrightarrow{IA}) \bullet (\overrightarrow{MI} + \overrightarrow{IB} )  \\
&=& MI^2 + \overrightarrow{MI}  \overrightarrow{IB} + \overrightarrow{IA} \bullet \overrightarrow{MI} + \overrightarrow{IA} \bullet \overrightarrow{IB} \\
&=& MI^2 +\overrightarrow{MI} (\overrightarrow{IA} + \overrightarrow{IB})  IA^2 \\
&=& MI^2  \dfrac{1}{4}AB^2 \end{array}$$


**Exemple :**  
Dterminer lensemble (E) des points M tel que : $\overrightarrow{MA} \bullet \overrightarrow{MB} = 7$ et $AB = 6$.  

$$\begin{array}{l l}
\overrightarrow{MA} \bullet \overrightarrow{MB} &=& 7 \\
\iff MI^2 \dfrac{1}{4}AB^2 &=& 7 \\
\iff MI^2 &=& 7 + \dfrac{1}{4} AB^2 \\
\iff &=& 7 + 9 \\
\iff &=& 16 \\
\iff MI^2 &=& 4^2 \end{array}$$

Lensemble $(E)$ est le cercle de centre $I$ et de rayon 4.

Dbuts de carrs parfaits : $(x  1)^2  1 + (y  1)^2  1 = 2$  
On obtient alors : $(x  1)^2 + (y  1)^2 = 4 = 2^2$  
Lensemble des points $M$ est un cercle de centre $\Omega(1; 1)$ et de rayon $2$.
