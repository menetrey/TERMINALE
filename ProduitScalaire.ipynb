# I) Rappels sur les vecteurs
## 1) Définition
<div class="alert alert-block alert-danger">
Un vecteur $\overrightarrow{u}$ ou $\overrightarrow{AB}$ est défini par :<br>
• une direction (la droite (AB)).<br>
• un sens (de A vers B)<br>
• Une longueur ou norme : $||\overrightarrow{u}||$ ou $AB$<br>  
<b>Égalité de deux vecteurs :</b>  $\overrightarrow{AB} = \overrightarrow{CD} \iff ABDC \mbox{  parallélogramme  }$  .
</div> 

## 2) Opérations sur les vecteurs 

### a) Somme de deux vecteurs  

La somme de deux vecteurs est définie par la relation de Chasles :  
$\overrightarrow{AC} = \overrightarrow{AB} + \overrightarrow{BC}$  

Inégalité triangulaire : $|| \overrightarrow{u} +\overrightarrow{v}|| \le ||\overrightarrow{u}||  + ||\overrightarrow{v}||$  


Construction de la somme de deux vecteurs de même origine. On trace un parallélogramme, afin de reporter le deuxième vecteur permettant d’appliquer la relation de Chasles.  



<div class="alert alert-block alert-danger">
<b>Propriété  :</b> Somme de deux vecteurs :<br>
• Commutativité : $\quad \quad \overrightarrow{u} + \overrightarrow{v} = \overrightarrow{v} + \overrightarrow{u}$<br>
• Associativité : $\quad \quad  (\overrightarrow{u} +\overrightarrow{v}) + \overrightarrow{w} = \overrightarrow{u} + (\overrightarrow{v} + \overrightarrow{w}) = \overrightarrow{u} + \overrightarrow{v} + \overrightarrow{w}$<br>
• Élément neutre $ \overrightarrow{0} : \quad \quad \overrightarrow{u} + \overrightarrow{0} = \overrightarrow{u}$<br>
• Opposé $\quad - \overrightarrow{u}$ : $\quad −\overrightarrow{AB} = \overrightarrow{BA}$  
 </div>   
 
### b) Multiplication d’un vecteur par un scalaire
Lorsqu’on multiplie un vecteur par un réel $k$, appelé scalaire, le vecteur ainsi formé $k\overrightarrow{u}$ est tel que :      
• Sa longueur est multiplié par $|k|$  
• Si $k \gt 0$ son sens est inchangé  
• Si $k \lt 0$ son sens est inversé.  
• Si $k = 0$ on a : $0\overrightarrow{u} = 0$      

<div class="alert alert-block alert-danger">
<b>Propriété :</b> Bilinéarité.<br>
La multiplication par un scalaire est distributive par rapport à l’addition de deux vecteurs ou la somme de deux réels.  
    $$• k(\overrightarrow{u} + \overrightarrow{v}) = k \overrightarrow{u} + k \overrightarrow{v}$$  <br>
    $$• (k + k′) \overrightarrow{u} = k \overrightarrow{u} + k′\overrightarrow{u}$$  
</div>    
    
## 3) Colinéarité de deux vecteurs
<div class="alert alert-block alert-danger">
<b>Définition  :</b><br>
On dit que deux vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ sont colinéaires, si et seulement si, il existe un réel $k$ tel que : <br>
    $\overrightarrow{v} = k \overrightarrow{u}$
</div>


**Remarque :**  
Le vecteur nul $\overrightarrow{0}$ est colinéaire à tout vecteur car : $\overrightarrow{0} = 0 \overrightarrow{u}$  

<div class="alert alert-block alert-danger">
<b>Propriété :</b>  
La colinéarité permet de montrer le parallélisme et l’alignement.<br>
$\overrightarrow{AB} \quad \mbox{et} \quad \overrightarrow{CD} \quad \mbox{ colinéaires} \quad \iff (AB) \ // \ (CD)$<br>
$\overrightarrow{AB} \quad \mbox{et} \quad \overrightarrow{AC} \quad \mbox{colinéaires} \quad  \iff A, B, C \quad \mbox{alignés}$    
</div>    
    
**Exemple :**    

Soit ABC un triangle, E, I et F tels que :  
$\overrightarrow{AE} = \dfrac{1}{3} \overrightarrow{BC},\quad  \overrightarrow{CI} =\dfrac{2}{3}\overrightarrow{CB} \quad \mbox{et} \quad \overrightarrow{AF}=\dfrac{1}{3}\overrightarrow{AC}$  . 

Démontrer que I, E et F sont alignés.



Exprimons $\overrightarrow{EI}$ et $\overrightarrow{EF}$ en fonction de $\overrightarrow{AB}$ .  

• d'après l'énoncé $\overrightarrow{CI} =\dfrac{2}{3} \overrightarrow{CB}$ donc $\overrightarrow{BI} = \dfrac{1}{3} \overrightarrow{BC}$  . 

On en déduit que $\overrightarrow{AE} = \overrightarrow{BI}$ donc que AEIB est un parallélogramme. On a alors : $\overrightarrow{EI} = \overrightarrow{AB}$  
  
• De plus :  $ \overrightarrow{EF} = \overrightarrow{EA} + \overrightarrow{AF} =
\dfrac{1}{3}\overrightarrow{CB} + \dfrac{1}{3} \overrightarrow{AC} = \dfrac{1}{3} ( \overrightarrow{AC} + \overrightarrow{CB}) = \dfrac{1}{3} \overrightarrow{AB}$  

On en déduit alors : $\overrightarrow{EF} = \dfrac{1}{3} \overrightarrow{EI}$ .  
Les vecteurs $\overrightarrow{EF}$ et $\overrightarrow{EI}$ sont colinéaires et donc les points E, F et I sont alignés.   

## 4) Géométrie analytique
<div class="alert alert-block alert-danger">
<b>Propriété:</b><br>
Mis à part les calculs de distance qui exige un repère orthonormé, les formules suivantes sont valable dans tout repère.<br>
• Soient $A(x_A ; y_A)$ et $B(x_B ; y_B)$, les coordonnées du vecteur $\overrightarrow{AB}$ vérifient :<br>
    $$\overrightarrow{AB} \begin{pmatrix}x_B − x_A \\ y_B − y_A \end{pmatrix}$$

    
<br>• Soient $A(x_A ; y_A)$ et $B(x_B ; y_B)$, les coordonnées du milieu I de [AB] vérifient :<br><br>
    
$$I \Big(\dfrac{x_B + x_A}{2}; \dfrac{y_B + y_A}{2}\Big)$$  


<br>• On appelle déterminant de deux vecteurs $\overrightarrow{u}(x ; y)$ et $\overrightarrow{v}(x′ ; y′)$, le nombre :  
$$det (\overrightarrow{u},\overrightarrow{v}) = \begin{vmatrix}x \quad x′  \\ y \quad y′ \end{vmatrix}= xy′ − x′y$$  

<br>• Deux vecteurs sont colinéaires si et seulement si, leur déterminant est nul.<br>
$\quad \quad \quad \quad \quad \quad \overrightarrow{u}$ et $\overrightarrow{v}$ colinéaires $\iff det (\overrightarrow{u},\overrightarrow{v}) = 0$  
 
<br>• Dans un repère orthonormé, la norme d’un vecteur $\overrightarrow{u}(x; y)$ et la distance entre les points $A(x_A ; y_A)$ et $B(x_B ; y_B)$ vérifient :<br>
    
$$||\overrightarrow{u}|| = \sqrt{x^2 + y^2} \quad \mbox{et} \quad  AB =\sqrt{(x_B − x_A)^2 + (y_B − y_A)^2}$$      
</div>


**Exemples:**  
Dans un repère orthonormé $(O, \overrightarrow{i}, \overrightarrow{j})$  

1) Soit $A(1; 4)$ et $B(−5; 2)$.  

Calculer les coordonnées de $\overrightarrow{AB}$ , du milieu de [AB] et la longueur AB. 


$$\overrightarrow{AB} \begin{pmatrix} −5 − 1 \\ 2 − 4 \end{pmatrix} = \begin{pmatrix}−6  \\ −2 \end{pmatrix}$$  
et $$I \Big( \dfrac{1 − 5}{2};\dfrac{4 + 2}{2} \Big) \quad \mbox{d'où} \quad I \Big( -2;3 \Big) $$
  
    
    
$$  AB =\sqrt{(−6)^2 + (−2)^2} = \sqrt{40} = 2\sqrt{10}$$  


2) On donne $\overrightarrow{u}(2 ; 3)$ et $\overrightarrow{v}(3 ; 4)$. Les vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ sont-ils colinéaires ?  

$$det(\overrightarrow{u} ; \overrightarrow{v}) = \begin{vmatrix} 2 \ 3 \\ 3 \  4 \end{vmatrix} = 8 − 9 = − 1 \neq 0$$  
Les vecteurs ne sont pas colinéaires.  
  
  
Dans un repère quelconque ABCD est un parallélogramme. M, N, Q sont tels que :
$$\overrightarrow{DM} = \dfrac{4}{5} \overrightarrow{DA} , \quad  \overrightarrow{AN} = \dfrac{3}{4} \overrightarrow{AB} , \quad  \overrightarrow{CQ} =\dfrac{2}{3}\overrightarrow{CD}$$  

La parallèle à $(MQ)$ menée par $N$ coupe $(BC)$ en $P$.  
Déterminer le coefficient $k$ de colinéarité tel que $\overrightarrow{BP} = k\overrightarrow{AD}$.    




Dans le repère $(A;\overrightarrow{AB} ,\overrightarrow{AD})$ :
Les coordonnées de $M$, $N$ et $Q$ sont :  
$$M \Big( 0;\frac{1}{5}\Big), \quad N\Big(\dfrac{3}{4}; 0\Big), \quad  Q\Big(\dfrac{1}{3}; 1\Big)$$  

$P$ est sur $(BC)$, son abscisse est 1.    
De plus comme $k$ est tel que : $\overrightarrow{BP} = k\overrightarrow{AD}$, son ordonnée vaut $k$.  
Les coordonnées de $P$ sont : $P(1; k)$  

Comme $(NP) // (MQ)$, le déterminant de $\overrightarrow{MQ}$ et $\overrightarrow{NP}$ est nul, on a :   
$$\begin{array}\\ det (\overrightarrow{MQ},\overrightarrow{NP} ) &=& 0  \iff \begin{vmatrix} \dfrac{1}{3} − 0 & 1 −\dfrac{3}{4} \\ 1 − \dfrac{1}{5} & k − 0 \end{vmatrix}&=& 0 \\ \iff \begin{vmatrix} \dfrac{1}{3} & \dfrac{1}{4} \\ \dfrac{4}{5} & k \end{vmatrix} &=& 0 \\ \iff \dfrac{k}{3} − \dfrac{1}{5} &=& 0 \\ \iff \dfrac{k}{3} &=& \dfrac{1}{5} \\ \iff k &=&\dfrac{3}{5} \end{array}$$   



# II)   Produit scalaire
## 1) Définition
<div class="alert alert-block alert-danger">
<b> formule connaissant un angle</b><br>
On appelle produit scalaire de deux vecteurs, non nuls, $\overrightarrow{u}$ et $\overrightarrow{v}$ le réel noté $\overrightarrow{u} \bullet \overrightarrow{v}$ tel que :<br>
    $$ \overrightarrow{u} \bullet \overrightarrow{v} = ||\overrightarrow{u}|| \times ||\overrightarrow{v}|| \times cos(\overrightarrow{u} , \overrightarrow{v})$$  
    
 </div>
 
**Exemple :**  
On donne la figure suivante, déterminer le produit scalaire :$\overrightarrow{AB} \bullet \overrightarrow{AC}$  





   $$ \overrightarrow{AB} \bullet \overrightarrow{AC} = ||\overrightarrow{AB}|| \times ||\overrightarrow{AC}|| \times cos(\overrightarrow{AB} , \overrightarrow{AC})$$   
     
   $$ \overrightarrow{AB} \bullet \overrightarrow{AC} = 2 \times 3 \times cos(60°)$$
   
   
   $$ \overrightarrow{AB} \bullet \overrightarrow{AC} = 2 \times 3 \times \dfrac{1}{2}= 3$$   
   
  
<div class="alert alert-block alert-danger">
<b> formule avec figure géométrique</b><br>

Vecteurs, non nuls, $\overrightarrow{u}$ et $\overrightarrow{v}$ de même origine.  
    
Soit $~\overrightarrow{u} = \overrightarrow{AB}$ , $\overrightarrow{v} = \overrightarrow{AC}$ et $H$ le projeté orthogonal de $C$ sur $(AB)$. On a alors
  $$ \overrightarrow{AB} \bullet \overrightarrow{AC} = \overrightarrow{AB} \times \overrightarrow{AH}$$  
    
- $\overrightarrow{AB}$ et $\overrightarrow{AH}$ sens contraire :     
    

    
alors    $$ \overrightarrow{AB} \bullet \overrightarrow{AC} = - AB \times AH$$ <br><br>  
    
-   $\overrightarrow{AB}$ et $\overrightarrow{AH}$ de même sens :   
 

alors     $$ \overrightarrow{AB} \bullet \overrightarrow{AC} =  AB \times AH$$ </div> 

** Remarque: **   

• Si $(\overrightarrow{AB} , \overrightarrow{AC}) \lt \dfrac{π}{2}$, les vecteurs $\overrightarrow{AB}$ et $\overrightarrow{AH}$ ont même sens alors, $\overrightarrow{AB} \bullet \overrightarrow{AC} \gt 0$     
  
  
• Si $(\overrightarrow{AB} , \overrightarrow{AC}) \gt \dfrac{π}{2}$
, les vecteurs   $\overrightarrow{AB}$ et $\overrightarrow{AH}$ sont des sens contraires alors, $\overrightarrow{AB} \bullet \overrightarrow{AC} \lt 0$  
  
  
• Si $(\overrightarrow{AB} , \overrightarrow{AC}) = \dfrac{π}{2}$, le vecteur $\overrightarrow{AH} = 0$ alors, $\overrightarrow{AB} \bullet \overrightarrow{AC} = 0$    
  
  
    
**Exemple:**  

Soit ABCD un rectangle tel que AB = 3. Le point C se projette orthogonalement en B donc:  
$$ \overrightarrow{AB} \bullet \overrightarrow{AC} =  AB \times AB = 3^2 = 9$$   
  
<div class="alert alert-block alert-danger">
 <b>Propriété :</b><br>   
Le produit scalaire est un forme bilinéaire symétrique :<br>
• symétrie : $\overrightarrow{u} \bullet \overrightarrow{v} =\overrightarrow{v} \bullet \overrightarrow{u}$<br>
• bilinéarité : pour tous réels $a$ et $b$ : $
\overrightarrow{u} \bullet (\overrightarrow{v}+ \overrightarrow{w})  = \overrightarrow{u} \bullet \overrightarrow{v}+ \overrightarrow{u} \bullet \overrightarrow{w}$ et $(a\overrightarrow{u}) \bullet (b\overrightarrow{v}) = ab \times (\overrightarrow{u} \bullet \overrightarrow{v})$  
</div>


**Remarque :**    
Ces propriétés permettent d’effectuer des opérations sur le produit scalaire comme le produit et la somme de quantités algébriques. C’est une sorte de distributivité.     


**Exemple :**  



En utilisant les renseignements portés sur la figure ci-dessus, calculer les produits scalaires suivants :
$$\begin{array}\\ (\overrightarrow{AB} + \overrightarrow{AH} )\bullet  \overrightarrow{AB} &=& AB^2 + \overrightarrow{AH} \bullet  \overrightarrow{AB} \\
&=& AB^2 + AH^2 \\
&=& AB^2 + (AB^2 − BH^2) \quad \mbox{th de Pythagore} \\
&=& 2AB^2 − BH^2 \\
&=& 2 \times 4 − 1\\
&=& 7\end{array}$$  

  
  
$$\begin{array}\\ 
(\overrightarrow{AH} + \overrightarrow{HC} )\bullet  \overrightarrow{AB} &=&  \overrightarrow{AH} \bullet  \overrightarrow{AB} + \overrightarrow{HC} \bullet  \overrightarrow{AB} \\
&=& AH^2 + \overrightarrow{HC} \bullet  \overrightarrow{AB} \\
&=& (AB^2 − BH^2) − \overrightarrow{HC} \times  \overrightarrow{HB}  \quad \mbox{ th de Pythagore et colinéarité} \\
&=& 4 − 1 − 2 \times 1 \\
&=& 1 
\end{array}$$


<div class="alert alert-block alert-danger">
<b>Propriété: </b><br>
    
Colinéarité et orthogonalité<br>
• $ \overrightarrow{u} \bullet \overrightarrow{v} = ||\overrightarrow{u}|| × ||\overrightarrow{v}|| \iff \overrightarrow{u}\mbox{ et } \overrightarrow{v}$ colinéaires de même sens. <br> 
    
• $\overrightarrow{u} \bullet \overrightarrow{v} = −||\overrightarrow{u}|| × ||\overrightarrow{v}|| \iff \overrightarrow{u}\mbox{ et } \overrightarrow{v}$ colinéaires de sens contraires.  <br>
    
• $\overrightarrow{u} \bullet \overrightarrow{v} = 0 \iff \overrightarrow{u}\mbox{ et } \overrightarrow{v}$ orthogonaux. <br> 
</div>
Remarque : On écrit : $\overrightarrow{u} ^2 = \overrightarrow{u}  \bullet \overrightarrow{u}  = ||\overrightarrow{u} ||^2$  
  
## 3) Autres définitions
### a) Définition par la norme
<div class="alert alert-block alert-danger">
Le produit scalaire de deux vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ est égal à :<br>
$\overrightarrow{u} \bullet \overrightarrow{v} = \dfrac{1}{2}\big( ||\overrightarrow{u}+ \overrightarrow{v}||^2  − || \overrightarrow{u}||^2 − ||\overrightarrow{v}||^2 \big)$  

</div>

**Remarque :**  
Cette définition mesure le défaut d’orthogonalité de deux vecteurs.  
En effet si le triangle ABC est rectangle en A, alors $BC^2 − AB^2 − AC^2 = 0$  
### b) Définition analytique
<div class="alert alert-block alert-danger">
Dans un repère orthonormé $(O,\overrightarrow{i},\overrightarrow{j})$, le produit scalaire de deux vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ de coordonnées respectives $(x; y)$ et $(x′; y′)$ est égal à :<br>
    $$ \begin{array}{r l} \\
    \overrightarrow{u} \bullet \overrightarrow{v} &=& xx′ + yy′ \quad \mbox{notation matricielle} \\
\iff \begin{pmatrix}x \\ y \end{pmatrix} \bullet \begin{pmatrix} x′ \\ y′ \end{pmatrix} &=& xx′ + yy′ \end{array}$$
</div>

**Exemple :**  
On donne la figure suivante, déterminer le produit scalaire :$\overrightarrow{AB} \bullet \overrightarrow{AC}$  






$\overrightarrow{AB} \bullet \overrightarrow{AC}   = \begin{pmatrix}3 − 2 \\ 0 − 2 \end{pmatrix} \bullet \begin{pmatrix} −1 − 2 \\ 1 − 2 \end{pmatrix} = \begin{pmatrix} 1 \\ −2 \end{pmatrix} \bullet \begin{pmatrix}−3  \\ −1 \end{pmatrix} = 1 \times (−3) + (−2) \times (−1) = −1$  
 
   
## 4)  Relation d’Al-Kashi
Cette relation a pour but de déterminer une relation entre les trois longueurs d’un triangle, il s’agit de la généralisation du théorème de Pythagore.   


<div class="alert alert-block alert-danger">
Soit un triangle ABC. On a :  <br>
    $$BC^2 = AB^2 + AC^2 − 2AB \times AC \times cos \widehat{A}$$
    
Avec a, b et c les longueurs des côtés opposés respectivement
à A, B et C. On a : <br>
    $$a^2 = b^2 + c^2 − 2 \ b \ c \ cos \widehat{A}$$
</div>    

  
      
 **Exemple :**  
Déterminer la longueur BC et les angles $\widehat{B}$ et $\widehat{C}$ du triangle ci-dessous.       

  

D’après la figure : $b = 3, c = 8 \ \mbox{et} \ \widehat{A} = 60°$.

• D’après la relation d’Al-Kashi, nous avons :  
$$a^2 = b^2+c^2−2 \ b \ c\  cos\widehat{A} = 3^2+8^2−2\times 3 \times 8 \times \dfrac{1}{2} = 9+64−24 = 49 \iff a = 7$$  
  
  
• Pour déterminer l’angle $\widehat{B}$, on effectue une permutation circulaire de la formule d’Al-Kashi, c’est à dire :  


On obtient donc : $b^2 = c^2 + a^2 − 2 \ a \ c \ cos\widehat{B} \iff 2\ a\ c \ cos \widehat{B} = a^2 + c^2 − b^2$  

$$cos \widehat{B} =\dfrac{a^2 + c^2 − b^2}{2ac}=\dfrac{49 + 64 − 9}{2 \times 7 \times 8}=\dfrac{104}{112}=\dfrac{13}{14}$$  


On obtient donc : $\widehat{B} = arccos\Big(\dfrac{13}{14}\Big) \approx 21, 7°$  

• Enfin, par complément à 180° : $\widehat{C} \approx 180 − 60 − 21, 79 \approx 98, 21°$   

## 5) Ensemble de points
<div class="alert alert-lbock alert-danger">
Soient deux points A et B et leur milieu I, pour tout point M on
a la relation :<br>
    $$
\overrightarrow{MA} \bullet \overrightarrow{MB} = MI^2 − \dfrac{1}{4}AB^2$$  
</div>


**Démonstration :**  
On introduit le point I dans le produit scalaire :  
$$\begin{array} \\
\overrightarrow{MA} \bullet \overrightarrow{MB}  
&= &   (\overrightarrow{MI} + \overrightarrow{IA}) \bullet (\overrightarrow{MI} + \overrightarrow{IB} )  \\
&=& MI^2 + \overrightarrow{MI} · \overrightarrow{IB} + \overrightarrow{IA} \bullet \overrightarrow{MI} + \overrightarrow{IA} \bullet \overrightarrow{IB} \\
&=& MI^2 +\overrightarrow{MI} (\overrightarrow{IA} + \overrightarrow{IB}) − IA^2 \\
&=& MI^2 − \dfrac{1}{4}AB^2 \end{array}$$


**Exemple :**  
Déterminer l’ensemble (E) des points M tel que : $\overrightarrow{MA} \bullet \overrightarrow{MB} = 7$ et $AB = 6$.  

$$\begin{array}{l l}
\overrightarrow{MA} \bullet \overrightarrow{MB} &=& 7 \\
\iff MI^2 −\dfrac{1}{4}AB^2 &=& 7 \\
\iff MI^2 &=& 7 + \dfrac{1}{4} AB^2 \\
\iff &=& 7 + 9 \\
\iff &=& 16 \\
\iff MI^2 &=& 4^2 \end{array}$$

L’ensemble $(E)$ est le cercle de centre $I$ et de rayon 4.  



<div class="alert alert-lbock alert-danger">
<b>Propriété:</b><br>  
Un cercle de diamètre [AB] est caractérisé par :$\overrightarrow{MA} \bullet \overrightarrow{MB} =0$<br>
</div>

**Démonstration :**  
En effet un point M du cercle de diamètre [AB] de milieu I est tel que :   
$$MI^2 =\dfrac{1}{2}AB^2=\dfrac{1}{4}AB^2$$ donc  $\overrightarrow{MA} \bullet \overrightarrow{MB} =0$.   
  
# III) Géométrie repérée
## 1) Équation cartésienne et équation réduite
<div class="alert alert-lbock alert-danger">
Une droite est définie par un point $A$ et un vecteur directeur $\overrightarrow{u}$ 
</div>

**Remarque :**  
La droite $(AB)$ est définie par le point A et le vecteur directeur $\overrightarrow{AB}$ .  
  
  
<div class="alert alert-block alert-danger">
<b> Théorème:   Équation cartésienne d’une droite</b><br>  
    
Toute droite $d$ du plan est déterminée par une équation de la forme :
$d : ax + by + c = 0$, avec $a$ et $b$ non tous nuls.  <br>
    
Un vecteur directeur de la droite $d$ est alors $\overrightarrow{u}(−b ; a)$
</div>

**Exemple :**  
Soit la droite $d$ définie par les point $A(2 ; 3)$ et $\overrightarrow{u}(−2 ; 1)$.  

Déterminer une équation cartésienne de la droite $d$.  
  
Soit $M(x ; y) \in d$, les vecteurs $\overrightarrow{AM}$ et $\overrightarrow{u}$ sont colinéaires donc :    

$$\begin{array} \\
det(\overrightarrow{AM},\overrightarrow{u}) &=& 0 \\
\iff \begin{vmatrix} x − 2 & −2 \\ y − 3 & 1 \end{vmatrix} &=& 0 \\
\iff   (x − 2) + 2(y − 3) &=& 0 \\
\iff x + 2y − 2 − 6 &=& 0 \\
\iff x + 2y − 8 &=& 0 \end{array}$$  

**Attention:** 
Une équation cartésienne n’est pas unique. On peut multiplier les coefficients
de l’équation par un facteur $k$ non nul.   
Par exemple, la droite d est définie par :  
$$d : −2x − 4y + 16 = 0$$  

<div class="alert alert-block alert-danger">
<b>Théorème: Équation réduite d’une droite </b><br>
Toute droite $d$, non verticale, admet une équation de la forme :
$d : y = mx + p$ où $\overrightarrow{u}(1 ;m)$ est vecteur directeur de $d$
</div>


**Exemple :**  
$d : x + 2y − 8 = 0$ admet comme équation réduite $y = −\dfrac{1}{2}x + 4$.  

## 2) Vecteur normal à une droite
<div class="alert alert-block alert-danger">
Un vecteur normal $\overrightarrow{n}$ à une droite $d$ est un vecteur orthogonal à tout vecteur directeur $\overrightarrow{u}$ de la droite $d$.  <br>On a alors : $\overrightarrow{n} \bullet \overrightarrow{u} = 0$  
</div>





<div class="alert alert-block alert-danger">

<b>Théorème: Dans un repère orthonormé </b><br>
• Si $d : ax + by + c = 0$, alors $\overrightarrow{n}(a ; b)$ vecteur normal à $d$.<br>
• Réciproquement si un vecteur $\overrightarrow{n}(a ; b)$, non nul, est un vecteur normal à une droite $d$, alors $ax + by + c = 0$ est une équation cartésienne de $d$.  
</div>


**Remarque :**  Le repère orthonormé est essentiel pour l’orthogonalité.   

**Démonstration :**  Soit $d : ax + by + c = 0$ de vecteur directeur $\overrightarrow{u}(−b ; a)$. 
$$\overrightarrow{n} \bullet \overrightarrow{u} = \begin{pmatrix} a \\ b \end{pmatrix} \bullet \begin{pmatrix} -b \\ a \end{pmatrix} = −ab + ab = 0 \implies \overrightarrow{n} \perp \overrightarrow{u}$$  
  
**Exemple :**  
On donne la droite $d : 3x − y + 5 = 0$. 

1) Déterminer une équation de la droite $\Delta$ passant par $A(1 ; 2)$ perpendiculaire à $d$. On proposera deux solutions.  

• Si $\Delta$ est perpendiculaire à $d$, un vecteur normal $\overrightarrow{n}(3 ;−1)$ à $d$ est un vecteur directeur de la droite $\Delta$. On a alors : $\Delta : −x − 3y + c = 0$  

$A \in \Delta$, ses coordonnées vérifient l’équation de $\Delta : −1−6+c = 0 \iff c = 7$  

Une équation cartésienne de $\Delta$ est : $−x − 3y + 7 = 0 \times(−1) \iff x + 3y − 7 = 0$   

• Soit $M(x ; y) \in \Delta$ et $\overrightarrow{u}(1 ; 3)$ un vecteur directeur de $d$. On a alors :  

$$\begin{array}
\overrightarrow{AM} \bullet \overrightarrow{u} &=& 0 \\
\iff \begin{pmatrix} x − 1 \\ y − 2 \end{pmatrix} \bullet \begin{pmatrix}1 \\ 
3 \end{pmatrix}  &=& 0\\
\iff  x−1+3y−6 &=& 0 \\
\iff x+3y−7 &=& 0 
\end{array}$$  

  
2) En déduire les coordonnées du point $H$, projeté orthogonal de $A$ sur la droite $d$.  

En déduire la distance de $A$ à la droite $d$.  

Si $H$ est le projeté orthogonal de $A$ sur $d$ alors, $H$ est l’intersection des droites $d$ et $D$.
Les coordonnées des $H$ vérifient le système :  
$$\begin{cases}
3x − y &=  -5 \quad (\times 3)\\
x + 3y &= 7
\end{cases}$$
 
$$\begin{array} \\
9x − 3y &= −15 \\
x + 3y &= 7 \\
\hline
10x + 0y &= −8 \quad \mbox{d'où} \quad x= -\dfrac{4}{5}
\end{array}$$  
  
de la $1^e$ équation  :  

$$y = 3x + 5 = −\dfrac{12}{5}+ 5 =\dfrac{13}{5}$$  

donc $H\Big(-\dfrac{4}{5};\dfrac{13}{5}\Big)$


La distance de $A$ à la droite $d$ est alors :  
$$ AH =\sqrt{\Big( −\dfrac{4}{5} − 1 \Big)^2 + \Big(\dfrac{13}{5} − 2 \Big)^2}
= \sqrt{ \Big( −\dfrac{9}{5} \big)^2 + \Big(\dfrac{3}{5} \Big)^2} = \dfrac{ \sqrt{81 + 9}}{5} = \dfrac{3\sqrt{10}}{5} $$

## 3) Équation d’un cercle
<div class="alert alert-block alert-danger">
<b>Théorème: Dans un repère orthonormé </b><br>
L’équation cartésienne d’un cercle, $\mathcal{C}$ de centre $\Omega(a ; b)$ et de rayon $r$ est de la forme :<br>
    $$(x − a)^2 + (y − b)^2 = r^2$$
</div>

**Exemple :**  
Montrer que l’ensemble des points $M(x; y)$ tels que :$x^2 + y^2 − 2x − 2y − 2 = 0$ est un cercle. Préciser le rayon et les coordonnées du centre.    
  
  
On regroupe les "x" et les "y" : $x^2 − 2x + y^2 − 2y = 2$  
Débuts de carrés parfaits : $(x − 1)^2 − 1 + (y − 1)^2 − 1 = 2$  
On obtient alors : $(x − 1)^2 + (y − 1)^2 = 4 = 2^2$  
L’ensemble des points $M$ est un cercle de centre $\Omega(1; 1)$ et de rayon $2$.
