## 2) Mthode des rectangles : la quadrature de la parabole

<div class="alert alert-block alert-danger">
    <b> Dfinition : Quadrature de la parabole</b><br>
  La quadrature de la parabole est le calcul de laire de la surface dlimite par un segment de parabole et une droite.<br>
</div>

**Exemple  :**  
Encadrer laire $A$ sous la parabole de la fonction $f:x \to x^2$ sur $[0 ; 1]$.  Lide de Riemann est dencadrer cette aire par deux sries de rectangles , on divise $[0; 1]$ en $n$ parties.   

     


Sur $[\dfrac{i}{n};\dfrac{i+1}{n}]$, comme $f$ est croissante :  
 la valeur minimale est $f(\dfrac{i}{n})$ et  
 la valeur maximale $f(\dfrac{i+1}{n})$.  
 On obtient 2 rectangles de largeur $\dfrac{1}{n} $ et de hauteurs $f(\dfrac{i}{n})$ et $f(\dfrac{i+1}{n})$.  
Soit $S_n$ et $T_n$ la somme des aires respectives des rectangles infrieurs et des rectangles suprieurs


  
**Algorithme    :**   
On automatise le calcul avec le programme en Python suivant :    


```python
from math import *

def f(x):
    return sqrt(1-x**2)

def A(n):
    s = 0
    t = 0
    for i in range(n):
        s = s + 1/n*f(i/n)
        t = t + 1/n*f((i+1)/n)
    return s, t    
```

### b) Calcul du volume dune sphre de rayon R
Compte tenu de la symtrie de la sphre, on calcule le volume dune demi-sphre quon multipliera ensuite par 2. On peut faire le schma suivant :  

On dcoupe ainsi la sphre avec des plans perpendiculaires  laxe $(Oz)$.   
Les surfaces obtenues sont alors des cercles de rayon $r(z)$.  La surface des ces cercles $S(z)$ vaut : $S(z) =\pi r^2(z)$.  
Il reste donc  dterminer le rayon $r(z)$ en fonction de $z$  laide du thorme de Pythagore dans le triangle OMN rectangle en $N(0, 0,z)$: $$r^2(z) =R^2z^2$$.  
On obtient alors le demi volume de la sphre:
$$\begin{array}\\ 
\dfrac{1}{2}V&=&\int_0^R S(z)dz \\
&=& \int_0^R \pi(R^2z^2)dz \\
&=&\pi[R^2z\dfrac{z^3}{3}]_0^R \\
&=&\pi (R^3\dfrac{R^3}{3})\\
&=&\dfrac{2}{3}\pi R^3\\
\end{array}$$<br>
On en dduit alors le volume de la sphre que tout le monde a appris :<br>
$$V=\dfrac{4}{3}\pi R^3$$<br>


### c) Caclul du volume d'un cne
On dcoupe le cne daxe $(Oz)$ avec des plans perpendiculaires  laxe $(Oz)$.  
Les surfaces obtenues sont alors des disques de rayon $r(z)$.  
La surface des ces disques $S(z)$ vaut : $S(z) = \pi r^2(z)$  


Il reste donc  dterminer le rayon $r(z)$ en fonction de $z$  laide du thorme de Thals dans les triangles :  
$AOC$ et $AAB$, on a avec $A(0, 0,z)$:  
$\frac{AA}{AO}=\frac{AB}{OC} \iff \frac{(hz)}{h}=\frac{r(z)}{R}.$    

On en dduit alors que : $$r(z) =\frac{R}{h}(hz)$$  
On obtient ainsi le volume du cne:  
$$\begin{array}\\
V&=& \int_0^h S(z)dz\\
&=& \int_0 ^h \pi \dfrac{R^2}{ h^2} (h-z)^2dz\\
&=&  \pi \dfrac{R^2}{h^2} \int_0^h(h-z)^2dz \\
&=&  \pi \dfrac{R^2}{h^2}\Big[ -\dfrac{(h-z)^3}{3}\Big]_0^h \\
&=& \pi \dfrac{R^2}{h^2} \Big(-\dfrac{(h-h)^3}{3}+\dfrac{(h-0)^3}{3} \Big)\\
&=& \pi \dfrac{R^2}{h^2} \Big(\dfrac{h^3}{3}\big) \\
&=& \pi \dfrac{R^2 h}{3}\\
&=& \dfrac{1}{3} \pi R^2h\\
\end{array}$$

On retrouve ainsi le volume du cne que tout le monde connat : $V=\dfrac{1}{3}\pi R^2h$
