




# I) Intégrale d'une fonction continue positive
## 1) Définition

<div class="alert alert-block alert-danger">
    Soit $f$ une fonction continue et positive sur un intervalle $[a;b]$. Soit $\mathcal{C_f}$ sa courbe représentative dans un repère orthogonal $(O, I, J)$. <br><br>
    On appelle: <br>
    • Unité d’aire (u.a.): l’aire du rectangle construit à partir des points O, I et J.<br><br>
    • Domaine sous la courbe: domaine délimité par la courbe $\mathcal{C_f}$, l’axe des abscisses, et les droites d’équation $x=a$ et $x=b$. <br>
    Ce qui se traduit par : $a\leq x \leq b$ et  $0 \leq y \leq f(x)$<br><br>
    • Intégrale de $f$ sur $[a;b]$: la mesure de l’aire en u.a. du domaine situé sous la courbe $\mathcal{C_f}$, noté : $\int_a^b f(x)dx$<br>
    

  </div>
  


**Remarque  :**  
• $\int_a^b f(x)dx$ se lit : « somme ou intégrale de $a$ à $b$ de $f(x)dx$».   

• La variable "x" est une variable muette, c’est à dire qu’elle n’est plus présente lorsque le calcul est effectué. Elle peut être remplacée par :t,u, ou toute autre lettre à l’exception de $a$ et $b$.  
On peut ainsi écrire indistinctement :$\int_a^b f(t)dt$, $\int_a^b f(u)du$    

  
**Exemple  :**   
Soit la courbe d’une fonction $f$ sur $[−2 ; 3]$, OI = 2 cm et OJ = 3 cm.   
L’unité d’aire vaut : $2×3=6 cm^2$ 



Pour calculer $\int_{-2}^3 f(x)dx$, on s’aide du quadrillage, on a:  
• 7 rectangles pleins  
• un demi rectangle en haut à gauche 
• un triangle en haut à droite de côté respectifs 2 et 1 soit $\dfrac{2×1}{2}=1$ rectangle.  
  
On en déduit donc : 

Dans le cas général d’une fonction dont la représentation n’est pas une ligne brisée mais une courbe, on ne peut plus utiliser cette méthode de quadrillage. On encadre alors l’aire entre deux séries de rectangles. C’est l’objet de paragraphe suivant : méthode des rectangles.  
  
## 2) Méthode des rectangles : la quadrature de la parabole

<div class="alert alert-block alert-danger">
    <b> Définition : Quadrature de la parabole</b><br>
  La quadrature de la parabole est le calcul de l’aire de la surface délimitée par un segment de parabole et une droite.<br>
</div>

**Exemple  :**  
Encadrer l’aire $A$ sous la parabole de la fonction $f:x \to x^2$ sur $[0 ; 1]$.  L’idée de Riemann est d’encadrer cette aire par deux séries de rectangles , on divise $[0; 1]$ en $n$ parties.   

     


Sur $[\dfrac{i}{n};\dfrac{i+1}{n}]$, comme $f$ est croissante :  
• la valeur minimale est $f(\dfrac{i}{n})$ et  
• la valeur maximale $f(\dfrac{i+1}{n})$.  
• On obtient 2 rectangles de largeur $\dfrac{1}{n} $ et de hauteurs $f(\dfrac{i}{n})$ et $f(\dfrac{i+1}{n})$.  
Soit $S_n$ et $T_n$ la somme des aires respectives des rectangles inférieurs et des rectangles supérieurs


  
**Algorithme    :**   
On automatise le calcul avec le programme en Python suivant :    


```python
from math import *

def f(x):
    return sqrt(1-x**2)

def A(n):
    s = 0
    t = 0
    for i in range(n):
        s = s + 1/n*f(i/n)
        t = t + 1/n*f((i+1)/n)
    return s, t    
```

```python
A(10000)
```

## 3) Intégrale d’une fonction continue positive

<div class="alert alert-block alert-danger">
    <b> Intégrale d’une fonction $f$ continue et positive sur $I= [a;b]$. </b><br>
 
 - On divise I en $n$ parties égales.<br><br>
 - Sur $[\dfrac{i}{n};\dfrac{i+1}{n}]$, on détermine les valeurs minimale et maximale de $f$.<br><br>
 - L’aire sous la courbe est alors encadrée par deux suites $(S_n)$ et $(T_n)$ correspondantes à l’aire des rectangles inférieurs et l’aire des rectangles supérieurs. $(S_n)$ et $(T_n)$ convergent vers la même limite correspondant à l’intégrale de $f$ sur I.<br>
 </div>  
 
 
$$\lim \int_a^b f(x)dx = \lim_{n \to \infty}S_n=\lim_{n \to \infty}T_n=\mathcal{A}$$

**Remarque :**  Le symbole $dx$ dans l’intégrale est une notation différentielle qui symbolise une très petite distance et représente la largeur de chaque petit rectangle.  
$f(x)dx$ représente l’aire d’un rectangle et le symbole $\int$ devant signifie que l’on fait la somme des aires de chaque petit rectangle.  

  
**Exemple  :**  
Calculer l’intégrale : $\int_{-1}^1 \sqrt{1−x^2}dx$  



## 4) Définition cinématique de l’intégrale
On a donné une définition géométrique de l’intégrale, on peut aussi en donner une définition cinématique. Pour un mobile se déplaçant sur un axe à la vitesse $v(t)$, la distance $d$
 parcourue par le mobile entre les instants $t_1$ et $t_2$ vaut : 
 $$d=\int_{t_1}^{t_2} v(t)dt$$
 


# II)  Intégrale et primitive
## 1) Théorème fondamental de l’intégration
<div class="alert alert-block alert-danger">  
Soit une fonction $f$ continue et positive sur un intervalle [a; b].  <br>
La fonction $F_a$ telle que $F_a(x) =\lim \int_a^x f(t)dt$ est une primitive de $f$ s'annulant en $a$ . 
</div>
 
<div class="alert alert-block alert-info">
    <b> Démonstration:</b><br>
    Dans le cas où $f$ est croissante sur $[a;b]$. <br><br> 
    On revient à la définition de la dérivée, il faut montrer que si $x_0 \in [a;b]$ : $F′=f \iff \lim_\limits{h \to 0 }\dfrac{F(x_0+h)−F(x_0)}{h}=f(x_0)$  <br><br>
    -  <b>Pour $h>0$</b>, par soustraction d’aire on a : <br><br>
    
   $$F(x_0+h)−F(x_0) = \int_a^{x_0+h} f(t)dt−\int_a^{x_0} f(t)dt= \int_{x_0}^{x_0+h} f(t)dt=\mathcal{A}$$<br>
    
   $f$ est croissante sur $[a;b]$, donc si $t\in [x_0;x_0+h]$: $f(x_0) \leq f(t) \leq f(x_0+h)$ donc en encadrant l’aire $\mathcal{A}$ par le rectangle inférieur (hachuré) et le rectangle supérieur (bleu),on a :  <br>
    
$$f(x_0)\times h \leq \mathcal{A} \leq f(x_0+h)\times h$$
      <br>
        
$$f(x_0) \leq \dfrac{\mathcal{A}}{h} \leq f(x_0+h)$$
   <br>
        
$$f(x_0) \leq \dfrac{F(x_0+h)−F(x_0)}{h} \leq f(x_0+h)$$
    <br> 
    
- <b>Pour $h<0$</b>, il suffit de faire de même, on trouverait alors:  
$$f(x_0+h) \leq \dfrac{F(x_0+h)−F(x_0)}{h} \leq f(x_0)$$    
    
    
<b>Conclusion</b><br>
On sait que $f$ est continue sur $[a;b]$, donc  $\lim\limits_{h\to0}f(x_0+h) =f(x_0)$ <br>
    D’après le théorème des gendarmes, on a :  $\lim\limits_{h\to 0} \dfrac{F(x_0+h)−F(x_0)}{h}=f(x_0)$
</div>

## 2) Définition
<div class="alert alert-block alert-danger">
    $f$ est une fonction définie sur un intervalle I. On dit que $f$ admet une primitive sur I si, et seulement si, il existe une fonction $F$ dérivable sur I telle que :<br>
    $$ \forall x \in I, F′(x) =f(x)$$
</div>    
  
  
**Exemples:**  
1) Soit la fonction $f$ définie sur $\mathbb{R}$ par $f(x) =2x$.    
Déterminer une primitive de $f$.    
$F$ dérivable sur $\mathbb{R}$ et définie par $F(x) =x^2$ est une primitive de $f$ car :  
$$\forall x \in \mathbb{R}, \quad F′(x) =2x$$  

  
    
      
<div class="alert alert-block alert-danger">
    <br>
Soit une fonction $f$ admettant une primitive $F$ sur $\mathcal{I}$, alors toute primitive $G$ de $f$ est de la forme :<br> $$\forall x \in \mathcal{I}, \quad    G(x) =F(x) +k \mbox{    avec   }   k\in \mathbb{R}$$  
    <br>
 </div>   
    
**Primitive vérifiant la condition initiale**  
<div class="alert alert-block alert-danger">
    Soit $f$ une fonction admettant une primitive sur un intervalle $\mathcal{I}$.Soit $x_0 \in \mathcal{I}$ et $y_0 \in \mathbb{R}$. Il existe une unique primitive $F$ de $f$ sur $\mathcal{I}$ tel que :  <br>
    $$F(x_0) =y_0$$
 </div>

### 4) Intégrale d'une fonction continue
#### a. Calcul à partir d'une primitive

<div class="alert alert-block alert-danger">
 $f$ est une fonction continue sur un intervalle $\mathcal{I}$.<br>
    $F$ est une primitive quelconque de $f$ sur $\mathcal{I}$, alors pour tous réels $a$ et $b$ de $\mathcal{I}$ on a :<br>
    $$\int_a^b f(x)dx=[F(x)]_a^b=F(b)−F(a)$$
  </div>
  
**Exemple:**  
1) Calculer l’intégrale suivante :$\int_{−1}^2(x^2−4x+3)dx$   

$\begin{array}\\ 
\int_{−1}^2(x^2−4x+3)dx &=& [\frac{1}{3}x^3-2x^2+3x]_{-1}^2\\
&=& (\frac{1}{3}(2)^3-2(2)^2+3(2))- (\frac{1}{3}(-1)^3-2(-1)^2+3(-1))\\
&=& (\frac{8}{3}-8+6)- (\frac{-1}{3}-2-3)\\
&=&-(\frac{-1}{3}-5-\frac{8}{3}+2)\\
&=&3+3\\
&=&+6 u.a. 
\end{array}$

#### b. Relation entre aire et intégrale
<div class="alert alert-block alert-info">
  - Soit $f$ une fonction continue sur un intervalle $[a;b]$ telle que $f \leq 0$. <br>
    Soit $\mathcal{A}$ l’aire délimitée par la courbe, l’axe des abscisses et les droites $x=a$ et $x=b$. On a alors :<br>
    $$\mathcal{A}=− \int_a^b f(x)dx $$<br>
    

    
<br><br>
    
 - Soient deux fonctions $f$ et $g$ continues sur $[a;b]$ telles que $f>g$. <br>
   Soit $\mathcal{D}$ l’aire comprise entre les deux courbes et les droites $x=a$ et $x=b$.On a alors :<br>
   $$\mathcal{D}=\int_a^b(f−g)(x)dx  $$
 
    
</div>

**Exemple:**
Trouver l’aire comprise entre la parabole d’équation $y=x^2$ et la droite $y=x$ entre les abscisses $x=0$ et $x=1$.  

On a alors comme la parabole est endessous de la droite :  
$$\begin{array}\\
\mathcal{A}&=& \int_0^1(x−x^2)dx\\
&=&\Big[\dfrac{1}{2}x^2−\dfrac{1}{3}x^3 \Big]_0^1\\
&=& \dfrac{1}{2}−\dfrac{1}{3}=\dfrac{1}{6}\\
\end{array}$$  

#### c.  Propriétés algébriques de l’intégrale
<div class="alert alert-block alert-success">
    Soit $f$ une fonction continue sur un intervalle $\mathcal{I}$ alors :  <br>
    1)$\forall a \in \mathcal{I}$  on a :    $\int_a^af(x)dx=0$<br>
    2)$\forall a,b \in \mathcal{I}$  on a : $\int_a^bf(x)dx=−\int_b^af(x)dx$<br>
    3) <b>Relation de Chasles</b> $\forall,b,c∈ \in \mathcal{I}$  on a : $\int_a^cf(x)dx=\int_a^bf(x)dx+\int_b^cf(x)dx$
 </div>
 
**Remarque  :**  
- Si une fonction est paire, alors d’après la relation de Chasles,on a:  
$$\int_{-a}^af(x)dx=\int_{-a}^0f(x)dx+\int_0^af(x)dx=\int_0^af(x)dx+\int_0^af(x)dx=2\int_0^af(x)dx$$  
- Si une fonction est impaire, alors d’après la relation de Chasles, on a :  
$$\int_{-a}^af(x)dx=\int_{-a}^0f(x)dx+\int_0^af(x)dx=−\int_0^af(x)dx+\int_0^af(x)dx=0$$  

#### d. Linéarité de l’intégrale
<div class="alert alert-block alert-danger">
    Soit $f$ et $g$ deux fonctions continues sur un intervalle $\mathcal{I}$ contenant $a$ et $b$, alors pour tous les réels $\alpha$ et $\beta$ , on a :<br>
    $$\int_a^b(\alpha f+\beta g)(x)dx=\alpha\int_a^bf(x)dx+ \beta \int_a^bg(x)dx$$
</div>



## 5) Intégration par parties
<div class="alert alert-block alert-danger">
    <b>Le principe</b><br>
    Soient $u$ et $v$ deux fonctions dérivables sur $[a;b]$ de dérivées $u′$ et $v′$ continues, alors :<br>
    $$\int_a^b u\  v′(x)dx=\Bigg[ u\ v(x) \Bigg]_a^b−\int_a^bu′v(x)dx$$  
 </div> 
 
 <div class="alert alert-block alert-info">
    <b>Démonstration  :</b><br>  
    si $u$ et $v$ sont dérivables sur $[a;b]$ alors, $u\ v$ est dérivable et :<br>
    $$(uv)′=u′v+uv′\iff uv′=(uv)′−u′v$$ <br>
  Comme $u′$ et $v′$ sont continues, on peut intégrer l’égalité, d’où :<br>
    $$\int_a^b u\ v′(x)dx=\int_a^b [(uv)′(x)−u′v(x)]dx \overset{linéarité}{=} \int_a^b (uv)′(x)dx−\int_a^bu′v(x)dx$$
    $$=[uv(x)]_a^b−\int_a^bu′v(x)dx$$
  </div>
  
 <div class="alert alert-block alert-warning"> 
    <b>Exemples d'intégration par parties: </b><br> 
 Le choix de $u$ et $v′$ est stratégique. Il faut pouvoir trouver une primitive de $u′v$<br> <br>
 <b>1) Calculer :$\int_0^1 x\ e^xdx$  </b><br>
    
  On ne peut pas trouver une primitive de $x\to x e^x$ car elle n'est pas de la forme $u′e^u$ .  Le "x" devant l’exponentielle est en trop. <br>
  Il faut donc intégrer par parties en posant : 
   $$\begin{array}\\
   u(x) =x& \quad u′(x) =1 \\
   v′(x) =e^x & \quad v(x) =e^x\\
   \end{array}$$<br>
   On obtient donc : $\int_0^1xe^xdx=\Big[xe^x \Big]_0^1−\int_0^1e^xdx=\Big[xe^x \Big]_0^1− \Big[e^x \Big]_0^1= (e−0)−(e−1)=1$
  <br><br>
  
 <b> 2) Déterminer une primitive de $\ln x$ sur $]0 ;+\infty[$ </b><br> 
  Soit $F$ la primitive de $\ln$ qui s’annule en 1. On a alors : $F(x) =\int_1^x lntdt$. La primitive de $ln$ n’est pas connue, on décompose $lnt$ en : $lnt=1×lnt$.<br><br>
    
  on pose alors :<br>
    $$\begin{array}\\
    u(t) =lnt & u′(t) =1/t\\
    v′(t) =1 & v(t) =t \\
    \end{array}$$<br>
    On a alors :$F(x) =\int_1^xlntdt=\Big[tlnt\Big]_1^x−\int_1^x 1 dt=\Big[tlnt\Big]_1^x−\Big[t\Big]_1^x=xlnx−x+1$ <br><br>
    Si l’on cherche la primitive $G$ de $lnx$ avec une constante d’intégration nulle,on a :<br> $G(x) =xlnx−x  $
 
 <b> 3) On peut être amené à faire une double intégration par partie.<br>
     Déterminer la primitive $F$ sur $\mathbb{R}$ qui s’annule en $−1$ de $f(x)=(x+1)^2e^{2x}$</b><br>
    On a alors : $F(x)=\int_{-1}^x(x+1)^2e^{2x}dx$.<br>
    
 On cherche à faire disparaître le polynôme devant l’exponentielle. On pose alors :<br>
 $$ \begin{array}
  u(t)=(1+t)^2 & \quad u'(t)=2(1+t) \\
  v'(t)=e^{2t} & \quad v(t)=\frac{1}{2}e^{2t}\\
  \end{array}$$   <br>
    
   $$ \begin{array}\\
   \int_{-1}^x (1+x)^2e^{2x}dx &=& \Big[(1+t)^2 \frac{1}{2} e^{2t} \Big]_{-1}^x - \int_{-1}^x 2(1+t)\frac{1}{2}e^{2t}dt\\
   &=& \Big[(1+t)^2 \frac{1}{2} e^{2t} \Big]_{-1}^x - \int_{-1}^x (1+t)e^{2t}dt\\
   \end{array}$$<br>
    
 On fait de nouveau une intégration par partie sur $\int_{-1}^x (1+t)e^{2t}dt$, on pose alors : <br><br>
    $$\begin{array}\\
    u(t)=(1+t) & \quad u'(t)=1\\
    v'(t)=e^{2t} & \quad v(t)=\frac{1}{2}e^{2t}\\
    \end{array}$$<br>
    
 On retrouve alors :   $$\int_{-1}^x (1+t)e^{2t}dt = \Big[\frac{1}{2}(1+t)e^{2t}\Big]_{-1}^x- \int_{-1}^x \frac{1}{2}e^{2t}dt$$<br><br>
    
    
    
    
  $$ \begin{array}\\
   \int_{-1}^x (1+x)^2e^{2x}dx&=& \Big[(1+t)^2 \frac{1}{2} e^{2t} \Big]_{-1}^x - \int_{-1}^x (1+t)e^{2t}dt\\
   &=& \Big[(1+t)^2 \frac{1}{2} e^{2t} \Big]_{-1}^x - \Big[\frac{1}{2}(1+t)e^{2t}\Big]_{-1}^x + \int_{-1}^x \frac{1}{2}e^{2t}dt\\
    &=& \Big[(1+t)^2 \frac{1}{2} e^{2t} \Big]_{-1}^x - \Big[\frac{1}{2}(1+t)e^{2t}\Big]_{-1}^x + \Big[ \frac{1}{4}e^{2t}\Big]_{-1}^x \\
    &=& \Big(\frac{1}{2}(1+x)^2e^{2x}-0 \Big)- \Big(\frac{1}{2}(1+x)e^{2x}-0 \Big) + \Big(\frac{1}{4}e^{2x}-\frac{1}{4}e^{-2}\Big)\\
    &=& \frac{1}{2}e^{2x}(x^2+x+\frac{1}{2})-\frac{1}{4}e^{-2}\\
   \end{array}$$  <br>
    
 Ainsi on obtient la primitive de f qui s'annule en $-1$: $F(x)=   \frac{1}{2}e^{2x}(x^2+x+\frac{1}{2})-\frac{1}{4}e^{-2}$.
    
</div>

## V) Calcul de volume
### a) Présentation d’une méthode de calcul
Une méthode pour déterminer le volume d’un solide, consiste à découper celui-ci par des plans parallèles. On intègre ensuite les surfaces obtenues par ce découpage suivant l’axe normal à ces plans. 

### b) Calcul du volume d’une sphère de rayon R
Compte tenu de la symétrie de la sphère, on calcule le volume d’une demi-sphère qu’on multipliera ensuite par 2. On peut faire le schéma suivant :  

On découpe ainsi la sphère avec des plans perpendiculaires à l’axe $(Oz)$.   
Les surfaces obtenues sont alors des cercles de rayon $r(z)$.  La surface des ces cercles $S(z)$ vaut : $S(z) =\pi r^2(z)$.  
Il reste donc à déterminer le rayon $r(z)$ en fonction de $z$ à l’aide du théorème de Pythagore dans le triangle OMN rectangle en $N(0, 0,z)$: $$r^2(z) =R^2−z^2$$.  
On obtient alors le demi volume de la sphère:
$$\begin{array}\\ 
\dfrac{1}{2}V&=&\int_0^R S(z)dz \\
&=& \int_0^R \pi(R^2−z^2)dz \\
&=&\pi[R^2z−\dfrac{z^3}{3}]_0^R \\
&=&\pi (R^3−\dfrac{R^3}{3})\\
&=&\dfrac{2}{3}\pi R^3\\
\end{array}$$<br>
On en déduit alors le volume de la sphère que tout le monde a appris :<br>
$$V=\dfrac{4}{3}\pi R^3$$<br>


### c) Caclul du volume d'un cône
On découpe le cône d’axe $(Oz)$ avec des plans perpendiculaires à l’axe $(Oz)$.  
Les surfaces obtenues sont alors des disques de rayon $r(z)$.  
La surface des ces disques $S(z)$ vaut : $S(z) = \pi r^2(z)$  


Il reste donc à déterminer le rayon $r(z)$ en fonction de $z$ à l’aide du théorème de Thalès dans les triangles :  
$AOC$ et $AA′B$, on a avec $A′(0, 0,z)$:  
$\frac{AA′}{AO}=\frac{AB}{OC} \iff \frac{(h−z)}{h}=\frac{r(z)}{R}.$    

On en déduit alors que : $$r(z) =\frac{R}{h}(h−z)$$  
On obtient ainsi le volume du cône:  
$$\begin{array}\\
V&=& \int_0^h S(z)dz\\
&=& \int_0 ^h \pi \dfrac{R^2}{ h^2} (h-z)^2dz\\
&=&  \pi \dfrac{R^2}{h^2} \int_0^h(h-z)^2dz \\
&=&  \pi \dfrac{R^2}{h^2}\Big[ -\dfrac{(h-z)^3}{3}\Big]_0^h \\
&=& \pi \dfrac{R^2}{h^2} \Big(-\dfrac{(h-h)^3}{3}+\dfrac{(h-0)^3}{3} \Big)\\
&=& \pi \dfrac{R^2}{h^2} \Big(\dfrac{h^3}{3}\big) \\
&=& \pi \dfrac{R^2 h}{3}\\
&=& \dfrac{1}{3} \pi R^2h\\
\end{array}$$

On retrouve ainsi le volume du cône que tout le monde connaît : $V=\dfrac{1}{3}\pi R^2h$


### d) Volume d'un tore
Un tore est un solide qui a la forme d’une "chambre à air" ou d’un "donut" pour les anglais. Il est caractérisé par $d$ et $R$ comme indiqué sur la figure suivante :  

Si l’axe du tore est l’axe $z$ alors lorsqu’on découpe le tore par des plans perpendiculaires à cet axe, on obtient des couronnes dont les rayons des cercles intérieur et extérieur sont respectivement $d+r(z)$ et $d−r(z)$, la surface d’une couronne d’altitude $z$ vaut :  
$$\begin{array}\\
S(z) &=&\pi \Big[(d+r(z))^2−(d−r(z))^2 \Big]\\
&=& \pi(d^2+2d r(z) +r^2(z)−d^2+2d r(z)−r^2(z))\\
&=& 4\pi d r(z)\\
\end{array}$$

Le volume d’un demi tore vaut donc :  
$$\dfrac{1}{2}V=\int_0^R S(z)dz = 4\pi d \int_0^R r(z)dz$$  
  
On a la figure ci-dessous :



On a :  
$$\int_0^R r(z)dz = \mbox{aire d’un quart de cercle} = \dfrac{\pi}{4}R^2$$  
On obtient alors le volume du tore :  
$$\dfrac{1}{2}V=4\pi d \times \dfrac{\pi}{4}R^2 \iff V=2 \pi^2R^2d$$


## VI) Intégrales et inégalités
<div class="alert alert-block alert-danger">
 Soient $f$ et $g$ deux fonctions continues sur un intervalle $[a;b]$.  <br>
    <b>1) Positivité</b><br> 
    Si $f>0$ sur $[a;b]$ alors : $\quad \int_a^b f(x)dx >0$  <br><br>
    <b>2) Intégration d’une inégalité </b><br>  
  Si $f>g$ sur $[a;b]$ alors : $\quad \int_a^b f(x)dx \gt \int_a^b g(x)dx$<br> <br> 
    <b>3) Inégalité de la moyenne </b> <br> 
    Si $$\forall x \in [a;b], \quad m \leq f(x)\leq M \quad \mbox{alors}\quad m \times(b−a)\leq \int_a^b f(x)dx \leq  M \times (b−a)$$
    </div>


**Exemple:** Encadrement de l’intégrale suivante : $I=\int_1^9 \dfrac{1}{1+\sqrt{x}}dx$.  

## V) Valeur moyenne
<div class="alert alert-block alert-danger">
Soit une fonction $f$ continue sur un intervalle $[a;b]$. <br>  
 La valeur moyenne de la fonction $f$ sur $[a;b]$ est le réel $\mu$ défini par : <br>
    $$\mu=\dfrac{1}{b−a}\int_a^b f(x)dx$$  
 </div>
