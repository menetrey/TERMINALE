# I) Rappels sur la fonction exponentielle
## 1)  Définition et propriétés
<div class="alert alert-block alert-danger">
La fonction exponentielle, notée exp, est l’unique fonction dérivable
sur $\mathbb{R}$ égale à sa dérivée et vérifiant : $exp(0) = 1$.<br>
 </div>  
 
   
**Remarque :** En admettant l’existence d’une telle fonction, on montre l’unicité en montrant que la fonction exp ne s’annule pas sur $\mathbb{R}$.  

   
<div class="alert alert-block alert-danger">

<b>Théorème 1 : Propriétés</b><br>
• Relation fonctionnelle : $\forall a, b \in  \mathbb{R}, exp(a + b) = exp(a) \times exp(b)$.<br><br>
    
• Positivité : $\forall x \in \mathbb{R}, exp(x) > 0$<br><br>
• Monotonie : La fonction exp est croissante sur $\mathbb{R}$<br><br>
• Notation d’Euler : On pose $exp(x) = e^x$ avec $e = exp(1) ≈ 2, 718$  
    
<br><br>
$\forall a, b \in \mathbb{R} : \begin{matrix} e^{a+b} = e^a \times e^b &   e^{−a} =
\dfrac{1}{e^a} \\ e^{a−b} = \dfrac{e^a}{e^b} &  e^{na} = (e^a)^n , n \in N \end{matrix}$  <br>
</div>


**Remarque :**  La relation fonctionnelle pourrait servir de définition à la fonction exponentielle : « unique fonction qui prend la valeur 1 en 0 et qui transforme une somme en produit ».  


**Algorithme :**  On obtient une approximation de e par l’approximation affine de l’exponentielle en a : ea+p ≈ ea(1 + p) où p est le pas.   

e(0.000 1) renvoie 2,718  




```python
def e(p) :
    e =1
    for i in range( int (1/p) ) :
        e=e *(1+p)
    return e
```

# II)  Fonction logarithme népérien
**Avant propos**
La création de la fonction logarithme népérien est, à l’origine, antérieure à la fonction exponentielle bien que dans notre progression elle suive l’étude de la fonction exponentielle.  
La fonction logarithme a été créée par un drapier écossais du XVIIe siècle. Ce drapier, Néper, cherche une fonction pour simplifier les
longs calculs des astronomes, des navigateurs et des financiers. Il crée alors une fonction f qui transforme le produit en somme : $f (ab) = f (a) + f (b)$.  

## 1) définition
<div class="alert alert-block alert-danger">
On appelle fonction logarithme népérien notée $ln$, la fonction
définie de $]0 ;+\infty[$ sur $\mathbb{R}$ telle que : <br>
    $$ x = e^y \iff y = ln x$$
On dit que la fonction $ln$ est la fonction réciproque de la fonction exponentielle.<br><br>
On a alors : $ln 1 = 0$ et $ln e = 1$ ainsi que<br>
$$\forall x \in \mathbb{R}, \quad ln (e^x) = x$$
    $$\forall x \in ]0 ;+\infty[,\quad  e^{ ln x} = x$$
</div>

**Exemple :** $ln (e^{−2}) = −2$ et $e ^{ln 5} = 5$  
  
## 2) Représentation
Les représentations de la fonction logarithme népérien et de la
fonction exponentielle sont symétriques par rapport à la droite d’équation y = x.  





## 3) Variation de la fonction logarithme
<div class="alert alert-block alert-danger">
La fonction $ln$ est strictement croissante sur $]0 ;+\infty[$
 </div>

<div class="alert alert-block alert-info">
    
Propriété : <br>
  Soit $a$ et $b$ deux réels strictement positifs<br>
• De la monotonie de la fonction $ln$, on a : $ln a = ln b \iff a = b$<br>
• De la croissance de la fonction $ln$, on a : $ln a \gt ln b \iff a \gt b$
</div>    
    
    
**Exemples :**  
Résoudre $ln(2 − 2x) = 1$.    

  
Résoudre $ln(2x + 1) \leq −1$.   

# 3 Propriétés algébriques de la fonction ln
## 1)  Relation fonctionnelle
<div class="alert alert-block alert-danger">
Pour tous $a, b \in ]0 ;+\infty[ : ln \ ab = ln a + ln b$
</div>    
Remarque : Cette propriété est à l’origine de la fonction logarithme.  

**Exemple :**  
$ln \  6 = ln(2 × 3) = ln 2 + ln 3$  

## 2) Quotient, inverse, racine carrée et puissance
<div class="alert alert-block alert-danger">  
 Pour tous $a, b \in ]0 ;+\infty[$, on a :  <br>
$\begin{cases} ln\dfrac{a}{b} &=& ln a − ln b \\ 
ln \dfrac{1}{b} &=& −ln b \\  
ln\sqrt{a} &=& \dfrac{1}{2}ln a \\
ln a^n &=& n ln a \quad \mbox{ avec } n \in \mathbb{N}
\end{cases}$
</div>
