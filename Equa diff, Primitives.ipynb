# I) Équation différentielle et primitive
## 1) Équation différentielle
<div class="alert alert-block alert-danger">
Une équation différentielle (E.D.) d’ordre $n$ est une équation :  <br>
• dont l’inconnue est une fonction $y$ de variable $x$, $n$ fois dérivable sur $I \in \mathbb{R}$;<br>
• liant $y$ et certaines de ses dérivées : $y′$, $y′′$, $\ldots$, $y^{(n)}$ et éventuellement $x$.<br>
• Résoudre une équations différentielle sur $I$, c’est déterminer l’ensemble des fonctions $y$  érivables sur I qui vérifient cette équation.
 </div>
 
 
**Remarque :**  
Il existe de nombreuses formes d’équations différentielles :
E.D. d’ordre 1, d’ordre 2, homogène, linéaire, incomplète, à variables séparées $\ldots$  

On se propose de résoudre les E.D. du premier ordre et quelques autres.  

**Exemples :**  
• La plus simple, vue en première : $y′ = y$.  
Les solutions y sont telles que $y(x) = ae^x$ avec $a \in \mathbb{R}$.   En effet $y′(x) = ae^x$.  
La fonction exp est la solution de l’E.D. $y′ = y $ telle que $y(0) = 1$.  

• E.D. du premier ordre :  
$y′ = 2y + 5$, $\quad $ $y′ = 2xy + 4x$,$\quad $  $y′ = y^2$  

• E.D. du second ordre :  
$y′′ + 4y = 0$,$\quad$ $x^2y′′ − 2y + 2x = 0$.  

## 2) Primitive
<div class="alert alert-block alert-danger">  
Soit $f$ est une fonction définie sur un intervalle $I$. <br>
On appelle primitive de $f$ sur $I$, une solution $F$ de l’E.D. $y′ = f (x)$.<br>
    On a alors : $ \forall x \in I \ ,  \ F′(x) = f (x)$  
</div>  

  
**Remarque :**  
Le mot « primitive » fait écho à la fonction dérivée : fonction « initiale » ou « origine » dont la fonction dérivée est connue.  

En physique, lorsqu’on connaît la vitesse v(t), l’équation horaire, x(t), est alors une primitive de la vitesse v.  

**Exemples :**  
1) Soit la fonction $f$ définie sur $\mathbb{R}$ par $f (x) = 2x$.  
$F : x \to x^2$ est une primitive de $f$ sur $\mathbb{R}$ car $F′(x) = 2x$.  

2) Soit la fonction $F$ définie sur $]0;+\infty[$ par $F(x) = x(ln x − 1)$.  
$F$ est dérivable sur $]0;+\infty[$ et $F′(x) = ln x − 1 + x \times \dfrac{1}{x}= ln x$.  

F est donc une primitive de la fonction $ln$ sur $]0;+\infty[$.  
  
<div class="alert alert-block alert-danger">
Soit une fonction $f$ admettant une primitive $F$ sur $I$, alors toute primitive $G$ de $f$ sur $I$ est de la forme : $G = F + k\ , k \in \mathbb{R}$  
</div>


**Démonstration : Par double implication.**   

• Soit $G$ une fonction définie sur $I$ par : $G = F + k\ , k \in \mathbb{R}$.  

$G$ est dérivable sur $I$ par somme de fonctions dérivables et $G′ = F′\overbrace{=}^{ F'=f} f$ .  
G est donc une primitive de f sur I.  

• Réciproquement soit $G$ une primitive de $f$ sur $I$, $G$ est dérivable sur $I$ et :  

$$(G − F)′ = G′ − F′ = f (x) − f (x) = 0$$
La fonction $(G − F)$ est constante, donc :  
$$\exists  k \in \mathbb{R}\ , G − F = k \iff G = F + k$$  

**Exemple :**   
Si $F : x \to x^2$ est une primitive de $f : x \to 2x$ alors $G : x \to x^2 + 3$  est aussi une primitive de $f$ sur $\mathbb{R}$.    

## 3) Primitive vérifiant une condition initiale
<div class="alert alert-block alert-danger">
Soit $f$ une fonction admettant une primitive sur $I$.<br>
Soit $x_0 \in I$ et $y_0 \in \mathbb{R}$, il existe une unique primitive $F$ de $f$ sur $I$ tel que : 
    $$F(x_0) = y_0$$.
</div>  

## 4) Existence de primitives
<div class="alert alert-block alert-danger">
 Toute fonction continue sur $I$ admet des primitives sur $I$. 
 </div>
    
**Remarque :**   
On admet ce théorème. Sa démonstration repose sur le théorème fondamental de l’intégration.  
Ces deux notions ont été liées dès le départ à la fin du XVIIe siècle avec Newton et Leibniz.  

La continuité est une condition suffisante mais non nécessaire pour qu’une fonction admette des primitives. Il existe des fonctions non continues qui admettent des primitives mais cela dépasse le cadre du programme.  

Ce théorème est un théorème d’existence, il ne donne pas l’expression des primitives d’une fonction continue.


Pour pouvoir déterminer les primitives d’une fonction, on répertorie les primitives des fonctions élémentaires par une lecture inverse du tableau des dérivées.

## 5) Résolution par une méthode numérique : méthode d’Euler

<div class="alert alert-block alert-danger">
<b>Théorème: Principe de la méthode d’Euler</b><br> 
Soit f une fonction dérivable sur $I$, d’après l’approximation affine, pour un pas $p$ :
    $$ \forall x, \ x+p \in I, \quad f (x + p) \approx f (x) + p f ′(x)$$
    
Si l’on dispose d’une relation entre f et f ′ par une équation différentielle ainsi que
d’une condition initiale, on peut, en appliquant l’approximation affine de façon itérative, déterminer de proche en proche des valeurs approchée de $f (x)$ sur $I$.
 </div>
 
**Remarque:**  
C’est par cette méthode que l’on a tracé, en première, la fonction
exponentielle définie par $y′ = y$ et $y(0) = 1$.  


 
 

```python
import matplotlib.pyplot as plt

def courbe(a,p):
    x = 0 ; X = [x]
    y = 1 ; Y = [y]
    borne = int(a/p)
    for i in range(borne):
        y= y+ p*y
        x+=p
        X.append(x)
        Y.append(y)
    plt.plot(X,Y)
    plt.show()
    return Y.pop(1000),X,Y
```

```python
courbe(3,0.1)
```

**Exemple :**  
Soit l’équation différentielle :  
$$\begin{cases} y′ + 2xy &=& 1 \\ y(0) &=& 0
\end{cases}$$

```python
import matplotlib.pyplot as plt

def courbe(a,p):
    x = 0 ; X = [x]
    y = 0 ; Y = [y]
    borne = int(a/p)
    for i in range(borne):
        y= y + p*(1-2*x*y)
        x+=p
        X.append(x)
        Y.append(y)
    plt.plot(X,Y)
    plt.show()
    return X,Y
```

```python
courbe(3,0.01)
```
