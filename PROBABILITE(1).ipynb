# I) Loi de probabilité
## 1) Conditions préalables
Il s’agit de construire une structure mathématique qui permette de repérer des
situations dans un domaine où notre intuition nous conduit souvent à la solution sans vraiment avoir conscience de notre démarche.


Dans tout calcul de probabilité, il faut :  

1) Une expérience aléatoire : il s’agit d’un protocole dont on ne peut prévoir l’issue.  

Exemples :  
• Lancer un dé   
• Lancer une pièce de monnaie  
• Distribuer 5 cartes à un joueur avec un jeu de 32 cartes  

2) Repérer toutes les issues possibles de l’expérience :le dénombrement.

Exemples :    
• Il y a 6 issues possibles pour un dé : {1; 2; 3; 4; 5; 6}.  
• Il y a deux issues possibles pour une pièce : face ou pile : {F; P}  
• Il y a 201 376 mains possibles de 5 cartes pour un jeu de 32 cartes  


3) Déterminer ce que l’on souhaite comme issues.  

Exemples :
• Obtenir un nombre pair avec un dé.  
• Obtenir face avec une pièce.  
• Obtenir 2 coeur dans une main de cinq cartes.  


## 2) Définition
<div class="alert alert-block alert-danger">
 On appelle univers d’une expérience aléatoire, l’ensemble de toutes les issues de cette expérience. On appelle cet ensemble : $\Omega$.  
    
Si $e_1, e_2,\ldots, e_n$ sont les issues de cette expérience, on a alors : 
    $$\Omega = \{e_1, e_2, \ldots , e_n\}$$
    
    
</div>


Exemples :  
• L’univers d’un dé : $\Omega = \{1; 2; 3; 4; 5; 6\}$  
• L’univers d’une pièce : $\Omega = \{F; P\}$  
• Parfois nommer toutes les issues est trop long comme l’univers d’une main de
5 cartes avec un jeu de 32 cartes. On se contente alors de compter les éléments
de cet ensemble $\Omega$  

<div class="alert alert-block alert-danger">
    
On appelle cardinal d’un ensemble fini $\Omega$, le nombre d’éléments
qui le compose. On le note $card(\Omega)$.  <br>
    
Si $e_1, e_2, \ldots, e_n$ sont les issues de cette expérience, on a alors : $card(\Omega) = n$
    
</div>


Exemples :  
• L’univers $\Omega$ d’un dé : $card(\Omega) = 6$  
• L’univers $\Omega$ d’une main de cinq cartes : $card(\Omega) = 201 376$  


<div class="alert alert-block alert-danger">
    
On appelle probabilité d’une issue $e_i$, noté $p(e_i)$ le nombre
compris entre 0 et 1 tel que :<br>
    
$$p(e_1) + p(e_2) + \ldots + p(e_n) = 1$$ 
 </div>
 
Définir la loi de probabilité d’une expérience, c’est déterminer les probabilités de tous les éléments de l’ensemble $\Omega$.  

</div>


**Exemples :**  
• Dans une urne qui contient 10 boules indiscernables au toucher, 3 sont vertes
(V), 3 sont bleues (B) et 4 sont jaunes (J), on tire une boule au hasard et on note sa couleur.  

Déterminer la loi de probabilité de cette expérience.  

L’univers de cette expérience est $\Omega = \{V, R, J\}$. Pour déterminer la loi probabilité de cette expérience, il faut calculer les probabilités suivantes :  
$$P(V) = \dfrac{3}{10}= 0, 3 $$
$$P(B) = \dfrac{3}{10}= 0, 3$$  
$$P(J) =\dfrac{4}{10}= 0, 4$$


On regroupe ces résultats dans un tableau représentant la loi de probabilité :  

|$e_i$| V |R |J |
|---|---|---|---|
| $P(e_i)$| 0,3 | 0,3 | 0,4 |



## 3)Loi équirépartie
<div class="alert alert-block alert-danger">
    
Lorsque toutes les issues ont la même probabilité d’apparition, on dit que la loi de probabilité est équirépartie (ou encore que l’on se situe dans
un cas d’équiprobabilité).<br>
Si $\ card(\Omega) = n \quad $ alors on a : $\quad \forall i \in \{1, 2, \ldots , n\} $, $p(e_i) =\dfrac{1}{n}$  
</div>



**Exemples :**   
• Dans un dé bien équilibré chaque face à la même probabilité d’apparition :  
$p(1) = p(2) = p(3) = p(4) = p(5) = p(6) = \dfrac{1}{6}$  

• Si une pièce est bien équilibrée chaque face à la même probabilité d’apparition :  

$p(F) = p(P) =\dfrac{1}{2}$  


# II) Probabilité d’un événement
## 1) Définition
<div class="alert alert-block alert-danger">
 On appelle événement un sous-ensemble de l’univers $\Omega$.<br>
Soit $A$ un événement donné. $p(A)$, sa probabilité, est alors la somme des probabilités des issues qui le composent.
    </div>
    
    
**Exemple :**  

On lance un dé équilibré dont on a calculé la loi de probabilité auparavant.    
Calculer la probabilité de l’événement $A$ : "Obtenir un multiple de 3".  

On a alors : $A = \{3, 6\}$, donc : $p(A) = p(3) + p(6) = \dfrac{1}{6} + \dfrac{1}{6} = \dfrac{2}{6} =\dfrac{1}{3}$  


**Remarque :**  

• Si l’ensemble $A$ est réduit à l’ensemble vide ?, il n’est composé d’aucune issue.  
On l’appelle alors l’événement impossible et $p(\emptyset) = 0$  

• Si l’ensemble $A$ représente tout l’univers $\Omega$, il est composé de toutes les issues.  
On l’appelle alors l’événement certain et $p(\omega) = 1$  
  
## 2) Événement d’une loi équirépartie
<div class="alert alert-block alert-danger">
Dans une loi équirépartie, la probabilité d’un événement $A$ vérifie
l’égalité suivante : <br>
    $$p(A) =\dfrac{\mbox{nbre d’issues de l’événement A}}{\mbox{nbre d’issues total}}=\dfrac{card(A)}{card(\Omega)}$$  
  </div>
  
  
**Exemple :**  

Comme pour l'exemple précédent. 

# III) Opération sur les événements
## 1) Événement contraire
<div class="alert alert-block alert-danger">
 On appelle événement contraire d’un événement $A$, l’événement
noté $A$ composé des éléments de $\Omega$ qui ne sont pas dans $A$.<br>
    $$x \in \overline{A} \iff x \in \Omega \quad  \mbox{et} \quad  x \notin A$$
</div>

**Remarque :**  
$\overline{A}$ se prononce "A barre". On peut visualiser $\overline{A}$ par le diagramme de Venn suivant :  
 



**Exemples :**   
• On lance un dé :  
$A$ : faire au moins 3 $\quad $ d’où $\overline{A}$ : faire au plus 2  

• On tire deux cartes dans un jeu de 32  
$B$ : Obtenir au moins un coeur$\quad $  d’où $\overline{B}$ : Obtenir aucun coeur  
  
## 2) Intersection de deux événements
<div class="alert alert-block alert-danger">
On appelle intersection de deux événements $A$ et $B$, l’événement
noté : <br>
    $A \cap B$ composé des éléments de $\Omega$ qui appartiennent à $A$ et à $B$.<br>
    
$$x \in A \cap B \iff x \in A \quad \mbox{et} \quad  x \in B$$
</div>


**Remarque :**  
$A \cap B$ se prononce "A inter B". 
On peut visualiser $A \cap B$ par le diagramme de Venn suivant :    






Lorsque l’événement $A$ est inclus dans l’événement $B$, on a alors :  
$$A \subset B \implies A \cap B = A$$  

Lorsque les événement $A$ et $B$ sont *incompatibles*, leur intersection est vide.  

$A$ et $B$ sont incompatibles $\iff A \cap B = \emptyset$  

C’est le cas des événements $A$ et $\overline{A}$ : $A \cap \overline{A} = \emptyset$    

**Exemple :**  
On tire une carte dans un jeu de 32 cartes, soient les événements suivants :  
$A$ : Obtenir un coeur  
$B$ : Obtenir une dame  

on en déduit: $A \cap B$ : Obtenir la dame de coeur    


## 3) Union de deux ensembles
<div class="alert alert-block alert-danger">
On appelle union de deux événements $A$ et $B$, l’événement
noté : $A\cup B$ composé des éléments de $\Omega$ qui appartiennent à $A$ ou à $B$ (éventuellement aux deux, le « ou » étant non exclusif).  
    $$x \in A \cup B \iff x \in A \quad \mbox{ou} \quad x \in B$$  
</div>


**Remarque :**  
$A \cup B$ se prononce "A union B".  
On peut visualiser $A \cup B$ par le diagramme de Venn suivant :  

  



Lorsque l’événement $A$ est inclus dans l’événement $B$, on a alors :  
$$A \subset B \implies A \cup B = B$$  

L’union des événements $A$ et $\overline{A}$ donne l’ensemble $\Omega$ :   $A \cup \overline{A} = \Omega$   

  
**Exemple:**  

On tire une carte dans un jeu de 32  
$A$ : Obtenir un coeur  
$B$ : Obtenir une dame  
On en déduit que $A \cup B$ : Obtenir une dame ou un coeur    

## 4) Utilisation de ces opérations dans une loi de probabilité
<div class="alert alert-block alert-danger">
Si $A$ et $B$ sont deux événements, alors on a :<br>
    $$p(A) = 1 − P(\overline{A})$$
    $$p(A \cup B) = P(A) + P(B) − P(A \cap B)$$  
    
Si $A$ et $B$ sont incompatibles, alors : $P(A \cup B) = P(A) + P(B)$
</div>


**Exemple :**   
$A$ et $B$ sont deux événements d’une même expérience aléatoire.  
Calculer $p(B)$ puis $p(\overline{B})$.  
On donne : $p(A) = 0, 3\quad $  ; $ p(A \cup B) = 0, 7\quad $ et $\quad p(A \cap B) = 0, 2$   


# IV) Intervalle de fluctuation
## 1) Loi des grands nombres
<div class="alert alert-block alert-danger">
Pour une expérience donnée, dans un modèle défini par une loi de probabilité, les distributions des fréquences calculées sur un échantillon de taille n, se rapproche de la loi de probabilité lorsque n devient grand.  
</div>



**Remarque :**  
En statistique, un échantillon de taille n est obtenu par répétition
indépendantes d’une même expérience aléatoire. Par exemple, on reproduit 120
fois une expériences consistant à lancer 100 fois une pièce de monnaie. On a ainsi
un échantillon de taille 120.  

Ce théorème nous affirme que si la taille de l’échantillon est suffisamment grande,
on observera une fréquence d’apparition du côté "face" de la pièce proche de 0,5.
Si ce n’est pas le cas alors on pourra affirmer que la pièce n’est pas équilibrée.  


## 2) Intervalle de fluctuation
<div class="alert alert-block alert-danger">
Soit une expérience aléatoire où la probabilité d’un événement $A$
est $p$. On reproduit cette expérience $n$ fois et l’on détermine la fréquence observée $f_{obs}$ d’apparition de l’événement $A$.<br>
    
Si $p \in [0, 2 ; 0, 8]$  et si $n \gt 25$ , alors, dans environ 95 % des cas, $f_{obs}$ est compris dans l’intervalle $I$ appelé intervalle de fluctuation au seuil de 95 % (ou au risque de 5 %).<br>
    $$p \in [0, 2 ; 0, 8]\quad \mbox{ et} \quad  n \gt 25 \quad \mbox{ alors}  \quad I = \bigg [ p −\dfrac{1}{\sqrt{n}}; p +\dfrac{1}{\sqrt{n}} \bigg ]$$  
</div>

   
**Exemple:** 
    
Pour savoir si une pièce est équilibrée, on la lance 200 fois. On a obtenu 120 fois "pile". Est-il vraisemblable que la pièce soit équilibrée ?  

On calcule l’intervalle de confiance à 95 % :    
$$I =\bigg [ 0,5 −\dfrac{1}{\sqrt{200}}; 0,5 +\dfrac{1}{\sqrt{200}} \bigg ]$$

$$ I \approx  [0, 43 ; 0, 57]$$


or $f_{obs} = \dfrac{120}{200}= 0, 6$. donc $f_{obs} \notin I$.  

Il y a de forte chance que la pièce ne soit pas équilibrée.
