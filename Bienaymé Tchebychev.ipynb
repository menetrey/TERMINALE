## 3) Ingalit de concentration
<div class="alert alert-block alert-danger">    
<b>Thorme :</b><br> 
Soit $(X_1, X_2, \ldots , X_n)$ un chantillon de variables alatoires desprance $\mu$ et de variance $V$ et $M_n$ la variable alatoire moyenne de cet chantillon.  
    $$\forall \delta \in ]0;+\infty[ ,P(|M_n -\mu|\ge \delta)\le \frac{V}{n \delta^2} $$

 </div>   
    
    
**Remarque :**  
    On rappelle que $M_n = \dfrac{X_1 + X_2 + \ldots+ X_n}{n}$

**Exemple :** On prend un d ttradrique bien quilibr dont on a dtermin
lesprance $\mu = 2, 5$ et la variance $V = 1, 25$.  
Combien de lancers du d ttradrique doit-on faire pour sassurer au
seuil de 95 % que la moyenne des rsultats des lancers est dans lintervalle
$]2, 45 ; 2, 55[$.  
Le rayon de lintervalle $]2, 45 ; 2, 55[$ est $\delta = 2, 55  2, 5 = 0, 05$.  
    
Soit $M_n$ la variable alatoire moyenne dun chantillon de n lancers. On a alors :  
    
$$ \begin{array} \\ 
P(|M_n  \mu| \ge \delta) \le \dfrac{V}{n \delta^2} &\iff&  P(|M_n  2,5| \ge 0, 05) \le \dfrac{1,25}{0,05^2 n } \\ 
&\iff& P(|M_n  2,5| \ge 0, 05) \le \dfrac{500}{n} \\
&\iff& \dfrac{500}{n} \le  0, 05 \\ 
&\iff&  n \ge \dfrac{500}{0,05} \\
&\iff& n \ge 10000
\end{array}$$ 
    
Il faut faire au moins 10 000 lancers pour sassurer que la moyenne des rsultats est  moins de 5 % de lesprance au seuil de 95 %.
